## 1. 核心思想

Smale 马蹄映射是一种把二维几何映射转化为符号动力系统的机制：

$$
\text{拉伸与折叠}
\Longrightarrow
\text{不变 Cantor 集}
\Longrightarrow
\text{符号序列上的移位}.
$$

它的重要性体现在四个方面：

- 存在无穷多个鞍型周期轨道；
- 存在不可数多个非周期轨道；
- 存在一条稠密轨道；
- 不变动力学是双曲的，因此在小扰动下具有稳健性。

本讲的概念主线为

$$
\text{拉伸-压缩-折叠}
\Longrightarrow
\text{符号编码}
\Longrightarrow
\text{移位动力学}
\Longrightarrow
\text{稳健的复杂动力学}.
$$

最后将说明横截同宿交点为什么会自然地产生马蹄结构。


## 2. 标准马蹄映射的构造

设

$$
\mathcal Y=[0,1]\times[0,1]
$$

为单位正方形。映射 $g$ 的几何构造分为三步：

1. 沿水平方向以因子 $0<\lambda<1/2$ 压缩 $\mathcal Y$；
2. 沿竖直方向以因子 $\mu>2$ 拉伸；
3. 将所得长矩形折叠后重新放置，使它穿过 $\mathcal Y$。

返回正方形的像由两个互不相交的竖直条带组成：

$$
g(\mathcal Y)\cap\mathcal Y
=
\mathcal V_0\cup\mathcal V_1.
$$

它们的原像是两个水平条带：

$$
\mathcal Y\cap g^{-1}(\mathcal Y)
=
\mathcal H_0\cup\mathcal H_1,
\qquad
g(\mathcal H_i)=\mathcal V_i.
$$

弯折的中间部分位于 $\mathcal Y$ 外，因此进入该部分的点会离开正方形。在两个水平条带上，导数为

$$
Dg=
\begin{pmatrix}
\pm\lambda&0\\
0&\pm\mu
\end{pmatrix}.
\tag{13.1}
$$

更具体地，

$$
Dg|_{\mathcal H_0}
=
\begin{pmatrix}\lambda&0\\0&\mu\end{pmatrix},
\qquad
Dg|_{\mathcal H_1}
=
\begin{pmatrix}-\lambda&0\\0&-\mu\end{pmatrix}.
$$

负号来自折叠造成的方向反转。双曲性取决于 $\lambda$ 和 $\mu$ 的模，而不是符号。

> **图示说明。** 原书图 32 画的是 $g^{-1}$ 而不是 $g$，所以图中可见的水平拉伸和竖直压缩与正向映射的描述恰好相反。

### $\mathcal Y$、$g^{-1}(\mathcal Y)$ 与条带的含义

$\mathcal Y$ 是观察动力学的区域。原像

$$
g^{-1}(\mathcal Y)=\{x:g(x)\in\mathcal Y\}
$$

表示下一次正向映射后落入 $\mathcal Y$ 的所有点。因此

$$
\mathcal Y\cap g^{-1}(\mathcal Y)
=
\{x\in\mathcal Y:g(x)\in\mathcal Y\}
=
\mathcal H_0\cup\mathcal H_1
$$

表示当前在正方形内且下一步仍在正方形内的点。另一方面，$g(\mathcal Y)\cap\mathcal Y=\mathcal V_0\cup\mathcal V_1$ 表示当前在正方形内、且前一步来自正方形的点。

水平条带经过 $g$ 后变成竖直条带，是因为水平宽度乘以 $\lambda<1/2$，竖直高度乘以 $\mu>2$：

$$
g(\mathcal H_i)=\mathcal V_i,
\qquad
g^{-1}(\mathcal V_i)=\mathcal H_i.
$$

并不是一个竖直条带有两个水平原像；而是两个返回的竖直分支 $\mathcal V_0,\mathcal V_1$ 各自对应一个水平原像 $\mathcal H_0,\mathcal H_1$。


## 3. 最大不变集

大多数点最终都会离开正方形。马蹄动力学只考虑在所有正向和反向迭代下始终留在 $\mathcal Y$ 中的点：

$$
\mathcal X
=
\left\{
x\in\mathcal Y:
g^k(x)\in\mathcal Y
\text{ 对每个 }k\in\mathbb Z
\right\}.
$$

等价地，

$$
\mathcal X
=
\bigcap_{k\in\mathbb Z}g^k(\mathcal Y).
$$

有限时间的正向存留集

$$
\mathcal Y_n^+
=
\bigcap_{j=0}^{n}g^{-j}(\mathcal Y)
$$

由 $2^n$ 个互不相交的水平条带组成，它们对应未来 $n$ 次迭代仍留在 $\mathcal Y$ 的点。类似地，

$$
\mathcal Y_n^-
=
\bigcap_{j=0}^{n}g^j(\mathcal Y)
$$

由 $2^n$ 个竖直条带组成，对应过去 $n$ 次迭代来自 $\mathcal Y$ 的点。

每增加一个存留条件，条带数目加倍而厚度减小。极限集合 $\mathcal X$ 是：

- 紧集；
- 完美集，即没有孤立点；
- 全不连通；
- 不可数。

因此 $\mathcal X$ 在拓扑上是 Cantor 集。几何上，它具有稳定 Cantor 集与不稳定 Cantor 集的乘积型结构。未来存留条件产生越来越薄的水平条带，过去存留条件产生越来越窄的竖直条带，二者的嵌套交集最终收缩为单个点。


## 4. 符号空间

令 $\Sigma_2$ 为所有双向无穷二符号序列构成的空间：

$$
\Sigma_2
=
\left\{
a=(a_j)_{j\in\mathbb Z}:
a_j\in\{0,1\}
\right\}.
$$

典型元素写作

$$
a=\ldots a_{-2}a_{-1}.a_0a_1a_2\ldots.
$$

在 $\Sigma_2$ 上定义度量

$$
d(a,b)
=
\sum_{j=-\infty}^{\infty}
|a_j-b_j|\,2^{-|j|}.
$$

两个序列越靠近，意味着它们在 $j=0$ 附近越长的中心符号块上一致。

左移映射定义为

$$
\sigma:\Sigma_2\to\Sigma_2,
\qquad
(\sigma(a))_j=a_{j+1}.
$$

即

$$
\sigma(\ldots a_{-1}.a_0a_1a_2\ldots)
=
\ldots a_{-1}a_0.a_1a_2\ldots.
$$

必须使用双向序列，因为马蹄映射在 $\mathcal X$ 上可逆，过去和未来的行程都属于动力学信息。


## 5. 马蹄轨道的编码

对 $x\in\mathcal X$，定义其行程

$$
\phi(x)=(a_j)_{j\in\mathbb Z}
$$

满足

$$
a_j=i
\quad\Longleftrightarrow\quad
g^j(x)\in\mathcal H_i,
\qquad i\in\{0,1\}.
$$

$a_j$ 记录第 $j$ 次迭代所在的水平条带。对点施加一次 $g$，等价于对符号序列左移一次：

$$
\phi(g(x))_j=a_{j+1},
\qquad
\phi\circ g=\sigma\circ\phi.
$$

符号与几何的对应关系如下：

| 符号信息 | 几何信息 |
|---|---|
| $a_0$ | 当前点所在的水平条带 |
| $a_0,a_1,a_2,\ldots$ | 未来行程 |
| $\ldots,a_{-2},a_{-1}$ | 过去行程 |
| 未来符号 | 确定竖直的不稳定坐标 |
| 过去符号 | 确定水平的稳定坐标 |
| 周期序列 | 周期轨道 |
| 很长的中心符号块一致 | 两点属于同一个很小的柱矩形 |
| 左移 $\sigma$ | 施加一次 $g$ |

具有相同未来行程的点只在稳定方向上不同，因此正向迭代时彼此接近；具有相同过去行程的点则通过不稳定方向联系起来。


## 6. 马蹄映射的符号动力学

### 定理 13.1

映射

$$
g:\mathcal X\to\mathcal X
$$

与移位映射

$$
\sigma:\Sigma_2\to\Sigma_2
$$

拓扑共轭。即存在同胚 $\phi:\mathcal X\to\Sigma_2$，使得

$$
\phi\circ g=\sigma\circ\phi.
$$

### 证明

对有限中心符号块

$$
b=b_{-m}\ldots b_{-1}.b_0\ldots b_n,
$$

定义柱矩形

$$
R(b)
=
\left\{
x\in\mathcal Y:
g^j(x)\in\mathcal H_{b_j}
\text{，其中 }-m\le j\le n
\right\}.
$$

未来条件限制竖直坐标，过去条件限制水平坐标。该矩形满足

$$
\operatorname{width}R(b)=\lambda^m,
\qquad
\operatorname{height}R(b)=\mu^{-(n+1)}.
$$

因此

$$
\operatorname{diam}R(b)
\le
\sqrt{\lambda^{2m}+\mu^{-2(n+1)}}
\longrightarrow0
$$

当 $m,n\to\infty$。

#### 单射性

若两个点具有相同的完整双向行程，则它们对所有 $m,n$ 都属于同一个柱矩形。由于这些矩形的直径趋于零，两点必须相同。

#### 满射性

给定任意 $a\in\Sigma_2$，矩形

$$
R(a_{-m}\ldots a_n)
$$

非空、紧致且彼此嵌套。因此

$$
\bigcap_{m,n\ge0}R(a_{-m}\ldots a_n)\neq\varnothing.
$$

又因为直径趋于零，该交集只包含一个点，所以每个符号序列都由 $\mathcal X$ 中唯一一点实现。

#### 连续性

若两个行程在足够长的中心符号块上一致，则对应点位于同一个任意小的柱矩形中。结合 $\mathcal X$ 的紧致性与 $\Sigma_2$ 的 Hausdorff 性，可知连续双射 $\phi$ 是同胚。

### 说明：嵌套柱集与连续性

加入一个未来符号等于增加一个限制，因此

$$
R(b_0\ldots b_{n+1})
=
R(b_0\ldots b_n)
\cap
g^{-(n+1)}(\mathcal H_{b_{n+1}})
\subseteq
R(b_0\ldots b_n).
$$

新的集合是原水平条带内部更薄的子条带。严格地说，需要累计迭代 $g^{n+1}$ 才能把 $R(b_0\ldots b_n)$ 拉成竖直横穿 $\mathcal Y$ 的条带；正是这一横穿性质保证下一个符号无论取 $0$ 还是 $1$ 都可以实现。

$\operatorname{diam}R(b)\to0$ 直接证明编码点的唯一性以及解码映射 $\phi^{-1}$ 的连续性。$\phi$ 本身的连续性还需要两个水平条带彼此分离：对有限多个时刻 $|j|\le N$，由 $g^j$ 的连续性可知，充分小的初值扰动不会改变这些时刻的条带编号。等价地，有限柱集在 $\phi$ 下的原像是 $\mathcal X$ 中的开集。


## 7. 动力学推论

拓扑共轭把二进制序列的性质转化为马蹄映射的几何性质。

### 周期轨道

每个周期序列对应一个周期点。例如

$$
\ldots000.000\ldots,
\qquad
\ldots111.111\ldots
$$

对应不动点，而

$$
\ldots0101.0101\ldots
$$

对应二周期轨道。对固定周期 $p$，只有有限个二进制周期序列，因此全部周期点构成可数无限集。

周期为 $p$ 的轨道具有乘子模长

$$
\lambda^p<1,
\qquad
\mu^p>1.
$$

所以每个周期轨道都是鞍型。周期点还是稠密的，因为任意有限符号块都可以周期重复。

### 非周期轨道

$\Sigma_2$ 不可数，而周期序列只有可数多个，因此 $\mathcal X$ 中存在不可数多个非周期点。

### 稠密轨道

构造一个正向尾部包含每个有限二进制单词的序列。对它不断移位，可以把任意给定单词移动到中心位置，因此其移位轨道进入每个柱集，并在 $\Sigma_2$ 中稠密。对应的马蹄轨道在 $\mathcal X$ 中稠密，但这并不表示每条轨道都稠密。

### 双曲性

稳定与不稳定子空间为

$$
E^s=\operatorname{span}(e_1),
\qquad
E^u=\operatorname{span}(e_2).
$$

对 $v^s\in E^s$ 和 $v^u\in E^u$，

$$
\|Dg^n v^s\|
=
\lambda^n\|v^s\|
\longrightarrow0,
$$

$$
\|Dg^{-n}v^u\|
=
\mu^{-n}\|v^u\|
\longrightarrow0.
$$

因此 $\mathcal X$ 是双曲不变集。由双曲集的持续性，充分小的 $C^1$ 扰动仍具有邻近的、符号动力学拓扑共轭的不变集。

### 说明：序列点、轨道与 T1-T3

一个完整双向序列 $a\in\Sigma_2$ 是符号相空间中的一个点，而不是整条轨道。它的完整轨道是

$$
\mathcal O_\sigma(a)=\{\sigma^n(a):n\in\mathbb Z\}.
$$

单个序列记录马蹄系统中一个点的完整行程，而该序列的所有移位对应马蹄轨道上的所有点。周期序列可数，是因为对每个周期 $p$ 至多只有 $2^p$ 个单词。非周期序列不可数；又由于每条轨道至多包含可数个点，因此不同的非周期轨道本身也必有不可数多条。包含所有有限二进制单词的序列则给出稠密轨道。


## 8. 广义马蹄映射

标准构造中的直条带对于实际应用过于特殊。由微分方程产生的返回映射通常给出弯曲条带。

### 弯曲条带

竖直曲线写成 $x=v(y)$，并满足

$$
0\le v(y)\le1,
\qquad
|v(y_l)-v(y_r)|
\le
\rho|y_l-y_r|,
\qquad
0<\rho<1.
$$

水平曲线写成 $y=h(x)$，并满足

$$
0\le h(x)\le1,
\qquad
|h(x_l)-h(x_r)|
\le
\rho|x_l-x_r|.
$$

两条不相交的竖直曲线 $v_1(y)<v_2(y)$ 围成竖直条带

$$
\mathcal V
=
\left\{
(x,y)\in\mathbb R^2:
v_1(y)\le x\le v_2(y),
\ 0\le y\le1
\right\}.
$$

类似地，$h_1(x)<h_2(x)$ 围成水平条带

$$
\mathcal H
=
\left\{
(x,y)\in\mathbb R^2:
h_1(x)\le y\le h_2(x),
\ 0\le x\le1
\right\}.
$$

### 锥条件

设有 $N$ 个互不相交的水平条带 $\mathcal H_i$ 与竖直条带 $\mathcal V_i$，满足

$$
g(\mathcal H_i)=\mathcal V_i,
\qquad
i=0,\ldots,N-1.
$$

为避免与标准马蹄中的竖直扩张因子 $\mu>2$ 混淆，使用 $0<\theta<1/2$ 表示锥宽。在基点 $p$ 处，对切向量 $\xi=(\xi_1,\xi_2)$ 定义

$$
K_p^u
=
\{\xi:|\xi_1|<\theta|\xi_2|\},
\qquad
K_p^s
=
\{\xi:|\xi_2|<\theta|\xi_1|\}.
$$

锥不变性要求

$$
Dg_p(K_p^u)\subset K_{g(p)}^u,
\qquad
Dg_p^{-1}(K_p^s)\subset K_{g^{-1}(p)}^s.
$$

一致扩张条件为

$$
|(Dg_p\xi)_2|
\ge
\frac1\theta|\xi_2|,
\qquad
\xi\in K_p^u,
$$

$$
|(Dg_p^{-1}\xi)_1|
\ge
\frac1\theta|\xi_1|,
\qquad
\xi\in K_p^s.
$$

原书还假设行列式满足

$$
|\det Dg|,
\ |\det Dg^{-1}|
\le
\frac12\theta^{-2}.
$$

### 为什么需要 A1、A2 与行列式条件

$\xi=(\xi_1,\xi_2)$ 表示切向量的分量；$(Dg_p\xi)_2$ 的下标 $2$ 表示映射后向量的第二个，即竖直分量。A1 取代了标准线性情形中坐标轴的严格不变性：近似竖直的向量经过 $Dg$ 后仍近似竖直，近似水平的向量经过 $Dg^{-1}$ 后仍近似水平。A2 则给出定量扩张：竖直分量正向至少扩张 $1/\theta$ 倍，而水平分量逆向扩张，等价于正向压缩。

对标准对角马蹄，单独的锥不变性对任意 $\theta>0$ 都成立；A2 还要求

$$
\mu\ge\frac1\theta,
\qquad
\frac1\lambda\ge\frac1\theta,
$$

即 $\theta\ge\max\{\lambda,1/\mu\}$。对非线性映射，导数中的非对角项会混合方向，因此连 A1 也只对适当选择的锥成立。

行列式是局部面积放大因子，但这里更具体地出现在斜率变换的导数中。若

$$
Dg_p=
\begin{pmatrix}a&b\\c&d\end{pmatrix},
\qquad
\widehat s=\frac{as+b}{cs+d},
$$

则

$$
\frac{d\widehat s}{ds}
=
\frac{\det Dg_p}{(cs+d)^2}.
$$

由 A2，$|cs+d|\ge1/\theta$，所以

$$
\left|\frac{d\widehat s}{ds}\right|
\le
|\det Dg_p|\theta^2
\le
\frac12.
$$

因此图变换对斜率是压缩的。常数 $1/2$ 并不特殊，任意严格小于 $1$ 的一致常数都可以。

### 定理 13.4：广义马蹄映射

在条带、锥不变性、扩张与行列式条件下，$g$ 存在双曲不变集 $\mathcal X$，并且 $g|_{\mathcal X}$ 与 $N$ 符号全移位

$$
\Sigma_N
=
\{0,1,\ldots,N-1\}^{\mathbb Z}
$$

拓扑共轭。证明沿用标准马蹄中的嵌套条带论证。


## 9. $\lambda$-引理

$\lambda$-引理又称倾斜引理，它说明横截圆盘如何逐渐与不稳定流形对齐。

设 $0$ 是微分同胚 $g$ 的双曲不动点，并且

$$
\dim W^u(0)=k,
\qquad
\dim W^s(0)=l.
$$

设 $k$ 维 $C^1$ 子流形 $\mathcal M$ 在 $q$ 处与 $W^s(0)$ 横截相交。

### 引理 13.6

对任意 $r>0$ 和 $\delta>0$，存在整数 $n_0$、$k$ 维圆盘 $D\subset\mathcal M$ 以及 $x\in W^u(0)$，使得对所有 $n\ge n_0$，

$$
d_1\left(
B(x,r)\cap W^u(0),
g^n(D)
\right)
<\delta.
$$

$d_1$ 表示 $C^1$ 距离：不仅点集位置接近，切空间也接近。

### 几何意义

横截稳定流形的圆盘先沿稳定方向被吸向不动点，同时沿不稳定方向被拉长，最后成为几乎与不稳定流形相切的长圆盘。这不是收敛到一个点，而是流形片段及其切方向的接近。

### 证明机制

使用局部稳定-不稳定坐标 $x=(x_s,x_u)$，写成

$$
g(x_s,x_u)
=
\left(
A^s x_s+\varphi_s(x),
A^u x_u+\varphi_u(x)
\right),
$$

其中

$$
\|A^s\|\le a<1,
\qquad
\|(A^u)^{-1}\|\le a<1,
\qquad
\varphi_s(x),\varphi_u(x)=o(\|x\|).
$$

对切向量 $v_n=(v_n^s,v_n^u)$，定义相对于不稳定方向的斜率

$$
\Lambda_n
=
\frac{\|v_n^s\|}{\|v_n^u\|}.
$$

在线性化情形中，

$$
\Lambda_{n+1}\le a^2\Lambda_n,
\qquad
\|v_{n+1}^u\|\ge a^{-1}\|v_n^u\|.
$$

因此 $\Lambda_n\to0$，同时不稳定方向的尺度增长。非线性项可通过缩小邻域并使用锥估计来控制。

### 两个估计分别证明什么

比值 $\Lambda_n=\|v_n^s\|/\|v_n^u\|$ 控制方向：$\Lambda_n\to0$ 表示切空间逐渐与 $E^u$ 对齐。$\|v_n^u\|$ 的增长控制尺度：充分小的初始圆盘可以被拉长到给定的不稳定方向长度。仅有其中一个估计，都不足以得到 $C^1$ 接近。

对完全分块对角的线性映射，斜率每次改进因子为 $a^2$。加入非线性交叉项后，稳定与不稳定分量会发生混合，通常只能保留较弱的估计，如 $\Lambda_{n+1}\le a\Lambda_n$；所谓损失一个 $a$，就是用这部分余量吸收高阶项。

原书中的计算是证明框架。要从切向量增长推出与半径为 $r$ 的不稳定圆盘接近，还需要在整个圆盘上一致控制导数、对导数估计积分，并结合前面已经假定的 $C^0$ 控制。


## 10. Smale-Birkhoff 同宿定理

设 $0$ 是微分同胚

$$
g:\mathbb R^d\to\mathbb R^d
$$

的双曲鞍型不动点。若 $q\neq0$ 满足

$$
q\in W^s(0)\cap W^u(0),
$$

则 $q$ 是同宿点。交点在下式成立时称为横截的：

$$
T_qW^s(0)+T_qW^u(0)
=
T_q\mathbb R^d.
$$

### 定理 13.5

若 $q$ 是横截同宿点，则存在双曲不变集 $\Lambda$ 与充分大的正整数 $\kappa$，使得

$$
g^\kappa|_\Lambda
$$

与 Smale 马蹄映射、从而与双符号全移位拓扑共轭。核心含义是

$$
\boxed{
\text{横截同宿交点}
\Longrightarrow
\text{马蹄动力学}
}.
$$

### 证明概要

1. 选择局部坐标，使 $W^s(0)$ 与 $W^u(0)$ 为坐标轴。
2. 由于 $q$ 是同宿点，$g^n(q)\to0$ 且 $g^{-n}(q)\to0$。
3. 在稳定流形附近、但与其分离的位置选择一个水平细矩形 $\mathcal R$。
4. 由 $\lambda$-引理，充分大的迭代把 $\mathcal R$ 变为近似平行于 $W^u(0)$ 的细长条带。
5. 全局同宿运动把该条带带回并横穿 $\mathcal R$，产生马蹄条带几何。
6. 增大迭代次数，使稳定压缩和不稳定扩张满足广义马蹄的锥条件。
7. 应用广义马蹄定理得到与二符号移位共轭的不变集。

### 拉伸、弯折与分支分别来自哪里

记 $q_j=g^j(q)$。当 $n\to+\infty$ 时，$q_n$ 沿 $W^s(0)$ 趋于 $0$，而 $q_{-n}$ 沿 $W^u(0)$ 趋于 $0$。取 $n_s,n_u>0$ 并令 $k=n_s+n_u$，则有精确的轨道对应

$$
g^k(q_{-n_u})=q_{n_s},
\qquad
g^k(q_{-n_u+1})=q_{n_s+1}.
$$

这些标记点只确定返回时间和返回位置；点本身没有长度，不能被拉伸，也不能仅凭两个点确定整条返回带的形状。

三个因素承担不同作用：

- 局部双曲性与 $\lambda$-引理使横截矩形变得细长，并在 $C^1$ 意义下与 $W^u(0)$ 对齐；
- 同宿轨道的全局部分把细带带出局部线性邻域，在外部使其弯曲，再送回 $W^s(0)$ 附近；
- 横截性保证返回条带真正横穿参考矩形，而不是与其相切。

因此，$\lambda$-引理只给出拉伸和方向对齐，并不产生弯折；弯折来自全局同宿返回。对靠近 $W^u$ 的竖直细矩形，两条长边跟随邻近的全局不稳定曲线，成为返回带的两条弯曲长边；两条短边是横截方向的端盖。返回到 $W^s$ 附近时，该细带近似竖直地穿过水平参考矩形。

要得到完整的二符号马蹄，需要选择充分大的迭代和适当的子矩形，使返回集具有两个互不相交的横穿分支 $V_0,V_1$，其水平原像为 $H_0,H_1$，并满足

$$
g^k(H_i)=V_i.
$$

图 34 只是示意同宿返回与边界方向，并没有单独画清最终二分支马尔可夫分割的每个连通分支。分离分支的存在与锥估计，是 proof sketch 中被压缩到几何验证里的部分。

定理结论只涉及某个迭代 $g^\kappa$ 在特定不变子集上的动力学，而不是整个原映射处处都是马蹄。

### 为什么流需要 Poincare 映射

对连续时间流，在非平衡同宿点处，向量场方向同时属于 $T_qW^s$ 与 $T_qW^u$，所以二者不能在整个相空间中横截。Poincare 截面去除了共同的流方向，因此其返回映射可以具有横截同宿点。


## 11. 面包师映射

面包师映射给出了符号移位的显式模型：

$$
g:[0,1)^2\to[0,1)^2,
$$

$$
g(x,y)
=
\begin{cases}
\left(2x,\dfrac12y\right),
&0\le x<\dfrac12,\\[4pt]
\left(2x-1,\dfrac12y+\dfrac12\right),
&\dfrac12\le x<1.
\end{cases}
$$

把坐标写成二进制展开：

$$
x=0.x_1x_2x_3\ldots,
\qquad
y=0.y_1y_2y_3\ldots.
$$

则

$$
g(x,y)
=
\left(
0.x_2x_3x_4\ldots,
0.x_1y_1y_2y_3\ldots
\right).
$$

也就是说，$x$ 的第一个二进制位从 $x$ 中移除，并插入 $y$ 的首位。把这些数位排列成双向序列后，这一操作就是移位。

对二进制有理数需要固定一致的展开约定，因为它们存在两种二进制表示。面包师映射使符号机制完全显式：拉伸读出下一个符号，堆叠则把过去的符号存入第二个坐标。


## 12. 常见误区

1. **混淆两类条带。** $g(\mathcal Y)\cap\mathcal Y$ 是竖直条带，而 $\mathcal Y\cap g^{-1}(\mathcal Y)$ 是水平条带。

2. **忘记图 32 画的是 $g^{-1}$。** 图中可见的拉伸方向与正向映射相反。

3. **把整个正方形当成不变集。** 只有 $\mathcal X$ 中的点对所有 $k\in\mathbb Z$ 都留在正方形内。

4. **使用单向序列。** 可逆马蹄必须同时编码过去与未来，所以需要双向序列。

5. **认为有限符号块确定一个点。** 有限块确定一个矩形，完整双向序列才确定一个点。

6. **颠倒稳定与不稳定方向。** 对 $g$ 而言，水平方向稳定，竖直方向不稳定。

7. **让折叠产生的负号掩盖双曲性。** 压缩和扩张取决于特征值的模。

8. **认为每条轨道都稠密。** 定理只断言至少存在一条稠密轨道。

9. **认为 Cantor 集可数。** 它全不连通，但不可数。

10. **把锥中的变量当成基点坐标。** 锥条件作用于基点处的切向量。

11. **误读 $\lambda$-引理。** 它给出流形片段的 $C^1$ 接近，而不是点态收敛到单个点。

12. **夸大 Smale-Birkhoff 的结论。** 马蹄动力学出现在某个迭代限制于特定不变子集之后。

## 14. 总结

本讲的逻辑结构是

$$
\boxed{
\begin{aligned}
&\text{拉伸、压缩与折叠}\\
&\qquad\Longrightarrow\text{不变 Cantor 集}\\
&\qquad\Longrightarrow\text{符号编码}\\
&\qquad\Longrightarrow\text{全移位动力学}\\
&\qquad\Longrightarrow\text{周期与非周期行为}\\
&\qquad\Longrightarrow\text{稳健的双曲性}.
\end{aligned}
}
$$

标准马蹄通过共轭关系

$$
\phi\circ g=\sigma\circ\phi
$$

把几何动力学转化为符号动力学。Smale-Birkhoff 定理给出基本几何判据：

$$
\boxed{
\text{横截同宿点}
\Longrightarrow
\text{嵌入的 Smale 马蹄}
}.
$$

因此，一个横截同宿交点就会迫使系统出现无穷多个周期轨道、不可数多个非周期轨道、稠密轨道行为以及稳健的双曲不变集。
