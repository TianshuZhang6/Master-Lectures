## 21.0 保测系统（Measure-Preserving System）

设 $(X,\mathcal F,\mu)$ 为概率空间（probability space），$g:X\to X$ 为可测映射（measurable map）。若对每个 $A\in\mathcal F$ 都有

$$
\mu(g^{-1}(A))=\mu(A),
$$

则称 $g$ 为保测映射（measure-preserving map）。因此，对任意 $h\in L^1$，

$$
\int_X h\circ g\,d\mu=\int_Xh\,d\mu.
$$

对示性函数（indicator function），这个结论直接来自

$$
\int_X\mathbf 1_A\circ g\,d\mu
=\mu(g^{-1}(A))
=\mu(A)
=\int_X\mathbf 1_A\,d\mu.
$$

再由线性性和逼近，将其推广到所有可积函数。每个迭代 $g^j$ 也保测，所以

$$
\int_X|f\circ g^j|\,d\mu=\int_X|f|\,d\mu.
$$

### 为什么必须假设 $g$ 可测？

如果 $g$ 不可测，即使 $f$ 可测，$f\circ g$ 也不一定可测。于是 Koopman 算子、轨道部分和、极大函数及其积分都可能无法定义。标准的“保测映射”定义本身已经包含 $g$ 可测。


## 21.1 遍历性的定义（Definition of Ergodicity）

若每个不变集合（invariant set）都只有零测度或满测度，即

$$
g^{-1}(A)=A
\quad\Longrightarrow\quad
\mu(A)=0\ \text{或}\ 1,
$$

则称 $g$ 是遍历的（ergodic）。这表示系统不能分成两个都具有正测度、并且互不交流的不变区域。

### 遍历性不等于轨道到达几乎所有点

从 $x$ 出发的轨道（orbit）是

$$
\mathcal O(x)=\{x,g(x),g^2(x),\ldots\}.
$$

在连续空间中，它通常是可数集，因而可能满足 $\mu(\mathcal O(x))=0$。遍历性的正确直觉是：

> 对几乎所有初始点，轨道会以正确的长期频率访问每个固定的正测度区域，而不是精确经过几乎所有点。

若 $X$ 是紧致度量空间（compact metric space），$\mu$ 具有满支撑（full support），且 $g$ 遍历，则几乎每条轨道在 $X$ 中稠密（dense）。这不是遍历性的定义，而是 Birkhoff 遍历定理的推论。若测度不满支撑，通常只能得到轨道在 $\operatorname{supp}\mu$ 中稠密；而且结论是“几乎每条”，并非“每条”。


## 21.2 不变函数刻画（Invariant-Function Characterization）

可测函数 $f:X\to\mathbb R$ 称为不变函数（invariant function），如果

$$
f\circ g=f
\qquad\text{几乎处处（almost everywhere, a.e.）}.
$$

### 引理 21.2（Lemma 21.2）

$g$ 遍历，当且仅当每个可测不变函数都几乎处处为常数。

### 证明思路

假设 $g$ 遍历。对任意 $c\in\mathbb R$，集合

$$
A_c=\{x:f(x)<c\}
$$

是不变的，因此 $\mu(A_c)\in\{0,1\}$。若 $f$ 不是几乎处处常数，可以在两个本质取值之间选择 $c$，使 $A_c$ 及其补集都具有正测度，这与遍历性矛盾。

反过来，若 $A$ 是不变集合，则

$$
\mathbf 1_A\circ g=\mathbf 1_A.
$$

若所有不变函数都几乎处处为常数，则 $\mathbf 1_A$ 只能几乎处处等于 $0$ 或 $1$，从而 $\mu(A)=0$ 或 $1$。

### 重要澄清

即使 $g$ 遍历，任意 $f\in L^1$ 也不一定是常数。引理只适用于满足 $f\circ g=f$ 的函数。在 Birkhoff 定理中，原函数 $f$ 可以任意；是时间平均的极限 $f^*$ 不变，因此遍历性使 $f^*$ 成为常数。


## 21.3 例 21.3：圆周旋转（Circle Rotation）

考虑圆周 $S^1\simeq[0,1]/(0\sim1)$ 上的旋转

$$
g(x)=x+r\pmod1,
$$

并取勒贝格测度（Lebesgue measure）。则

$$
g\text{ 遍历}
\quad\Longleftrightarrow\quad
r\notin\mathbb Q.
$$

设 $f\in L^2(S^1)$ 是不变函数，并写成 Fourier 级数（Fourier series）

$$
f(x)=\sum_{n\in\mathbb Z}a_ne^{2\pi inx}.
$$

由 $f(x+r)=f(x)$ 及 Fourier 系数的唯一性，

$$
a_ne^{2\pi inr}=a_n.
$$

若 $r$ 无理，则对所有 $n\neq0$，$e^{2\pi inr}\neq1$，所以 $a_n=0$。因此只剩常数项 $a_0$，由引理 21.2 可知 $g$ 遍历。

若 $r=p/q$ 有理，则非恒定函数

$$
f(x)=\cos(2\pi qx)
$$

满足 $f\circ g=f$，所以 $g$ 不遍历。

无理旋转的每条轨道都在圆周上稠密，但每条轨道仍是可数集且勒贝格测度为零。稠密表示可以任意接近每个点，并不表示精确经过几乎所有点。


## 21.4 Koopman 算子（Koopman Operator）

定义

$$
(\mathcal L_gf)(x)=f(g(x)).
$$

其迭代满足

$$
\mathcal L_g^jf=f\circ g^j.
$$

若 $g$ 保测，则

$$
\|\mathcal L_gf\|_{L^p}=\|f\|_{L^p},
$$

所以 $\mathcal L_g$ 是等距算子（isometry）。它也是正算子（positive operator）：

$$
f\ge0\quad\Longrightarrow\quad\mathcal L_gf\ge0.
$$

保测性保证 $\mathcal L_g$ 把 $L^p$ 函数映射回 $L^p$，并且保持相应范数与积分。


## 21.5 极大遍历定理（Maximal Ergodic Theorem）

对 $f\in L^1$，定义部分和（partial sums）

$$
f_0=0,
\qquad
f_n=f+\mathcal Lf+\cdots+\mathcal L^{n-1}f,
$$

即

$$
f_n(x)=\sum_{j=0}^{n-1}f(g^j(x)).
$$

定义极大函数（maximal function）

$$
F_N(x)=\max_{0\le n\le N}f_n(x)\ge0.
$$

$F_N(x)$ 比较的是不同长度的累积和，而不是单个轨道值。令

$$
A_N=\{x:F_N(x)>0\}.
$$

则

$$
x\in A_N
\iff
\exists\,1\le n\le N:\ f_n(x)>0.
$$

因此 $A_N$ 是“前 $N$ 步中曾出现正累积和”的初始点集合。定理积分的是原函数 $f$：

$$
\boxed{\int_{A_N}f\,d\mu\ge0.}
$$

### 证明的关键步骤

由 $F_N\ge f_{n-1}$ 和 Koopman 算子的正性，

$$
\mathcal LF_N+f
\ge\mathcal Lf_{n-1}+f
=f_n.
$$

在 $A_N$ 上，最大值不是 $f_0=0$，所以

$$
\mathcal LF_N+f\ge F_N,
\qquad
f\ge F_N-\mathcal LF_N.
$$

在 $X\setminus A_N$ 上，$F_N\ge0$ 且 $F_N\not>0$，故 $F_N=0$，从而

$$
\int_{A_N}F_N\,d\mu=\int_XF_N\,d\mu.
$$

最后，利用 $\mathcal LF_N\ge0$ 和保测性，

$$
\int_{A_N}f\,d\mu
\ge
\int_XF_N\,d\mu-\int_X\mathcal LF_N\,d\mu
=0.
$$


## 21.6 极大不等式（Maximal Inequality）

对 $h\in L^1$，定义

$$
B_\alpha
=
\left\{
x:
\sup_{m\ge1}\frac1m\sum_{j=0}^{m-1}h(g^j(x))>\alpha
\right\}.
$$

它表示：存在某个有限时间 $m$，使从 $x$ 出发的时间平均超过 $\alpha$。若 $A$ 是不变集合，则推论 21.6 给出

$$
\int_{B_\alpha\cap A}h\,d\mu
\ge
\alpha\,\mu(B_\alpha\cap A).
$$

证明是在不变子系统 $A$ 上，对 $h-\alpha$ 应用极大遍历定理，并令有限最大值的时间范围趋于无穷。这个不等式随后用于排除 Birkhoff 时间平均的上极限与下极限之间存在间隙。


## 21.7 Birkhoff 遍历定理（Birkhoff Ergodic Theorem）

定义时间平均（time average）

$$
S_nf(x)=\frac1n\sum_{j=0}^{n-1}f(g^j(x)).
$$

若 $f\in L^1$，则：

1. 存在 $f^*\in L^1$，使 $S_nf(x)\to f^*(x)$ 对几乎所有 $x$ 成立；
2. $f^*\circ g=f^*$，并且 $\int_Xf^*\,d\mu=\int_Xf\,d\mu$；
3. 若 $g$ 遍历，则

   $$
   f^*(x)=\int_Xf\,d\mu
   \qquad\text{a.e.}
   $$

因此，遍历情形下，时间平均（time average）等于空间平均（space average）。

### 证明结构

定义

$$
\overline f=\limsup_{n\to\infty}S_nf,
\qquad
\underline f=\liminf_{n\to\infty}S_nf.
$$

由

$$
S_nf(g(x))
=\frac{n+1}{n}S_{n+1}f(x)-\frac{f(x)}n
$$

可知 $\overline f$ 和 $\underline f$ 都不变。这里的不变性来自时间平均平移一个时间步，本身不需要保测性。

对有理数 $\beta<\alpha$，令

$$
E_{\alpha,\beta}
=\{x:\underline f(x)<\beta<\alpha<\overline f(x)\}.
$$

由于 $E_{\alpha,\beta}\subseteq B_\alpha$，推论 21.6 给出

$$
\int_{E_{\alpha,\beta}}f\,d\mu
\ge\alpha\mu(E_{\alpha,\beta}).
$$

对 $-f$ 应用同一结论，得到

$$
\int_{E_{\alpha,\beta}}f\,d\mu
\le\beta\mu(E_{\alpha,\beta}).
$$

因为 $\beta<\alpha$，所以 $\mu(E_{\alpha,\beta})=0$。对有理数对取可数并可知 $\underline f=\overline f$ 几乎处处，故极限存在。

Fatou 引理（Fatou's Lemma）和保测性给出

$$
\int_X|f^*|\,d\mu
\le\liminf_n\int_X|S_nf|\,d\mu
\le\int_X|f|\,d\mu,
$$

所以 $f^*\in L^1$。再将 $f^*$ 的值域分成长度为 $1/n$ 的区间，在相应不变集合上应用推论 21.6，并对 $-f$ 重复论证，可得

$$
\int_Xf^*\,d\mu=\int_Xf\,d\mu.
$$

最后，若 $g$ 遍历，则不变函数 $f^*$ 几乎处处为常数；因为 $\mu(X)=1$，该常数为 $\int_Xf\,d\mu$。

### 访问频率与稠密轨道

取 $f=\mathbf 1_B$。若 $g$ 遍历，则对每个固定可测集合 $B$，

$$
\lim_{n\to\infty}
\frac1n\sum_{j=0}^{n-1}\mathbf 1_B(g^j(x))
=\mu(B)
$$

对几乎所有 $x$ 成立。若 $\mu(B)>0$，典型轨道会无穷多次进入 $B$，访问频率为 $\mu(B)$。若空间具有可数拓扑基且测度满支撑，对所有基元素取可数交即可推出典型轨道稠密。


## 21.8 von Neumann $L^p$ 遍历定理（von Neumann $L^p$ Ergodic Theorem）

设 $1\le p<\infty$ 且 $f\in L^p$。则存在不变函数 $f^*\in L^p$，使

$$
\|S_nf-f^*\|_{L^p}\longrightarrow0.
$$

### 证明思路

取 $h\in L^\infty$，使

$$
\|f-h\|_{L^p}<\frac{\varepsilon}{3}.
$$

Birkhoff 定理给出 $S_nh$ 的几乎处处收敛。因为 $|S_nh|\le\|h\|_\infty$，支配收敛定理（Dominated Convergence Theorem）将其提升为 $L^p$ 收敛。

又因为时间平均是压缩算子（contraction），

$$
\|S_nu\|_{L^p}\le\|u\|_{L^p},
$$

所以标准的 $\varepsilon/3$ 论证表明 $(S_nf)$ 是 $L^p$ 中的柯西序列（Cauchy sequence）。极限的不变性来自

$$
(S_nf)\circ g-S_nf
=\frac{f\circ g^n-f}{n},
$$

其 $L^p$ 范数不超过 $2\|f\|_{L^p}/n$。

### 该定理不能单独推出时间平均等于空间平均

von Neumann 定理只保证 $S_nf$ 收敛到某个不变函数 $f^*$。在非遍历系统中，$f^*$ 不一定是常数；更准确地说，

$$
f^*=\mathbb E[f\mid\mathcal I],
$$

其中 $\mathcal I$ 是不变 $\sigma$-代数（invariant sigma-algebra）。只有再假设 $g$ 遍历，才能得到

$$
f^*=\int_Xf\,d\mu.
$$

因此，核心逻辑为

$$
\boxed{
\text{保测}\Longrightarrow\text{时间平均收敛到不变函数};
\qquad
\text{遍历}\Longrightarrow\text{该极限等于空间平均}.
}
$$
