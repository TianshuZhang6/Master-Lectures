# Lecture 1 — Dynamical Systems

## 1. Dynamical systems and examples

A **dynamical system** is a triple $(\mathcal{X},\mathcal{T},\phi)$ consisting of a phase space $\mathcal{X}$, a time set $\mathcal{T}$, and an evolution map

$$
\phi : \mathcal{T}\times\mathcal{X}\longrightarrow\mathcal{X},
\qquad (t,x)\longmapsto\phi_t(x),
$$

such that, whenever the expressions are defined,

$$
\phi_0=\operatorname{Id}_{\mathcal{X}},
\qquad
\phi_{t+s}=\phi_t\circ\phi_s.
$$

Autonomous differential equations therefore give rise naturally to dynamical systems.

### Continuous-time example

For the linear ODE $\dot{x}=Ax$ with $x_0\in\mathbb{R}^d$,

$$
\phi_t(x_0)=e^{tA}x_0,
\qquad
e^{tA}=\sum_{k=0}^{\infty}\frac{(tA)^k}{k!}.
$$

### Discrete-time example

Let $g:\mathbb{R}\to\mathbb{R}$. The iterated map generated by $g$ is

$$
x_{k+1}=g(x_k),
\qquad
x_k=\phi_k(x_0)=g^k(x_0),
\qquad
\phi_0=\operatorname{Id}.
$$

A **cobweb diagram** is a standard graphical method for visualizing these iterates.

## 2. Discrete-time and continuous-time systems

A system is classified by its time set:

- **Discrete time:** $\mathcal{T}=\mathbb{N}_0$ for forward iteration, or $\mathcal{T}=\mathbb{Z}$ when backward iterates are also defined.
- **Continuous time:** $\mathcal{T}$ is an interval in $\mathbb{R}$; common choices are $\mathbb{R}$ and $[0,\infty)$.

The notation $\phi_k$ is generally used in discrete time, whereas $\phi_t$ is used in continuous time.

## 3. Invertibility, flows, and semiflows

If $\mathcal{T}$ is an additive group and every $\phi_t$ has an inverse, then

$$
\phi_t^{-1}=\phi_{-t},
\qquad
\phi_t\circ\phi_{-t}=\phi_{-t}\circ\phi_t=\operatorname{Id}_{\mathcal{X}}.
$$

A continuous-time system with $\mathcal{T}=\mathbb{R}$ is called a **flow**. A system defined only for forward time, usually with $\mathcal{T}=[0,\infty)$, is called a **semiflow**.

## 4. Existence, uniqueness, and smooth dependence

Consider the autonomous initial-value problem

$$
\dot{x}=f(x),
\qquad
x(0)=x_0,
\qquad
f:\mathbb{R}^d\to\mathbb{R}^d.
$$

If $f$ is smooth, then for every $x_0$ there is a unique maximal solution

$$
x(t)=\phi_t(x_0),
\qquad
t\in I(x_0)=\bigl(-t_-(x_0),t_+(x_0)\bigr).
$$

The solution map $(t,x_0)\mapsto\phi_t(x_0)$ is smooth on its domain. If $I(x_0)=\mathbb{R}$ for every $x_0$, the solutions define a global flow.

### Banach fixed-point theorem

Let $(\mathcal{X},d)$ be a non-empty complete metric space. If $g:\mathcal{X}\to\mathcal{X}$ satisfies

$$
d\bigl(g(x),g(y)\bigr)\leq\lambda d(x,y)
\qquad\text{for some }0\leq\lambda<1,
$$

then $g$ has a unique fixed point $x_*\in\mathcal{X}$, and the iteration $x_{n+1}=g(x_n)$ converges to $x_*$ for every initial point $x_0\in\mathcal{X}$.

## 5. Smooth dynamical systems

A continuous-time system $(\mathbb{R}^d,\mathbb{R},\phi_t)$ is called **smooth** when the map

$$
(t,x)\longmapsto\phi_t(x)
$$

is smooth on its domain of definition. Smoothness permits differential tools—such as derivatives, linearization, and invariant manifolds—to be applied to the flow.

## 6. Orbits and trajectories

The **orbit** (or **trajectory**) through $x_0\in\mathcal{X}$ is

$$
\operatorname{Orb}(x_0)
=
\left\{\phi_t(x_0)\in\mathcal{X}:t\in\mathcal{T}\text{ and }\phi_t(x_0)\text{ is defined}\right\}.
$$

It is the collection of all states visited by the solution starting at $x_0$.

## 7. Steady states, equilibria, and fixed points

A point $x_*\in\mathcal{X}$ is a **steady state**, **equilibrium**, or **fixed point** if

$$
\phi_t(x_*)=x_*
\qquad\text{for all admissible }t.
$$

For a continuous-time system $\dot{x}=f(x)$, this is equivalent to $f(x_*)=0$. The term *equilibrium* is customary for flows; *fixed point* is customary for discrete maps.

## 8. Nullclines

For the ODE $\dot{x}=f(x)$ on $\mathbb{R}^d$, the $k$th **nullcline** is

$$
N_k=\left\{x\in\mathbb{R}^d:f_k(x)=0\right\},
\qquad k=1,\ldots,d.
$$

On $N_k$, the $k$th component of the velocity is zero. Every equilibrium lies at the intersection of all nullclines.

## 9. Invariant sets

A subset $\mathcal{S}\subseteq\mathcal{X}$ is **invariant** if an orbit that starts in $\mathcal{S}$ remains in $\mathcal{S}$:

$$
x_0\in\mathcal{S}
\quad\Longrightarrow\quad
\phi_t(x_0)\in\mathcal{S}
\quad\text{for every admissible }t.
$$

It is **positively invariant** if this holds for $t\geq0$, and **negatively invariant** if it holds for $t\leq0$.

## 10. Periodic orbits

A **periodic orbit** is a non-equilibrium orbit $\gamma$ for which there exists $T>0$ such that

$$
\phi_{t+T}(x_0)=\phi_t(x_0)
\qquad
\text{for all }t\text{ and every }x_0\in\gamma.
$$

The smallest such $T$ is the **minimal period** of $\gamma$.

## 11. Phase portraits

A **phase portrait** is the geometric representation of the collection of orbits in the phase space. It shows the qualitative behavior of trajectories, including equilibria, periodic orbits, and invariant sets.
