# Lecture 12: Global Bifurcations

## Learning Objectives

After this lecture, you should be able to:

- distinguish heteroclinic orbits, homoclinic orbits, homoclinic loops, and limit cycles;
- explain what the split function measures and why $\xi'(0)\neq0$ is the generic breaking condition;
- derive the local and global maps used in the Andronov-Leontovich theorem;
- determine the existence and stability of the bifurcating limit cycle from the return map;
- explain the role of nonresonance in Sternberg linearization;
- describe how a rectangle becomes a spiral band near a Shilnikov saddle-focus;
- interpret homoclinic and periodic travelling-wave profiles in the FitzHugh-Nagumo PDE.


## 1. Global Connections

Consider the smooth one-parameter family

$$
\dot{x}=f(x,p),
\qquad x\in\mathbb R^d,
\quad p\in\mathbb R,
\quad d\in\{2,3\},
$$

where $f(0,p)=0$ and $x_*=0$ is a hyperbolic saddle. Local bifurcations concern dynamics near an equilibrium or periodic orbit. Global bifurcations involve global invariant objects, especially stable and unstable manifolds.

An orbit $\gamma(t)$ is **heteroclinic from $x_*$ to $y_*$** if

$$
\lim_{t\to-\infty}\gamma(t)=x_*,
\qquad
\lim_{t\to+\infty}\gamma(t)=y_*,
\qquad x_*\neq y_*.
$$

Equivalently,

$$
\gamma\subset W^u(x_*)\cap W^s(y_*).
$$

A nonconstant orbit $\gamma_0(t)$ is **homoclinic to $0$** if

$$
\lim_{t\to-\infty}\gamma_0(t)
=
\lim_{t\to+\infty}\gamma_0(t)
=0.
$$

Thus $\gamma_0\subset W^u(0)\cap W^s(0)$. In two dimensions, such saddle connections are structurally unstable: a small perturbation generally separates the invariant manifolds.

### Why $\gamma_0\cup\{0\}$ is written explicitly

If the orbit is defined as

$$
\gamma_0=\{\gamma_0(t):t\in\mathbb R\},
$$

then the equilibrium is not an element of $\gamma_0$. If $\gamma_0(t_0)=0$ at a finite time, uniqueness of ODE solutions would imply $\gamma_0(t)\equiv0$, contradicting the assumption that the orbit is nonconstant. Therefore,

$$
0\notin\gamma_0,
\qquad
\overline{\gamma_0}=\gamma_0\cup\{0\}.
$$

The closed geometric object $\gamma_0\cup\{0\}$ is usually called the **homoclinic loop**.

A homoclinic orbit is not a limit cycle. A limit cycle $\Gamma_p$ is a genuine periodic orbit with a finite period $T_p>0$:

$$
x_p(t+T_p)=x_p(t).
$$

It cannot contain an equilibrium. When a limit cycle bifurcates from a homoclinic loop, it approaches $\gamma_0\cup\{0\}$ geometrically along the cycle-existence side. Since it passes increasingly close to the saddle,

$$
T_p\longrightarrow+\infty.
$$

Thus the homoclinic loop is an infinite-period limiting object, but it is not itself a periodic orbit.


## 2. The Split Function

Assume that a homoclinic orbit exists at $p=0$. Choose a transverse section $\Sigma$ through a regular point of the loop and place the stable-manifold intersection at coordinate zero. Let the returned unstable manifold intersect $\Sigma$ at

$$
q=\xi(p).
$$

The function $\xi$ is the **split function**. It measures the signed separation of the invariant manifolds:

$$
\xi(p)=0
\quad\Longleftrightarrow\quad
\text{the homoclinic connection exists},
$$

whereas $\xi(p)\neq0$ means that the connection has been broken. The generic breaking condition is

$$
\xi(0)=0,
\qquad
\xi'(0)\neq0.
$$

Locally,

$$
q=\xi(p)=\xi'(0)p+o(p).
$$

The condition is not $\xi'(0)=0$. The derivative being nonzero means that the invariant manifolds separate to first order when the parameter crosses zero.

Three different conditions must be distinguished:

- $\xi(p)=0$ detects a homoclinic connection.
- $\xi'(0)\neq0$ guarantees a nondegenerate parameter unfolding.
- A limit cycle is detected by a nonzero fixed point of a Poincare return map:

  $$
  P_p(z)=z,
  \qquad z>0.
  $$

A periodic orbit is not a piece of $W^u$ returning to $W^u$. It consists of ordinary points near $W^s$ and $W^u$, and is found by returning a point of the transverse section to itself.

The sign of $\xi$ depends on the orientation chosen on $\Sigma$. Therefore, neither the parameter direction nor the words "inside" and "outside" have meaning until the positive direction of the split coordinate has been specified.


## 3. The Andronov-Leontovich Theorem

Suppose $d=2$ and the saddle eigenvalues satisfy

$$
\lambda_s(0)<0<\lambda_u(0).
$$

Define the **saddle quantity**

$$
\sigma(p)=\lambda_s(p)+\lambda_u(p).
$$

In two dimensions, $\sigma(p)=\operatorname{tr}D_xf(0,p)$. Assume that $\gamma_0$ is a homoclinic orbit at $p=0$ and that

$$
\sigma(0)\neq0,
\qquad
\xi'(0)\neq0.
$$

Then, on one side of $p=0$, there is a unique limit cycle near $\gamma_0\cup\{0\}$. Its stability is determined by $\sigma(0)$:

- If $\sigma(0)<0$, the cycle is stable.
- If $\sigma(0)>0$, the cycle is unstable.

The condition $\sigma(0)\neq0$ excludes the case in which stable contraction and unstable expansion have equal leading strength.

### Proof: straightening the invariant manifolds

Near the saddle, write

$$
\begin{aligned}
\dot{x}_1&=\lambda_sx_1+f_1(x_1,x_2),\\
\dot{x}_2&=\lambda_ux_2+f_2(x_1,x_2).
\end{aligned}
$$

The local invariant manifolds have graph representations

$$
W^s_{\mathrm{loc}}(0)=\{x_2=S(x_1)\},
\qquad
W^u_{\mathrm{loc}}(0)=\{x_1=U(x_2)\},
$$

where $S(0)=S'(0)=U(0)=U'(0)=0$. The transformation

$$
y_1=x_1-U(x_2),
\qquad
y_2=x_2-S(x_1)
$$

straightens the manifolds into the coordinate axes. The transformed system has the form

$$
\begin{aligned}
\dot y_1&=y_1\bigl(\lambda_s+g_1(y)\bigr),\\
\dot y_2&=y_2\bigl(\lambda_u+g_2(y)\bigr),
\end{aligned}
$$

where $g(y)=O(\lVert y\rVert)$. The factors $y_1$ and $y_2$ occur because the two coordinate axes are invariant.

### Exit-time linearization

A local $C^1$ transformation is constructed by matching the forward and backward exit times of nonlinear and linear trajectories from a small rectangle. If $\tau_s$ is the backward exit time through the stable boundary and $\tau_u$ is the forward exit time through the unstable boundary, the corresponding linear coordinates can be written as

$$
X=e^{\lambda_s\tau_s},
\qquad
Y=e^{-\lambda_u\tau_u}.
$$

In these coordinates,

$$
\dot X=\lambda_sX,
\qquad
\dot Y=\lambda_uY.
$$

Variation of constants and $g(y)=O(\lVert y\rVert)$ show that the nonlinear correction is higher order. In particular, $D\Phi(0)=I$, so the transformation is a local $C^1$ diffeomorphism. This special planar construction does not require the full nonresonance assumptions used for $C^\infty$ linearization.

### Local and global maps

Choose

$$
\Sigma_{\mathrm{in}}=\{X=1,\ 0<Y\leq1\},
\qquad
\Sigma_{\mathrm{out}}=\{0<X\leq1,\ Y=1\}.
$$

Starting from $(1,Y)$, the linear solution is

$$
X(t)=e^{\lambda_st},
\qquad
Y(t)=Ye^{\lambda_ut}.
$$

The orbit reaches $\Sigma_{\mathrm{out}}$ when $Ye^{\lambda_uT}=1$, so

$$
T=-\frac{1}{\lambda_u}\log Y,
\qquad
X(T)=Y^{-\lambda_s/\lambda_u}.
$$

Define

$$
\rho=-\frac{\lambda_s}{\lambda_u}>0.
$$

The local map is $L(Y)=Y^\rho$. The global excursion defines

$$
G(X,p)=q+aX+O(X^2),
\qquad
q=\xi(p),
$$

with $a>0$ after choosing compatible section orientations. The full return map is

$$
P_p(Y)=G\circ L(Y)=q+aY^\rho+\text{higher-order terms}.
$$

A genuine limit cycle corresponds to

$$
P_p(Y_*)=Y_*,
\qquad Y_*>0.
$$

Equivalently, it is a positive zero of $D(Y,p)=P_p(Y)-Y$. The boundary value $Y=0$ corresponds to the separatrix, has infinite passage time, and is not an ordinary point of the Poincare map. The condition $q=0$ at this boundary produces the homoclinic connection.

### Existence, uniqueness and stability

The exponent satisfies

$$
\sigma=\lambda_s+\lambda_u=\lambda_u(1-\rho).
$$

If $\sigma<0$, then $\rho>1$ and $Y^\rho=o(Y)$. In the proof coordinates, a unique small positive root exists for $q>0$:

$$
Y_*=q+o(q),
\qquad
P_p'(Y_*)\sim a\rho Y_*^{\rho-1}\longrightarrow0.
$$

Hence the cycle is stable.

If $\sigma>0$, then $0<\rho<1$ and $Y^\rho$ dominates $Y$. In the same proof coordinates, a unique small positive root exists for $q<0$:

$$
Y_*\sim\left(\frac{-q}{a}\right)^{1/\rho},
\qquad
P_p'(Y_*)\sim a\rho Y_*^{\rho-1}\longrightarrow+\infty.
$$

Hence the cycle is unstable.

These signs belong to the oriented coordinates used in the proof. Reversing the section coordinate reverses the sign of $q$. To determine the sign of the original parameter, first identify how the physical split coordinate is related to the proof coordinate, then use $q=\xi'(0)p+o(p)$. The inside/outside location likewise requires the geometric orientation of the section.

In Exercise E12.1, the $x$-axis coordinate points outward, while the positive coordinate used in the return-map proof points inward. Since

$$
\sigma=-2,
\qquad
\xi'(0)<0,
$$

increasing the exercise parameter $\alpha$ breaks the separatrix inward and produces a stable cycle for $\alpha>0$. That cycle lies inside the former homoclinic loop.


## 4. Sternberg Linearization and Resonance

Let $g:\mathbb R^d\to\mathbb R^d$ be a smooth map with a fixed point at the origin and

$$
Dg(0)=\Lambda=\operatorname{diag}(\lambda_1,\ldots,\lambda_d).
$$

For a multi-index $k=(k_1,\ldots,k_d)$, define

$$
|k|=k_1+\cdots+k_d,
\qquad
\lambda^k=\lambda_1^{k_1}\cdots\lambda_d^{k_d}.
$$

The map nonresonance condition is

$$
\lambda_j\neq\lambda^k,
\qquad |k|\geq2.
$$

Under the hypotheses of the Sternberg theorem, this condition gives a local $C^\infty$ transformation $H$ satisfying

$$
H\circ g=\Lambda\circ H.
$$

To remove a nonlinear monomial $cx^ke_j$, the corresponding homological equation requires division by $\lambda^k-\lambda_j$. If this quantity is nonzero, the term can be removed. If

$$
\lambda^k=\lambda_j,
$$

the term is resonant and generally remains in the normal form. Dynamically, the nonlinear monomial then scales at the same rate as the $j$-th linear eigendirection.

Hartman-Grobman gives a topological conjugacy near a hyperbolic fixed point. Sternberg nonresonance gives the stronger smooth conjugacy needed to preserve differential and asymptotic information.

For a continuous-time system with eigenvalues $\mu_j$, the corresponding resonance relation is additive:

$$
\mu_j=k_1\mu_1+\cdots+k_d\mu_d.
$$

The chapter's planar proof only requires a specially constructed $C^1$ linearization, so it does not conflict with the stronger Sternberg theorem.


## 5. Shilnikov Homoclinic Orbits

Suppose a three-dimensional saddle-focus has eigenvalues

$$
\lambda_1=\lambda>0,
\qquad
\lambda_{2,3}=\alpha\pm i\omega,
\quad \alpha<0,
\quad \omega>0.
$$

Here

$$
\alpha=\operatorname{Re}\lambda_2=\operatorname{Re}\lambda_3.
$$

$\alpha$ is the common real part of each complex eigenvalue, not their sum. The sum of the two real parts is $2\alpha$, while the trace is $\lambda+2\alpha$. The Shilnikov saddle quantity is instead

$$
\sigma=\lambda+\alpha,
$$

because the radius in the stable plane satisfies $\dot r=\alpha r$. The Shilnikov condition is

$$
\lambda+\alpha>0
\quad\Longleftrightarrow\quad
\lambda>|\alpha|.
$$

At the homoclinic parameter, this condition produces horseshoe-type invariant sets and infinitely many saddle periodic orbits near the loop. A saddle periodic orbit has one trivial Floquet multiplier equal to $1$ and two nontrivial multipliers satisfying

$$
|\mu_s|<1<|\mu_u|.
$$

A fixed compact horseshoe persists under sufficiently small perturbations. This does not mean that the homoclinic connection itself persists for every nearby parameter.

### How a rectangle becomes a spiral band

In suitable coordinates,

$$
\begin{aligned}
\dot\zeta&=\lambda\zeta,\\
\dot\xi_1&=\alpha\xi_1-\omega\xi_2,\\
\dot\xi_2&=\omega\xi_1+\alpha\xi_2.
\end{aligned}
$$

Let $z=\xi_1+i\xi_2$. Then

$$
\zeta(t)=\zeta_0e^{\lambda t},
\qquad
z(t)=z_0e^{(\alpha+i\omega)t}.
$$

On the incoming section, parameterize a small rectangle by

$$
(\zeta,\xi_1,\xi_2)=(s,c_h,u),
\qquad
0<s<\varepsilon,
\quad |u|<\delta.
$$

Let the outgoing section be $\Sigma_v=\{\zeta=c_v\}$. The flight time is

$$
T(s)=\frac1\lambda\log\frac{c_v}{s}.
$$

As $s\to0^+$, $T(s)\to+\infty$. Points closer to the stable manifold remain near the saddle-focus longer and rotate more times. Define

$$
\nu=-\frac{\alpha}{\lambda}>0.
$$

The stable output coordinate has the form

$$
z_{\mathrm{out}}
=
\left(\frac{s}{c_v}\right)^\nu
e^{i\theta(s)}(c_h+iu),
\qquad
\theta(s)=\frac{\omega}{\lambda}\log\frac{c_v}{s}.
$$

For fixed $s$, varying $u$ produces a short line segment. It is rotated through $\theta(s)$ and scaled by a factor of order $s^\nu$. As $s$ decreases:

- its rotation angle tends to infinity;
- its distance from the spiral center decreases like $s^\nu$;
- its absolute length also decreases like $s^\nu$.

The union of these rotating segments is a two-dimensional spiral band, not a single trajectory. It accumulates at

$$
W^u_{\mathrm{loc}}(0)\cap\Sigma_v=(c_v,0,0),
$$

not at the equilibrium $(0,0,0)$. The value $s=0$ lies on the stable manifold and never reaches the outgoing section. Therefore, the spiral center is the limit of exit positions from a family of different trajectories.

The absolute width satisfies $\operatorname{width}\sim\delta s^\nu$. Its center radius is also of order $s^\nu$, so its relative thickness need not tend to zero.

### The source of stretching

The band is not literally stretched along the unstable axis, because $\zeta=c_v$ is fixed on the outgoing section. The unstable coordinate instead controls the flight time. Indeed,

$$
\theta'(s)=-\frac{\omega}{\lambda s}.
$$

The local map satisfies

$$
\left\lVert\frac{\partial L}{\partial u}\right\rVert\asymp s^\nu,
\qquad
\left\lVert\frac{\partial L}{\partial s}\right\rVert\asymp s^{\nu-1}.
$$

The Shilnikov condition gives $0<\nu<1$. Hence $s^\nu\to0$ while $s^{\nu-1}\to+\infty$. The rectangle is compressed across the band and strongly stretched along its spiral direction. This is derivative-level stretching caused by sensitivity of the exit time, rather than direct elongation along the unstable axis.

The local and global maps satisfy

$$
L:\Sigma_h\to\Sigma_v,
\qquad
G:\Sigma_v\to\Sigma_h.
$$

Therefore, the return map based on $\Sigma_h$ is $P=G\circ L:\Sigma_h\to\Sigma_h$. Figure 31's caption that labels $G\circ L$ as a map from $\Sigma_v$ to itself is inconsistent; a return map based on $\Sigma_v$ would be $L\circ G$.

The global map produces only smooth finite distortion and does not remove the infinitely many turns. Their repeated intersections with the original rectangle produce horseshoe dynamics and infinitely many saddle periodic orbits.


## 6. FitzHugh-Nagumo Travelling Waves

Consider

$$
\begin{aligned}
u_t&=u_{xx}+f(u)-v,\\
v_t&=\varepsilon(u-v),
\end{aligned}
$$

where

$$
f(u)=u(1-u)(u-p),
\qquad
0<p<\frac12.
$$

Seek a travelling wave

$$
u(x,t)=U(s),
\qquad
v(x,t)=V(s),
\qquad
s=x-ct.
$$

Then $u_t=-cU'$, $u_{xx}=U''$, and $v_t=-cV'$. Setting $W=U'$ gives, for $c\neq0$,

$$
\begin{aligned}
U'&=W,\\
W'&=V-cW-f(U),\\
V'&=\frac{\varepsilon}{c}(V-U).
\end{aligned}
$$

The prime denotes differentiation with respect to the travelling coordinate $s$, not the original time $t$.

### Interpretation of the ODE orbits

A homoclinic profile satisfies

$$
(U(s),W(s),V(s))\longrightarrow(0,0,0)
\qquad(s\to\pm\infty).
$$

In the PDE, this is a localized travelling pulse whose profile returns to the resting state at both spatial ends. A periodic orbit of the profile ODE gives a spatially periodic travelling wave, or wavetrain. Stability in this spatial ODE does not by itself establish temporal or spectral stability of the PDE solution.

### Equilibria

For $\varepsilon\neq0$ and $c\neq0$, an equilibrium satisfies $W=0$ and $V=U=f(U)$. For $U\neq0$, this would require

$$
(1-U)(U-p)=1.
$$

However,

$$
\max_{U\in\mathbb R}(1-U)(U-p)=\frac{(1-p)^2}{4}<1.
$$

Thus the only equilibrium is $(U,W,V)=(0,0,0)$.

At $\varepsilon=0$, this uniqueness fails. The full three-dimensional system has the equilibrium curve

$$
\{(U,0,f(U)):U\in\mathbb R\},
$$

and the invariant plane $V=0$ contains $(0,0,0)$, $(p,0,0)$, and $(1,0,0)$.

### Hamiltonian singular limit

When $\varepsilon=0$, the variable $V$ is constant. Restricting to $V=0$ gives

$$
U'=W,
\qquad
W'=-cW-f(U).
$$

Only after this reduction may one set $c=0$. Direct substitution of $c=0$ into the full three-dimensional equation is invalid because that equation was obtained by dividing by $c$.

For $c=0$,

$$
U'=W,
\qquad
W'=-f(U).
$$

This is Hamiltonian with

$$
H(U,W)=\frac12W^2+F(U),
$$

where

$$
F(U)=-\frac14U^4+\frac{1+p}{3}U^3-\frac p2U^2.
$$

Along solutions, $dH/ds=0$. Since $f'(0)=-p$, the origin is a saddle with eigenvalues $\pm\sqrt p$. The zero-energy curve $H(U,W)=0$ contains a homoclinic loop to the origin. In the PDE, this is a stationary localized pulse.

For $\varepsilon=0$ but $c\neq0$,

$$
\frac{dH}{ds}=-cW^2,
$$

so the Hamiltonian loop is generally destroyed. In the full system,

$$
\frac{dH}{ds}=VW-cW^2.
$$

The recovery variable $V$ can supply or remove energy on different parts of the orbit, making a global balance possible. Constructing an exact three-dimensional homoclinic connection nevertheless requires the stable and unstable manifolds to match and is substantially harder.

For suitable parameters, the origin may have one positive real eigenvalue and a stable complex pair $\alpha\pm i\omega$. If a homoclinic orbit exists and $\lambda+\alpha>0$, the Shilnikov theorem applies to the travelling-wave ODE.

The source's final reference to parameters $(c,a)$ is inconsistent with its earlier notation. In the displayed system, the relevant parameters are $c$, $p$, and $\varepsilon$.
