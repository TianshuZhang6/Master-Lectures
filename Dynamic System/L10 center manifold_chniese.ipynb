# 第 10 讲：中心流形

## 学习目标

学习本讲后，你应当能够：

- 解释为什么在非双曲平衡点附近需要中心流形；
- 区分线性中心子空间 $E^c$ 与非线性中心流形 $W^c$；
- 陈述中心流形定理与中心流形约化定理；
- 理解中心流形的非唯一性与正则性限制；
- 推导并解释中心流形不变性方程；
- 计算中心流形的有限阶近似；
- 解释为什么中心流形约化保留局部分岔信息；
- 理解中心流形约化与正规形理论之间的关系。


## 1. 动机与谱分解

考虑原点处平衡点附近的光滑常微分方程：

$$
x'
=
Ax+f^{(2)}(x)+f^{(3)}(x)+\cdots
=
f(x),
\qquad
x\in\mathbb R^d,
\qquad
f\in C^k,
\tag{10.1}
$$

其中

$$
A=Df(0),
\qquad
x_*=0.
$$

此前的局部分岔计算往往隐含假设：系统已经写在最小的临界维数中。当系统还含有额外的稳定或不稳定变量时，中心流形理论为这种降维提供了依据。

中心谱子空间是虚轴上各特征值所对应广义特征子空间的直和：

$
E^c(0)
=
\bigoplus_{\operatorname{Re}\lambda=0}
E_\lambda^{\mathrm{gen}}.
$

对于共轭复特征值，$E^c$ 应理解为相应的实不变子空间。

谱可以分成三组：

$$
\begin{aligned}
d_c &: \quad \operatorname{Re}\lambda_c=0,\\
d_s &: \quad \operatorname{Re}\lambda_s<0,\\
d_u &: \quad \operatorname{Re}\lambda_u>0.
\end{aligned}
$$

相应地，

$$
\mathbb R^d
=
E^s\oplus E^c\oplus E^u.
$$

若 $d_c=0$，平衡点是双曲的，可以应用稳定/不稳定流形定理。若 $d_c>0$，线性化不能决定中心方向上的行为，因此必须考虑非线性项。

图 23 展示了：

- 中心流形 $W^c$ 的基本几何结构；
- $E^c$、$E^s$ 与 $W^c$ 之间的关系；
- 中心流形可能具有的非唯一性与正则性限制。

图中画出的是平衡点附近的局部放大情形。


## 2. 中心流形定理

### 定理 10.1：中心流形定理

假设方程 (10.1) 的向量场属于 $C^k$。则存在一个局部不变的 $d_c$ 维 $C^k$ 流形

$$
W_{\mathrm{loc}}^c(0),
$$

满足

$$
T_0W_{\mathrm{loc}}^c(0)=E^c(0).
$$

此外：

- 任意对所有 $t\geq0$ 都停留在某个充分小邻域内的轨道，在 $t\to+\infty$ 时趋近 $W_{\mathrm{loc}}^c(0)$；
- 任意对所有 $t\leq0$ 都停留在该邻域内的轨道，在 $t\to-\infty$ 时趋近 $W_{\mathrm{loc}}^c(0)$。

这里“趋近中心流形”的含义是

$$
\operatorname{dist}
\left(
\phi_t(z),W_{\mathrm{loc}}^c(0)
\right)
\longrightarrow0.
$$

它不一定表示轨道收敛到中心流形上的某一个特定点。

### 定义 10.2

流形 $W_{\mathrm{loc}}^c(0)$ 称为局部中心流形，通常简写为

$$
W^c(0)
\qquad\text{或}\qquad
W^c.
$$

其证明采用不动点构造，与稳定/不稳定流形定理的证明有关，但更加细致。


## 3. 图 23(a) 的几何解释

图 23(a) 表示最简单的分解：

$$
\mathbb R^2=E^c\oplus E^s.
$$

水平方向 $E^c$ 对应实部为零的特征值；竖直方向 $E^s$ 对应实部为负的特征值。因此，线性化流在 $E^c$ 上既不指数收缩也不指数膨胀，而在 $E^s$ 上指数收缩。

### 3.1 从中心特征子空间到中心流形

线性系统的中心动力学位于直的子空间 $E^c$ 上。加入非线性项后，真正的不变集合通常弯曲成

$$
W^c_{\mathrm{loc}}
=
\{(x,h(x)):|x|\text{ 足够小}\}.
$$

它满足

$$
h(0)=0,
\qquad
Dh(0)=0,
\qquad
T_0W^c=E^c.
$$

因此 $W^c$ 在原点与 $E^c$ 相切，但一般并不等于 $E^c$。图中的形状只表示非线性弯曲，并不说明中心流形一定是抛物线。

### 3.2 附近轨道与流形上的动力学

稳定方向使轨道到 $W^c$ 的横向偏差衰减，所以图中流形上方的箭头向下、下方的箭头向上。严格地说，趋于零的是 $y-h(x)$ 或轨道到 $W^c$ 的距离，而不一定是坐标 $y$ 本身。

轨道一旦位于 $W^c$ 上，只要仍留在所选邻域内，局部不变性就保证它继续留在流形上，其后运动由中心流形上的约化方程决定。中心流形不是平衡点集合：轨道仍可能沿它趋向原点、远离原点或产生其他局部行为。


## 4. 中心流形的局限性

### 定理 10.3：中心流形的局限性

中心流形有两个重要限制：

1. 中心流形不一定唯一。
2. 即使向量场是解析的，中心流形也不一定解析。

下面的例子同时说明这两个结论。


## 5. 系统 (10.2)：非唯一性与非解析性

考虑

$$
\begin{aligned}
x'&=x^2,\\
y'&=-y.
\end{aligned}
\tag{10.2}
$$

原点处有一个中心特征值 $0$ 和一个稳定特征值 $-1$，故

$$
E^c=\{(x,0):x\in\mathbb R\},
\qquad
E^s=\{(0,y):y\in\mathbb R\}.
$$

局部中心流形可写成 $y=h(x)$，且 $h(0)=h'(0)=0$。

### 5.1 显式解

当 $x_0\neq0$ 时，分离变量给出

$$
\frac{dx}{x^2}=dt,
\qquad
-\frac1{x(t)}=t-\frac1{x_0},
$$

所以

$$
x(t)=\frac{1}{\frac1{x_0}-t}
=\frac{x_0}{1-x_0t},
\qquad
y(t)=y_0e^{-t}.
$$

当 $x_0=0$ 时，$x(t)\equiv0$。若 $x_0>0$，解在 $t=1/x_0$ 处发散；中心流形只讨论轨道留在小邻域内的局部部分。

### 5.2 轨道族

若图像 $y=h(x)$ 不变，则 $y'=h'(x)x'$，从而

$$
x^2h'(x)=-h(x).
$$

对非零分支积分可得下式，而零解可用 $p=0$ 一并表示：

$$
h(x)=p e^{1/x},
\qquad
p\in\mathbb R.
$$

同一结果也可由显式解消去 $t$ 得到：

$$
y(t)
=
y_0\exp\left(\frac1{x(t)}-\frac1{x_0}\right)
=
p e^{1/x(t)},
\qquad
p=y_0e^{-1/x_0}.
$$

因此 $p$ 是区分不同轨道的积分常数。

### 5.3 为什么左侧是一族曲线，右侧只能是 $x$ 轴

对每个 $p\in\mathbb R$，定义

$$
h_p(x)=
\begin{cases}
p e^{1/x},&x<0,\\
0,&x\ge0.
\end{cases}
$$

关键极限是

$$
e^{1/x}\to0
\quad(x\to0^-),
\qquad
e^{1/x}\to+\infty
\quad(x\to0^+).
$$

所以左侧任意有限的 $p$ 都能平滑接到原点；左侧也可以为零，这正是 $p=0$。右侧若 $p\neq0$ 则在原点附近发散，故只能取 $p=0$，即 $y=0$。

时间方向给出同一解释。若 $x_0<0$，则当 $t\to+\infty$ 时 $x(t)\to0^-$ 且 $y(t)\to0$，所以一族左侧轨道都可趋向原点。若 $x_0>0$，只有在 $t\to-\infty$ 时 $x(t)\to0^+$，但此时非零的 $y_0e^{-t}$ 会发散，因此只能有 $y_0=0$。根本原因是 $x'=x^2>0$ 使运动始终向右，而 $y'=-y$ 在正向时间收缩、反向时间膨胀。

右侧轨道的参数化为

$$
\gamma(t)
=
\left(
\frac{x_0}{1-x_0t},0
\right),
\qquad
-\infty<t<\frac1{x_0},
\quad x_0>0.
$$

其几何像正是 $\{(x,0):x>0\}$；它只在反向无限时间趋近原点，并不在有限时间穿过原点。因此

$$
W_p^c
=
\{(x,p e^{1/x}):x<0\}
\cup\{(0,0)\}
\cup\{(x,0):x>0\}
$$

由左侧轨道、原点平衡轨道和右侧轨道拼成。

### 5.4 光滑性、非唯一性与非解析性

当 $x<0$ 时，$e^{1/x}$ 的任意阶导数都具有形式

$$
P_n\left(\frac1x\right)e^{1/x},
$$

其中 $P_n$ 是多项式。由于指数项在 $x\to0^-$ 时比任意幂衰减得更快，所有左导数都趋于零；右侧函数恒为零。因此

$$
h_p\in C^\infty,
\qquad
h_p^{(n)}(0)=0
\quad(n\ge0).
$$

不同的 $p$ 给出不同的中心流形，所以中心流形不唯一。当 $p\neq0$ 时，$h_p$ 在原点的 Taylor 级数恒为零，但函数在 $x<0$ 上不为零，因此它不是解析函数。


## 6. 中心-双曲分解

经过适当的线性坐标变换，系统可以写成

$$
\begin{aligned}
x'&=Bx+F(x,y),
&
F&=\mathcal O(x^2,xy,y^2),\\
y'&=Cy+G(x,y),
&
G&=\mathcal O(x^2,xy,y^2),
\end{aligned}
\tag{10.3}
$$

其中

$$
x\in\mathbb R^{d_c},
\qquad
y\in\mathbb R^{d_s+d_u}.
$$

矩阵满足

$$
B\in\mathbb R^{d_c\times d_c},
\qquad
\operatorname{spec}(B)\subset i\mathbb R,
$$

以及

$$
C
\in
\mathbb R^{(d_s+d_u)\times(d_s+d_u)},
\qquad
\operatorname{spec}(C)\cap i\mathbb R=\varnothing.
$$

因此：

- $x$ 包含中心变量或临界变量；
- $y$ 包含稳定和不稳定变量，即非临界变量；
- $B$ 的全部特征值位于虚轴上；
- $C$ 在虚轴上没有特征值；
- $F$ 与 $G$ 只包含总次数至少为二的项。

记号

$$
\mathcal O(x^2,xy,y^2)
$$

是

$$
\mathcal O\left(\|(x,y)\|^2\right)
$$

的简写。


## 7. 中心流形约化

### 定理 10.4：中心流形约化

系统 (10.3) 在原点附近局部拓扑等价于

$$
\begin{aligned}
x'&=Bx+F(x,H(x)),\\
y'&=Cy,
\end{aligned}
\tag{10.4}
$$

其中

$$
W_{\mathrm{loc}}^c(0)
=
\left\{
(x,y)
\in
\mathbb R^{d_c}
\times
\mathbb R^{d_s+d_u}
:
y=H(x)
\right\},
$$

并且

$$
H(x)=\mathcal O(\|x\|^2).
$$

因此，原非线性系统在局部被分解为

$$
\boxed{
\text{非线性中心动力学}
+
\text{线性双曲动力学}.
}
$$

本质的非双曲动力学由

$$
x'=Bx+F(x,H(x))
$$

控制。

本讲后续考虑最重要的情形：

$$
d_u=0.
$$

此时 $y$ 只包含稳定变量。


## 8. 图像映射 $H$ 的含义

写成

$$
W^c_{\mathrm{loc}}
=
\{(x,H(x)):x\in E^c\}
$$

并不是说动力学把中心特征子空间映射到稳定子空间。这里用 $x$ 作为中心流形上的坐标，而 $H(x)$ 给出流形相对于切空间 $E^c$ 在稳定或双曲方向上的偏移量。

当 $d_u=0$ 时，

$$
H:E^c\longrightarrow E^s.
$$

若还存在不稳定方向，则图像映射的值域应为 $E^s\oplus E^u$。由于中心流形经过原点并与 $E^c$ 相切，

$$
H(0)=0,
\qquad
DH(0)=0,
$$

并且当 $H$ 至少为 $C^2$ 时，

$$
H(x)=\mathcal O(\|x\|^2).
$$

所以 $E^c=\{(x,0)\}$ 只是线性切空间，而非线性中心流形通常是弯曲的。精确函数 $H$ 满足非线性不变性方程，通常无法显式求出，因此实际中只计算满足约化动力学所需阶数的多项式近似 $h(x)$。


## 9. 为什么中心流形能够描述局部分岔

考虑带参数系统

$$
z'=f(z,\mu),
\qquad
f(0,0)=0.
$$

若线性化的所有特征值实部均非零，平衡点是双曲的，其局部相图对小扰动结构稳定。因此局部分岔要求双曲性失效，即某些特征值满足

$$
\operatorname{Re}\lambda=0.
$$

零特征值对应平衡点分岔的临界方向，一对 $\pm i\omega$ 对应 Hopf 分岔的临界方向；这些方向组成 $E^c$。

将系统写成

$$
\begin{aligned}
x'&=Bx+F(x,y,\mu),\\
y'&=Cy+G(x,y,\mu).
\end{aligned}
$$

若中心流形为 $y=H(x,\mu)$，则临界动力学约化为

$$
x'=Bx+F(x,H(x,\mu),\mu).
$$

双曲方向的收缩或膨胀已由线性部分决定；在只有稳定方向时，附近轨道先较快靠近中心流形，再由较慢的中心动力学控制。

附近的平衡点和足够小的周期轨道都是局部有界的完整轨道，因此位于中心流形上。这就是平衡点分岔和 Hopf 分岔能够在约化系统中研究的原因。例如 Hopf 正规形的径向方程为

$$
r'
=
\alpha(\mu)r
+\operatorname{Re}(\ell_1)r^3+\cdots.
$$

其中 $r>0$ 的平衡点对应小周期轨道，其存在性和稳定性由该方程决定。


## 10. 中心流形与正规形

两种约化依次完成不同任务：

$$
\text{原系统}
\longrightarrow
\text{中心流形约化}
\longrightarrow
\text{低维中心动力学}
\longrightarrow
\text{正规形}.
$$

中心流形约化用 $y=H(x)$ 消去独立的双曲变量；正规形再通过近恒等坐标变换消去可去掉的非共振项，保留决定局部分岔的系数。

不能简单地令 $y=0$，因为稳定变量会通过 $H(x)$ 改变中心方程的系数。例如，在所略项不影响三阶计算时，

$$
\begin{aligned}
x'&=a x^3+bxy+\cdots,\\
y'&=-\lambda y+c x^2+\cdots,
\qquad
\lambda>0.
\end{aligned}
$$

设

$$
y=H(x)=\alpha x^2+\cdots.
$$

不变性方程在二阶给出 $\alpha=c/\lambda$，代回中心方程便得到

$$
x'
=
\left(
a+\frac{bc}{\lambda}
\right)x^3+\cdots.
$$

因此稳定变量本身不产生临界方向，却会改变有效正规形系数，必须先求 $H$ 或足够精确的近似。实际中只需计算到能确定首个关键非线性项的阶数；中心流形即使不唯一，所需的有限阶展开和正规形系数仍是确定的。


## 11. 中心流形不变性方程

假设

$$
W^c=\{(x,y):y=h(x)\}.
$$

沿图像上的轨道，

$$
y'=Dh(x)x'.
$$

利用

$$
x'=Bx+F(x,h(x)),
$$

得到

$$
y'
=
Dh(x)
\left[
Bx+F(x,h(x))
\right].
$$

原系统的 $y$ 方程给出

$$
y'=Ch(x)+G(x,h(x)).
$$

令两者相等：

$$
Dh(x)
\left[
Bx+F(x,h(x))
\right]
=
Ch(x)+G(x,h(x)).
$$

### 定义 10.5

定义不变性残差

$$
\boxed{
\mathcal N[h](x)
=
Dh(x)
\left[
Bx+F(x,h(x))
\right]
-
Ch(x)
-
G(x,h(x)).
}
$$

中心流形不变性方程为

$$
\boxed{
\mathcal N[h](x)=0,
}
\tag{10.5}
$$

并附带条件

$$
h(0)=0,
\qquad
Dh(0)=0.
$$

当 $d_c>1$ 时，这通常是非线性一阶偏微分方程。即使中心维数为一，也很少能够得到闭式解。


## 12. 一个简短的系数匹配例子

考虑

$$
\begin{aligned}
x'&=x^2,\\
y'&=-y+x^2.
\end{aligned}
$$

$x$ 为中心变量，$y$ 为稳定变量。设

$$
y=h(x)
=
ax^2+bx^3+\mathcal O(x^4).
$$

不变性方程为

$$
h'(x)x^2=-h(x)+x^2.
$$

因为

$$
h'(x)
=
2ax+3bx^2+\mathcal O(x^3),
$$

左侧为

$$
h'(x)x^2
=
2ax^3+\mathcal O(x^4).
$$

右侧为

$$
-h(x)+x^2
=
(1-a)x^2-bx^3+\mathcal O(x^4).
$$

比较二次项得到

$$
a=1.
$$

比较三次项得到

$$
2a=-b,
$$

所以

$$
b=-2.
$$

因此

$$
\boxed{
h(x)
=
x^2-2x^3+\mathcal O(x^4).
}
$$

该近似描述中心流形在稳定方向上的偏移，而不是描述一个把中心子空间动力学映射到稳定子空间的映射。


## 13. 中心流形的近似

### 定理 10.6：近似定理

假设 $h$ 满足

$$
h(0)=0,
\qquad
Dh(0)=0,
$$

并且对某个 $r>1$，

$$
\mathcal N[h](x)
=
\mathcal O(\|x\|^r)
\qquad
(x\to0).
$$

那么某个精确的中心流形图像 $y=H(x)$ 满足

$$
\boxed{
H(x)
=
h(x)+\mathcal O(\|x\|^r).
}
\tag{10.6}
$$

因此，

$$
\boxed{
r\text{ 阶不变性残差会产生 }r\text{ 阶图像误差}.
}
$$

该定理使形式上的系数匹配成为严格的近似方法。


## 14. 近似定理的详细证明

证明思路是把中心流形化为不动点问题，再说明 $p$ 阶不变性残差只需 $p$ 阶修正。以下假设 $d_u=0$，即 $C$ 的谱严格位于左半平面。

### 14.1 Lyapunov--Perron 表示

若 $y=H(x)$ 是中心流形，且 $x(t,x_0,H)$ 满足

$$
x'=Bx+F(x,H(x)),
\qquad
x(0,x_0,H)=x_0,
$$

则常数变易公式与稳定性给出

$$
H(x_0)
=
\int_{-\infty}^0
e^{-Cs}
G\left(
x(s,x_0,H),
H(x(s,x_0,H))
\right)\,ds.
\tag{10.7}
$$

把候选图像 $\Phi$ 代入右端便定义算子 $K$，所以真正的中心流形满足

$$
KH=H.
$$

积分从 $-\infty$ 开始，是为了消去自由的稳定齐次模态，只保留由中心运动强迫产生的稳定响应。

### 14.2 近似图像与残差

设候选函数 $h$ 满足

$$
h(0)=0,
\qquad
Dh(0)=0,
\qquad
\mathcal N(h(x))=\mathcal O(\|x\|^p).
$$

取一个在原点附近等于 $h$ 的紧支撑 $C^1$ 延拓 $\theta$，并定义

$$
N(x)
=
D\theta(x)
\bigl[Bx+F(x,\theta(x))\bigr]
-C\theta(x)-G(x,\theta(x)).
$$

于是原点附近有 $N(x)=\mathcal O(\|x\|^p)$。延拓只用于保证后面的轨道与估计有界，不改变局部问题。

### 14.3 残差如何变成不动点误差

令

$$
x_\theta'
=
Bx_\theta+F(x_\theta,\theta(x_\theta)),
\qquad
x_\theta(0)=x_0.
$$

由链式法则和乘积法则，

$$
\begin{aligned}
\frac{d}{dt}
\left[e^{-Ct}\theta(x_\theta(t))\right]
&=
e^{-Ct}
\left[D\theta(x_\theta(t))x_\theta'(t)
-C\theta(x_\theta(t))\right]\\
&=
e^{-Ct}
\left[G(x_\theta(t),\theta(x_\theta(t)))
+N(x_\theta(t))\right].
\end{aligned}
$$

先从有限的 $T<0$ 积分到 $0$，再令 $T\to-\infty$。由于 $C$ 稳定且 $\theta$ 有界，边界项 $e^{-CT}\theta(x_\theta(T))$ 趋于零；含 $G$ 的积分正是 $K\theta(x_0)$。因此得到精确恒等式

$$
K\theta(x_0)-\theta(x_0)
=
-
\int_{-\infty}^0
e^{-Cs}N(x_\theta(s))\,ds.
\tag{14.1}
$$

这说明不动点误差就是不变性残差经过稳定传播后的积分。

### 14.4 为什么积分仍是 $p$ 阶

对 $s\le0$，可取常数 $M_C,M_x,\alpha>0$ 及充分小的 $\eta>0$，使

$$
\|e^{-Cs}\|\le M_Ce^{\alpha s},
\qquad
\|x_\theta(s)\|\le M_xe^{-\eta s}\|x_0\|.
$$

由 $N(x)=\mathcal O(\|x\|^p)$，

$$
\|N(x_\theta(s))\|
\le
M_NM_x^p e^{-p\eta s}\|x_0\|^p.
$$

选择 $p\eta<\alpha$，式 (14.1) 给出

$$
\begin{aligned}
\|K\theta(x_0)-\theta(x_0)\|
&\le
M_CM_NM_x^p\|x_0\|^p
\int_{-\infty}^0 e^{(\alpha-p\eta)s}\,ds\\
&=
\frac{M_CM_NM_x^p}{\alpha-p\eta}\|x_0\|^p.
\end{aligned}
$$

故

$$
K\theta-\theta=\mathcal O(\|x\|^p).
$$

稳定积分只改变常数，不改变关于 $\|x\|$ 的幂次。

### 14.5 从近似图像修正到真实中心流形

写

$$
H=\theta+w,
\qquad
Lw:=K(\theta+w)-\theta.
$$

在函数类

$$
\mathcal L
=
\{w:\|w(x)\|\le c\|x\|^p\}
$$

上，标准的截断、指数和 Lipschitz 估计表明：邻域足够小时，$L$ 将 $\mathcal L$ 映入自身并成为压缩映射。因此存在不动点 $w_*$，且

$$
w_*(x)=\mathcal O(\|x\|^p).
$$

于是 $H=\theta+w_*$ 满足 $KH=H$，并在原点附近满足

$$
H(x)=h(x)+\mathcal O(\|x\|^p).
$$

这就说明：若候选函数的不变性残差为 $p$ 阶，它与某个真实中心流形的误差也是 $p$ 阶。


## 15. 实际的有限阶计算方法

计算中心流形近似时，可采用以下步骤：

1. 将平衡点平移到原点。
2. 计算 Jacobian 矩阵。
3. 识别中心、稳定与不稳定特征值。
4. 作线性变换，将中心变量与双曲变量分离。
5. 将中心流形写成 $y=h(x)$。
6. 使用相切条件

   $$
   h(0)=0,
   \qquad
   Dh(0)=0.
   $$

7. 选取多项式设定：

   $$
   h(x)
   =
   h^{(2)}(x)+h^{(3)}(x)+\cdots+h^{(m)}(x).
   $$

8. 将该设定代入不变性方程。
9. 比较相同次数的系数。
10. 将近似代入中心方程。
11. 分析约化动力学中的第一个非零项。

### 截断阶数的限制

截断阶数可根据问题选择，但并非完全任意。

首先，因为

$$
h(0)=0,
\qquad
Dh(0)=0,
$$

展开从二阶开始。

其次，向量场和所选择的中心流形必须有足够的可微性。例如，$C^3$ 正则性不能支持五阶 Taylor 展开。

第三，所需阶数由约化动力学决定。通常只计算到足以找到控制稳定性或分岔的第一个非零项。若预期的首项消失，就必须继续计算。

对多项式或解析向量场，形式系数计算通常可以继续到更高的有限阶。但解析向量场不一定具有解析或 $C^\infty$ 的中心流形。Taylor 近似只能在中心流形正则性所允许的阶数内得到严格保证。


## 16. 例 10.7

考虑

$$
\begin{aligned}
x'&=xy,\\
y'&=-y+\mu x^2.
\end{aligned}
$$

特征值为 $0$ 和 $-1$，所以 $x$ 是中心变量，$y$ 是稳定变量。令中心流形为 $y=h(x)$，不变性方程为

$$
h'(x)\,x\,h(x)+h(x)-\mu x^2=0.
\tag{16.1}
$$

### 16.1 为什么展开从二阶开始

由

$$
h(0)=0,
\qquad
h'(0)=0
$$

可知 Taylor 展开没有常数项和一次项。取三阶设定

$$
h(x)=a_2x^2+a_3x^3+\mathcal O(x^4).
$$

此时

$$
h'(x)\,x\,h(x)
=
2a_2^2x^4+\mathcal O(x^5),
$$

所以该乘积在二、三阶都没有贡献。代入式 (16.1) 得

$$
(a_2-\mu)x^2+a_3x^3+\mathcal O(x^4)=0,
$$

故

$$
a_2=\mu,
\qquad
a_3=0,
\qquad
h(x)=\mu x^2+\mathcal O(x^4).
$$

二阶计算已经确定 $a_2=\mu$；做到三阶的作用是确认 $a_3=0$，从而也确认约化方程没有四次项。

### 16.2 为什么选择三阶，能否计算到四阶

截断阶数由所需的约化动力学决定，并非固定为三阶。若再设

$$
h(x)
=
a_2x^2+a_3x^3+a_4x^4+\mathcal O(x^5),
$$

则四阶系数满足

$$
2\mu^2+a_4=0,
$$

所以

$$
h(x)
=
\mu x^2-2\mu^2x^4+\mathcal O(x^5).
$$

但研究当前稳定性时，三阶计算已足够，因为中心流形上的方程为

$$
x'=xh(x)
=
\mu x^3+\mathcal O(x^5).
$$

因此 $\mu<0$ 时原点局部渐近稳定，$\mu>0$ 时原点不稳定。一般地，$h$ 中的 $a_kx^k$ 会在本例的约化方程中变成 $a_kx^{k+1}$。所需阶数取决于要确定的首个关键系数，并受向量场和中心流形光滑性的限制；四阶可以计算，但对这里的稳定性结论不是必需的。


## 17. 例 10.8：化为标准坐标

考虑

$$
\begin{aligned}
x'&=y-1,\\
y'&=-(y-1)+\mu x^2+x(y-1).
\end{aligned}
\tag{10.9}
$$

平衡点为

$$
(x,y)=(0,1).
$$

先令

$$
\xi=x,
\qquad
\eta=y-1
$$

将其平移到原点。于是

$$
\begin{aligned}
\xi'&=\eta,\\
\eta'&=-\eta+\mu\xi^2+\xi\eta.
\end{aligned}
\tag{10.10}
$$

Jacobian 矩阵为

$$
A=
\begin{pmatrix}
0&1\\
0&-1
\end{pmatrix},
$$

其特征值为 $0$ 和 $-1$。

使用变换

$$
\begin{pmatrix}
\xi\\
\eta
\end{pmatrix}
=
T
\begin{pmatrix}
u\\
v
\end{pmatrix},
\qquad
T=
\begin{pmatrix}
1&1\\
0&-1
\end{pmatrix}
=
T^{-1}.
$$

所以

$$
\xi=u+v,
\qquad
\eta=-v.
$$

定义

$$
Q(u,v)
=
\mu(u+v)^2-(u+v)v.
$$

变换后的系统为

$$
\begin{aligned}
u'&=Q(u,v),\\
v'&=-v-Q(u,v).
\end{aligned}
$$

现在系统已经处于标准中心-稳定形式。令

$$
v=h(u)
$$

并像例 10.7 一样应用不变性方程即可。本讲将最后的中心流形计算留作练习。


## 18. 离散时间系统

对于映射也有类似的中心流形结论。中心方向对应单位圆上的乘子。

考虑分裂形式的映射

$$
\begin{aligned}
x_{n+1}&=Bx_n+F(x_n,y_n),\\
y_{n+1}&=Cy_n+G(x_n,y_n).
\end{aligned}
$$

若中心流形为

$$
y=h(x),
$$

则图像不变性要求

$$
\boxed{
h\left(
Bx+F(x,h(x))
\right)
=
Ch(x)+G(x,h(x)).
}
$$

与连续时间的不变性方程不同，这里出现的是函数复合，而不是导数。

## 20. 主要逻辑链

中心流形方法可以概括为

$$
\boxed{
\text{谱分解}
\longrightarrow
\text{中心流形}
\longrightarrow
\text{不变性方程}
\longrightarrow
\text{有限阶近似}
\longrightarrow
\text{约化动力学}
\longrightarrow
\text{正规形}
\longrightarrow
\text{稳定性与分岔}.
}
$$

核心原理为：

$$
\boxed{
\begin{aligned}
&\text{双曲方向主要由其线性动力学控制；}\\
&\text{双曲性的丧失发生在中心方向；}\\
&\text{中心流形包含临界的局部动力学；}\\
&\text{稳定变量通过图像 }y=H(x)\text{ 改变约化系数；}\\
&\text{有限阶近似足以研究局部分岔；}\\
&\text{正规形只保留本质的共振非线性项。}
\end{aligned}
}
$$
