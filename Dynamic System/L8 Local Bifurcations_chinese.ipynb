## 1. 主要思想与学习目标（Main idea and learning goals）

依赖参数的常微分方程（ordinary differential equation, ODE）具有如下形式

$$
x' = f(x,p), \qquad x \in \mathbb{R}^d,\quad p \in \mathbb{R}^m.
$$

当参数的微小变化使相图（phase portrait）不再与先前的相图拓扑等价（topologically equivalent）时，就会发生**局部分岔（local bifurcation）**。本讲中的典型机制是**双曲性丧失（loss of hyperbolicity）**：对于流（flow），线性化（linearization）的某个特征值到达虚轴；对于映射（map），某个乘子（multiplier）到达单位圆（unit circle）。

本节阐述四个反复出现的思想：

1. 检测双曲性在何处丧失。
2. 施加非退化性（non-degeneracy）和横截性（transversality）条件，以确保该事件具有泛型性（genericity）。
3. 通过坐标变换和参数变换化简系统。
4. 将所得系统与动力学易于分析的低阶**正规形（normal form）**进行比较。

对于流，主要的余维一（codimension-one）情形是折叠分岔（fold bifurcation）和霍普夫分岔（Hopf bifurcation）。对于映射，本节讨论折叠分岔、翻转分岔（flip bifurcation，即倍周期分岔（period-doubling bifurcation））和奈马克–萨克分岔（Neimark-Sacker bifurcation）。叉形分岔（pitchfork bifurcation）和跨临界分岔（transcritical bifurcation）则作为重要的对称性驱动或非泛型（non-generic）替代情形出现。

### 澄清：分岔何时必须具有零特征值？（Clarification: when must a bifurcation have a zero eigenvalue?）

**平衡点的局部分岔（local bifurcation of an equilibrium）**的一般要求是双曲性丧失，而不一定要求存在等于零的特征值。

对于连续时间系统（continuous-time system）

$$
\dot{x}=f(x,p), \qquad A(p)=D_xf(x^*(p),p),
$$

当每个特征值都满足 $\operatorname{Re}\lambda_j\neq0$ 时，平衡点是双曲的（hyperbolic）。因此，平衡点发生局部分岔的一个必要谱条件是：至少有一个临界特征值满足

$$
\operatorname{Re}\lambda_j=0
$$

在一维常微分方程中，特征值为实数，因此双曲性丧失确实意味着 $\lambda=0$。这是折叠分岔、跨临界分岔和叉形分岔的临界条件。然而，在二维或更高维系统中，霍普夫分岔（Hopf bifurcation）满足

$$
\lambda_{1,2}=\pm i\omega, \qquad \omega>0,
$$

它们是实部为零的非零特征值。

对于离散映射（discrete map）$x_{k+1}=g(x_k,p)$，当某个乘子到达单位圆时，双曲性丧失：

$$
|\mu|=1.
$$

主要情形是：折叠分岔对应 $\mu=1$，翻转分岔对应 $\mu=-1$，奈马克–萨克分岔（Neimark-Sacker bifurcation）对应 $\mu=e^{\pm i\theta}$。乘子 $\mu=0$ 通常表示强收缩，而不是临界状态。

最后，这一判据针对的是依附于平衡点或不动点（fixed point）的局部分岔。同宿分岔（homoclinic bifurcation）或异宿分岔（heteroclinic bifurcation）等全局事件可能在所涉及的平衡点仍保持双曲性时发生，因为发生变化的是不变流形（invariant manifolds）的全局配置，而不是局部谱。

## 2. 结构性术语：等价、分层、分岔图、展开与余维（Structural language: equivalence, strata, diagrams, unfoldings, and codimension）

考虑两个参数化常微分方程

$$
x'=f(x,p), \qquad y'=g(y,q).
$$

如果参数空间之间存在同胚映射（homeomorphism）$q=\xi(p)$，并且对于每一对固定的对应参数，两个相图均拓扑等价，那么它们就是**纤维拓扑等价的（fiber topologically equivalent）**。因此，可以在允许重新标记参数的同时，逐纤维检验等价性。

**层（stratum）** $S_{p_0}$ 是包含 $p_0$ 的最大连通参数区域，在该区域上所有相图都拓扑等价。**分岔图（bifurcation diagram）**将参数空间划分为这些层，并为每一层附上一个代表性相图。

**展开（unfolding）**是一个持久地包含某种分岔的参数化族。**这意味着该参数族会使分岔发生。** 它的**余维（codimension）**是这种展开所需的最少参数个数。因此，通过改变一个参数通常就会遇到余维一分岔。

在讲义的分岔图中，实线平衡点分支是稳定的，虚线分支是不稳定的。

## 3. 叉形分岔（Pitchfork bifurcation）

入门示例为

$$
x'=px-x^3=x(p-x^2).
$$

平衡点满足 $x(p-x^2)=0$。因此：

- 当 $p\leq 0$ 时，唯一的平衡点是 $x^*=0$。它是全局渐近稳定的（globally asymptotically stable）。
- 当 $p>0$ 时，原点仍是一个平衡点，但会变得不稳定，同时出现两个稳定平衡点：

$$
x_{\pm}=\pm\sqrt{p}.
$$

这是一个**超临界叉形分岔（supercritical pitchfork bifurcation）**。它与反射对称性（reflection symmetry）$x\mapsto -x$ 密切相关：如果将 $x(t)$ 变换为 $-x(t)$，方程仍保持相同的形式。该对称性迫使两个非零平衡点分支成对出现。

关键的定性事件并不只是平衡点的出现，而是系统从一个稳定对称态转变为两个稳定的对称破缺态（symmetry-broken states），同时原来的状态在 $p=0$ 处失去稳定性。

## 4. 折叠（鞍结）分岔：基本示例（Fold (saddle-node) bifurcation: basic example）

标准示例为

$$
x'=p-x^2.
$$

当 $p>0$ 时，平衡点为

$$
x_{\pm}=\pm\sqrt{p}.
$$

在平衡点处的线性化为

$$
X' = \partial_x f(x_{\pm},p)X=-2x_{\pm}X.
$$

因此，$x_+$ 是稳定的，而 $x_-$ 是不稳定的。在 $p=0$ 处，两个平衡点在原点相撞，特征值变为零，双曲性丧失。当 $p<0$ 时，不再存在平衡点。

方程 $x'=p+x^2$ 具有相同的局部分岔类型，但方向相反。二者都称为**折叠分岔（fold bifurcation）**或**鞍结分岔（saddle-node bifurcation）**。该参数族是余维一展开，因为足够小的光滑扰动仍会表现出相同的局部过程，即一对稳定—不稳定平衡点的产生或消失。

**注：叉形分岔与折叠分岔是不同的分岔；叉形分岔始终具有平衡点 x=0，而在折叠分岔中，平衡点 $x=+-\sqrt{q}$ 会随着 p 的改变而消失和出现**

## 5. 为什么最低阶项决定折叠分岔（Why the lowest-order terms determine the fold）

引理 8.7（Lemma 8.7）指出，在 $(x,p)=(0,0)$ 附近，

$$
x'=p\pm x^2+O(x^3)
$$

局部拓扑等价于

$$
x'=p\pm x^2.
$$

记号 $a(x)=O(b(x))$ 在 $x\to0$ 时表示 $|a(x)|/|b(x)|$ 保持有界。余项也可以包含适当的依赖参数的高阶项，例如 $px^3$。

对于负号情形，写成

$$
y'=g(y,p)=p-y^2+\psi(y,p), \qquad \psi=O(y^3).
$$

由于 $\partial_p g(0,0)=1$，隐函数定理（Implicit Function Theorem）给出附近的一条平衡点曲线

$$
\mathcal{M}=\{(y,p):p=y^2+G(y)\}, \qquad G(y)=O(y^3).
$$

当 $p>0$ 时，它的两个平衡点仍接近 $\pm\sqrt{p}$。依赖参数的同胚映射可以将正规形的两个平衡点映射到扰动方程的两个平衡点，同时保持它们的次序和局部流方向。因此，高阶项会使分岔图弯曲，但不会改变其局部拓扑类型。

## 6. 流的折叠分岔定理（Fold bifurcation theorem for flows）

对于一个光滑的一维单参数常微分方程

$$
x'=f(x,p),
$$

定理 8.8（Theorem 8.8）假设

$$
f(0,0)=0, \qquad \partial_x f(0,0)=0,
$$

并满足泛型性条件（genericity conditions）

$$
\partial_{xx}f(0,0)\neq0, \qquad \partial_p f(0,0)\neq0.
$$

那么，在原点附近，该系统拓扑等价于以下两个正规形之一

$$
y'=q\pm y^2.
$$

这些条件的含义：

- $f(0,0)=0$：该临界点是平衡点。
- $\partial_x f(0,0)=0$：线性特征值为零，因此双曲性丧失。
- $\partial_{xx}f(0,0)\neq0$：第一个重要的非线性项确实是二次项；这是非退化性条件（non-degeneracy condition）。
- $\partial_p f(0,0)\neq0$：参数以非零速度穿过分岔集；这是横截性条件（transversality condition）。

最后两个条件共同确保折叠分岔是鲁棒的，并且余维为一。


**前几个条件确保平衡点失去双曲性，$\partial_{xx}f(0,0)\neq0$ 确保高阶项能够形成折叠，最后一个条件确保存在某个关于 $x(p)=p$ 的方程，作为新的平衡点**

### 详细推导以及每个变换的目的

目标是证明：任何满足折叠条件（fold conditions）的系统，都可以通过状态变量与参数的局部可逆变换化简为

$$
y'=q\pm y^2.
$$

每个变换都会消去泰勒展开中的一个非本质特征，同时保持局部相图（local phase portrait）不变。

#### 步骤 0——非双曲平衡点（nonhyperbolic equilibrium）附近的泰勒展开

写成

$$
f(x,p)=f_0(p)+f_1(p)x+f_2(p)x^2+O(x^3).
$$

折叠假设意味着

$$
f_0(0)=0, \qquad f_1(0)=0, \qquad f_2(0)\neq0, \qquad f_0'(0)\neq0.
$$

这里，$f_1(0)=0$ 表示双曲性丧失（loss of hyperbolicity），$f_2(0)\neq0$ 保证二次项是第一个起作用的状态非线性项（state nonlinearity），而 $f_0'(0)\neq0$ 则保证参数使矢量场穿越临界构型（critical configuration）。

#### 步骤 1——平移状态变量并消去线性项

引入参数依赖平移

$$
y=x+\delta(p).
$$

等价地，

$$
\boxed{x=y-\delta(p).}
$$

沿每条轨线参数均保持不变，因此 $p'=0$，且 $\delta(p)$ 不随时间变化。于是

$$
y'=x'.
$$

将 $x=y-\delta$ 代入泰勒展开：

$$
\begin{aligned}
y'=x'
&=f_0(p)+f_1(p)(y-\delta)+f_2(p)(y-\delta)^2
+O\bigl((y-\delta)^3\bigr)\\
&=C(p,\delta)+L(p,\delta)y+Q(p,\delta)y^2+O(y^3).
\end{aligned}
$$

这三个系数的局部形式为

$$
\begin{aligned}
C(p,\delta)
&=f_0(p)-f_1(p)\delta+f_2(p)\delta^2+O(\delta^3),\\
L(p,\delta)
&=f_1(p)-2f_2(p)\delta+\delta^2\psi(p,\delta),\\
Q(p,\delta)
&=f_2(p)+O(\delta),
\end{aligned}
$$

其中 $\psi$ 是光滑函数。为了消去关于 $y$ 的线性项，选择 $\delta$ 使得

$$
F(p,\delta):=L(p,\delta)=0.
$$

在分岔点处，

$$
F(0,0)=0,
$$

并且

$$
\partial_\delta F(0,0)=-2f_2(0)\neq0.
$$

因此，隐函数定理给出唯一的光滑函数 $\delta=\delta(p)$，满足 $\delta(0)=0$ 和 $F(p,\delta(p))=0$。

**此步骤的目的。** 该平移将坐标中心置于矢量场对状态变量的导数为零的参数依赖点。它消去了关于 $y$ 的线性项，并显露出二次折叠几何（quadratic fold geometry）。这并不意味着每个邻近平衡点都被移到 $y=0$；当 $q>0$ 时，正规形仍在原点之外有两个平衡点。

因为 $\delta(p)=O(p)$，将选定的 $\delta(p)$ 代入 $C$ 和 $Q$ 可得

$$
C(p,\delta(p))=f_0'(0)p+O(p^2)
$$

以及

$$
Q(p,\delta(p))=f_2(0)+O(p).
$$

因此，变换后的方程为

$$
y'=\bigl[f_0'(0)p+O(p^2)\bigr]
+\bigl[f_2(0)+O(p)\bigr]y^2+O(y^3).
$$

#### 步骤 2——以常数项作为新参数

定义

$$
\alpha(p):=C(p,\delta(p))=f_0'(0)p+p^2\mu(p),
$$

并令

$$
\boxed{q=\alpha(p).}
$$

由于

$$
\alpha(0)=0, \qquad \alpha'(0)=f_0'(0)=\partial_pf(0,0)\neq0,
$$

反函数定理给出光滑局部逆映射 $p=p(q)$。

**此步骤的目的。** 原参数可能通过一个复杂的光滑函数进入常数项。将 $p$ 替换为 $q=\alpha(p)$，可以拉直这种依赖关系，使常数项恰好成为 $q$。因为该变换局部可逆，所以它只会重新标记邻近的系统族，不会将分岔的两侧合并。

代入 $p=p(q)$ 后，定义

$$
b(q):=Q(p(q),\delta(p(q))).
$$

于是

$$
y'=q+b(q)y^2+O(y^3), \qquad b(0)=f_2(0)\neq0.
$$

#### 步骤 3——规范化二次项系数的大小

因为 $b(0)\neq0$，连续性意味着 $b(q)$ 对充分小的 $q$ 具有恒定符号。令

$$
B(q)=|b(q)|>0
$$

并引入

$$
\widetilde y=B(q)y, \qquad \widetilde q=B(q)q.
$$

对于固定参数 $q$，$B(q)$ 不随时间变化，因此

$$
\widetilde y'=B(q)y'.
$$

利用 $y=\widetilde y/B(q)$ 可得

$$
\begin{aligned}
\widetilde y'
&=B(q)q+\frac{b(q)}{B(q)}\widetilde y^2+O(\widetilde y^3)\\
&=\widetilde q+s\widetilde y^2+O(\widetilde y^3),
\end{aligned}
$$

其中

$$
s=\frac{b(q)}{|b(q)|}=\operatorname{sign}(b(0))\in\{-1,+1\}.
$$

**此步骤的目的。** 该缩放消去了二次项系数的数值大小，只留下其符号，这表明折叠正规形有两个定向版本（oriented versions）。由于 $B(q)>0$，状态变量缩放是可逆的；又因为参数缩放在零点处的导数为 $B(0)>0$，所以参数缩放也是局部可逆的。

去掉波浪号后得到

$$
y'=q+sy^2+O(y^3), \qquad s=\pm1.
$$

#### 步骤 4——消去拓扑上无关的高阶项

引理 8.7 指出

$$
y'=q\pm y^2+O(y^3)
$$

局部拓扑等价于

$$
\boxed{y'=q\pm y^2.}
$$

最后这一步并不是声称高阶项在代数意义上为零，而是说它们只会弯曲局部平衡分支，并且可以通过局部同胚（local homeomorphism）消去，而不改变邻近平衡点的数量、稳定性或排列方式。

因此，完整逻辑为

$$
\text{general Taylor expansion}
\longrightarrow
\text{remove the linear state term}
\longrightarrow
\text{straighten the parameter}
\longrightarrow
\text{normalize the quadratic term}
\longrightarrow
\text{discard topologically inessential higher-order terms}.
$$

## 8. 跨临界分岔（transcritical bifurcation）与稳定性交换（exchange of stability）

如果去掉折叠横截性条件（fold transversality condition），就可能出现另一种分岔。讲义采用

$$
x'=x(p-x).
$$

平衡分支（equilibrium branches）为

$$
x^*_1=0, \qquad x^*_2=p.
$$

在 $x=0$ 处，导数为 $\partial_x f(0,p)=p$，因此原点在 $p<0$ 时稳定，在 $p>0$ 时不稳定。在 $x=p$ 处，

$$
\partial_x f(p,p)=p-2p=-p,
$$

因此该分支在 $p<0$ 时不稳定，在 $p>0$ 时稳定。两个分支在原点相交并交换稳定性。这就是**跨临界分岔（transcritical bifurcation）**。

该例说明了泛型性条件（genericity conditions）为何重要：仅有双曲性丧失（loss of hyperbolicity）并不能确定分岔类型。非线性项的结构以及参数进入方程的方式，决定了平衡点是碰撞后消失、彼此相交，还是对称分裂。

## 9. 霍普夫分岔定理（Hopf bifurcation theorem）

在泛型二维单参数流（generic two-dimensional, one-parameter flow）中，一对共轭复特征值（complex-conjugate eigenvalue pair）可能穿过虚轴。假设

$$
x'=f(x,p), \qquad x\in\mathbb{R}^2,\quad p\in\mathbb{R},
$$

其中 $f(0,0)=0$，并且

$$
\lambda_{1,2}(p)=\mu(p)\pm i\omega(p), \qquad \mu(0)=0,\quad \omega(0)>0.
$$

在泛型性假设（genericity assumptions）下，该系统局部拓扑等价于

$$
\begin{pmatrix}y_1'\\y_2'\end{pmatrix}
=
\begin{pmatrix}q&-1\\1&q\end{pmatrix}
\begin{pmatrix}y_1\\y_2\end{pmatrix}
+l_1(y_1^2+y_2^2)\begin{pmatrix}y_1\\y_2\end{pmatrix},
$$

其中非零的三次项系数 $l_1\neq0$ 称为**第一李雅普诺夫系数（first Lyapunov coefficient）**。讲义用符号 $l_1=\pm1$ 写出两个拓扑正规形（topological normal forms）。

横截性条件（transversality condition）是 $\mu'(0)\neq0$：特征值对的实部以非零速度穿过虚轴。非退化条件（non-degeneracy condition）$l_1\neq0$ 保证三次项控制邻近周期轨道（periodic orbit）的产生或消失。霍普夫分岔（Hopf bifurcation）的余维为一（codimension one）。

## 10. 霍普夫动力学的复数形式和极坐标形式（Hopf dynamics in complex and polar form）

令 $z=y_1+iy_2$。霍普夫正规形（Hopf normal form）变为

$$
z'=(q+i)z+l_1z|z|^2.
$$

使用 $z=re^{i\varphi}$ 可得极坐标方程（polar equations）

$$
r'=r(q+l_1r^2), \qquad \varphi'=1.
$$

角向方程描述匀速旋转。所有分岔信息都包含在标量径向方程（scalar radial equation）中。除 $r=0$ 外，还存在一条周期轨道（periodic orbit），其存在条件为

$$
r^2=-\frac{q}{l_1}>0.
$$

因此：

- **超临界霍普夫分岔（Supercritical Hopf，$l_1<0$）：** 平衡点在 $q<0$ 时稳定，在 $q>0$ 时不稳定。当 $q>0$ 时，存在一条小的**稳定**周期轨道，其半径为 $r=\sqrt{-q/l_1}$。
- **亚临界霍普夫分岔（Subcritical Hopf，$l_1>0$）：** 平衡点在 $q<0$ 时稳定，在 $q>0$ 时不稳定。当 $q<0$ 时，存在一条小的**不稳定**周期轨道。

对于一般矢量场（general vector field），可以利用直至三阶的导数计算 $l_1$。它的符号决定分岔周期轨道的临界性质（criticality）和稳定性。

### 霍普夫极坐标形式的详细推导（Detailed derivation of the Hopf polar form）

从复数形式的霍普夫正规形开始

$$
z'=(q+i)z+l_1z|z|^2.
$$

写成

$$
z=re^{i\varphi}, \qquad r\geq0.
$$

因为 $|z|=r$，右端变为

$$
(q+i)re^{i\varphi}+l_1re^{i\varphi}r^2
=re^{i\varphi}(q+i+l_1r^2).
$$

变量 $r=r(t)$ 和 $\varphi=\varphi(t)$ 都依赖于时间。对极坐标表示（polar representation）求导可得

$$
\begin{aligned}
z'&=(re^{i\varphi})'\\
&=r'e^{i\varphi}+ir\varphi'e^{i\varphi}\\
&=e^{i\varphi}(r'+ir\varphi').
\end{aligned}
$$

令 $z'$ 的两个表达式相等，并约去非零因子 $e^{i\varphi}$，得到

$$
r'+ir\varphi'=r(q+l_1r^2)+ir.
$$

两个复数相等，意味着其实部和虚部分别相等。因此

$$
r'=r(q+l_1r^2), \qquad r\varphi'=r.
$$

当 $r>0$ 时，除以 $r$ 得到 $\varphi'=1$。在 $r=0$ 处，角度没有定义，但径向方程仍然有效。因此

$$
\boxed{r'=r(q+l_1r^2), \qquad \varphi'=1.}
$$

从实变量出发也能直接得到相同结果。由 $r^2=y_1^2+y_2^2$，

$$
rr'=y_1y_1'+y_2y_2',
$$

且旋转交叉项（rotational cross terms）相互抵消。角速度（angular velocity）可以由下式计算

$$
\varphi'=\frac{y_1y_2'-y_2y_1'}{y_1^2+y_2^2}=1.
$$

这种分解将动力学拆分为径向增长或衰减，以及匀速逆时针旋转（uniform counterclockwise rotation）。

### 霍普夫径向方程的解释（Interpreting the Hopf radial equation）

完整的局部行为由下式刻画

$$
r'=F(r)=r(q+l_1r^2), \qquad r\geq0.
$$

径向平衡点（radial equilibrium）$r=0$ 代表平衡点 $(y_1,y_2)=(0,0)$。正的径向平衡点满足

$$
r_*^2=-\frac{q}{l_1}, \qquad r_*=\sqrt{-\frac{q}{l_1}}.
$$

当 $r(t)\equiv r_*$ 时，角向方程给出 $\varphi(t)=t+\varphi_0$。因此

$$
y_1(t)=r_*\cos(t+\varphi_0), \qquad y_2(t)=r_*\sin(t+\varphi_0),
$$

这是一条周期为 $2\pi$ 的周期轨道。因此，径向方程的正平衡点对应于平面系统（planar system）的一条周期轨道。

在原点附近，$r'\approx qr$。因此，当 $q<0$ 时原点具有吸引性，当 $q>0$ 时原点具有排斥性。$l_1$ 的符号决定非零轨道存在于参数的哪一侧，以及它是否稳定。

**超临界情形（Supercritical case），$l_1<0$。** 写成 $l_1=-a$，其中 $a>0$。当 $q<0$ 时，所有充分小的正半径都减小并趋于零。在 $q=0$ 时，$r'=-ar^3$，因此非双曲原点（nonhyperbolic origin）仍然渐近稳定（asymptotically stable），但收敛速度慢于指数收敛（exponential convergence）。当 $q>0$ 时，半径

$$
r_*=\sqrt{q/a}
$$

具有吸引性：比它小的半径增大，而在其上方的邻近半径减小。平衡点变得不稳定，同时一条稳定周期轨道诞生。

**亚临界情形（Subcritical case），$l_1>0$。** 当 $q<0$ 时，原点稳定，但在局部被一条排斥性周期轨道 $r_*=\sqrt{-q/l_1}$ 包围。在 $q=0$ 时，$r'=l_1r^3>0$ 对 $r>0$ 成立，因此非双曲原点不稳定。当 $q>0$ 时，所有充分小的非零半径都增大，并且不存在小周期轨道。

非零轨道的稳定性还可以由下式检验

$$
F'(r)=q+3l_1r^2, \qquad F'(r_*)=-2q.
$$

因此，该轨道在 $q>0$ 时稳定，在 $q<0$ 时不稳定；同时还要结合存在条件 $-q/l_1>0$。

## 11. 从流（flows）到映射（maps）：单位圆判据（unit-circle criterion）

对于参数化迭代映射（parameterized iterated map）

$$
x_{k+1}=g(x_k,p),
$$

设 $x^*$ 为不动点（fixed point），并令

$$
A(p)=Dg(x^*,p)
$$

为线性化映射（linearized map）。它的特征值称为**乘子（multipliers）**。当没有乘子位于单位圆上时，映射的不动点是双曲的（hyperbolic）。因此，当

$$
|\mu|=1\quad\text{or}\quad|\mu_{1,2}|=1.
$$

时，双曲性丧失（loss of hyperbolicity）。这取代了流中关于虚轴的条件。在低维情形下，主要的余维一（codimension-one）可能性为：

- 一个实乘子穿过 $+1$：映射的折叠分岔（fold bifurcation）；
- 一个实乘子穿过 $-1$：翻转分岔（flip bifurcation）或倍周期分岔（period-doubling bifurcation）；
- 一对非实共轭乘子穿过单位圆：奈马克–萨克分岔（Neimark-Sacker bifurcation）。

## 12. 映射的折叠分岔（Fold bifurcation of maps）

对于满足 $g(0,0)=0$ 且临界乘子（critical multiplier）为 $\mu=1$ 的泛型一维映射（generic one-dimensional map），定理 8.12 给出正规形（normal form）

$$
y\mapsto q+y\pm y^2.
$$

泛型性条件（genericity conditions）为

$$
\partial_{xx}g(0,0)\neq0, \qquad \partial_p g(0,0)\neq0.
$$

不动点满足

$$
y=q+y\pm y^2,
$$

因此，不动点方程可化为与流的折叠分岔中相同的二次平衡。于是，一对稳定—不稳定不动点被产生或消灭。其证明与流的折叠定理相似：平移不动点分支、重参数化、缩放二次项系数，并舍去在拓扑意义上无关的高阶项。


**注意：在映射中，不动点应满足 $x=g(x,p)$，由此得到 $x-g(x,p)=0$**

## 13. 翻转（倍周期）分岔定理（Flip (period-doubling) bifurcation theorem）

当一个实乘子穿过 $-1$ 时，就会发生翻转分岔（flip bifurcation）。对于满足 $g(0,0)=0$ 且 $\mu=-1$ 的泛型一维映射（generic one-dimensional map），定理 8.13 给出正规形（normal form）

$$
y\mapsto -(1+q)y\pm y^3.
$$

横截性条件（transversality condition）为

$$
\partial_{xp}g(0,0)\neq0,
$$

它保证乘子以非零的参数变化速率穿过 $-1$。非零的有效三次项系数给出非退化条件（non-degeneracy condition）。

**公式说明（Formula note）。** 摘录中的定理陈述写为

$$
\frac12(\partial_{xx}g(0,0))^2+\frac16\partial_{xxx}g(0,0)\neq0,
$$

而其后的推导得到

$$
c(0)=\left(\frac12\partial_{xx}g(0,0)\right)^2+\frac16\partial_{xxx}g(0,0)\neq0.
$$

这两个表达式的平方项相差一个因子。由 $a(0)=\tfrac12\partial_{xx}g(0,0)$ 和 $c(0)=a(0)^2+b(0)$ 可直接得到证明中的表达式，因此应对照完整原文或勘误表核查这一差异。

---

### 为什么把翻转乘子写成 $-(1+\kappa(p))$？

将不动点分支平移到原点后，把映射写成

$$
g(x,p)=g_1(p)x+g_2(p)x^2+g_3(p)x^3+O(x^4).
$$

由于翻转点处的乘子为 $-1$，

$$
g_1(0)=\partial_xg(0,0)=-1.
$$

完全可以写成

$$
g_1(p)=-k(p), \qquad k(0)=1.
$$

而讲义写成

$$
g_1(p)=-(1+\kappa(p)), \qquad \kappa(0)=0,
$$

其中 $\kappa(p)=k(p)-1$。这两种记号在数学上等价，但第二种更适合局部分岔分析（local bifurcation analysis）：

- $\kappa$ 直接度量乘子相对于临界值 $-1$ 的偏移。
- 临界参数由方便的零值 $\kappa=0$ 表示。
- 由于 $\kappa(0)=0$，它是一个小量，展开式为 $\kappa(p)=\kappa'(0)p+O(p^2)$。
- 若 $\kappa'(0)\neq0$，便可立即用 $q=\kappa(p)$ 作为新参数，得到正规形乘子 $\mu(q)=-(1+q)$。

对于充分小的 $q$，

$$
q<0\Rightarrow |\mu(q)|<1, \qquad q=0\Rightarrow\mu=-1, \qquad q>0\Rightarrow|\mu(q)|>1.
$$

因此，$q$ 的符号直接区分稳定侧与不稳定侧。

还有一个值得记录的符号细节。由于 $g_1(p)=-(1+\kappa(p))$，

$$
g_1'(0)=\partial_{xp}g(0,0)=-\kappa'(0).
$$

因此，精确恒等式中含有一个负号，尽管该符号对于非零横截性条件并无影响。

## 14. 翻转正规形约化与二周期轨道（Flip normal-form reduction and period-two orbit）

由于 $\partial_x g(0,0)=-1\neq1$，隐函数定理（Implicit Function Theorem）首先给出唯一的邻近不动点分支（fixed-point branch）。一个预备坐标变换将该分支变为 $x=0$。泰勒展开（Taylor expansion）给出

$$
g(x,p)=g_1(p)x+g_2(p)x^2+g_3(p)x^3+O(x^4).
$$

横截性（transversality）允许引入新参数 $q$，使线性系数为 $\mu(q)=-(1+q)$。近恒等坐标变换（near-identity coordinate change）

$$
x=y+\delta(q)y^2
$$

可以消去二次项，因为 $\mu(0)^2-\mu(0)=2\neq0$。对剩余的非零三次项系数作缩放，即可得到正规形（normal form）。

对于三次项取正号的情形，令

$$
f(y,q)=-(1+q)y+y^3.
$$

不动点 $y^*=0$ 的乘子为 $-1-q$：当 $q<0$ 时它稳定，当 $q>0$ 时它不稳定。二次迭代（second iterate）$f^2$ 的不动点揭示出一条新的二周期轨道（period-two orbit）。对于小的 $q>0$，它的两个点为

$$
y_{\pm}=\pm\sqrt{q}\quad\text{(plus higher-order corrections in a general system)}.
$$

对于三次项精确取正号的正规形，这两个点相互映射，并构成一条稳定的二周期轨道。这就是该事件被称为倍周期的原因：当一个二周期轨道诞生时，原不动点失去稳定性。

### 参数与状态空间的局部可逆性（Parameter and state-space local invertibility）

翻转分岔的证明使用了两种不同的局部可逆变换，它们各有不同的用途。

**1. 重参数化（Parameter reparameterization）。** 由

$$
\kappa(0)=0, \qquad \kappa'(0)\neq0,
$$

反函数定理（Inverse Function Theorem）给出下列二者之间的光滑局部逆：

$$
q=\kappa(p) \qquad\text{and}\qquad p=\kappa^{-1}(q).
$$

这意味着，$q$ 只是在分岔附近对参数轴重新命名。邻近的参数值不会丢失，也不会被合并。若 $\kappa'(0)>0$，则保持定向；若 $\kappa'(0)<0$，则参数方向反转，但分岔类型不变。

该条件十分重要。例如，$q=p^2$ 会合并 $p$ 与 $-p$，因而无法区分分岔的两侧。映射 $q=p^3$ 是一一对应的，但其逆 $p=q^{1/3}$ 在零点不光滑，因此它不是一个光滑的局部参数坐标。

**2. 状态空间坐标变换（State-space coordinate change）。** 随后，证明使用

$$
x=h_q(y)=y+\delta(q)y^2.
$$

这里的局部可逆性由 $\partial_yh_q(0)=1$ 得到。它的目的不是重新标记参数，而是简化状态空间中的泰勒展开。变换后的映射是局部共轭（local conjugate）

$$
\widetilde g_q=h_q^{-1}\circ g_q\circ h_q.
$$

由于 $h_q$ 在局部是一一对应的，不动点与周期轨道在不丢失信息的情况下相互对应。在相应的不动点处，线性化映射彼此相似（一维情形下相等），因此乘子和局部稳定性得到保持。

简言之，$q=\kappa(p)$ 将参数标准化，而 $x=h_q(y)$ 则简化状态动力学。

### 为什么可以消去二次项（Why the quadratic term can be removed）

重参数化后，设

$$
g_q(x)=\mu(q)x+a(q)x^2+b(q)x^3+O(x^4), \qquad \mu(q)=-(1+q).
$$

使用近恒等变换

$$
x=h_q(y)=y+\delta(q)y^2,
$$

其逆具有展开式

$$
h_q^{-1}(x)=x-\delta x^2+2\delta^2x^3+O(x^4).
$$

共轭映射 $h_q^{-1}\circ g_q\circ h_q$ 的二次项系数为

$$
a+\delta\mu-\delta\mu^2.
$$

令其等于零，得到

$$
\delta(q)=\frac{a(q)}{\mu(q)^2-\mu(q)}.
$$

在临界点 $\mu(0)=-1$ 处，

$$
\mu(0)^2-\mu(0)=1-(-1)=2\neq0,
$$

因此，二次项是非共振的（nonresonant），可以被光滑地消去。结果为

$$
\widetilde g_q(y)=-(1+q)y+c(q)y^3+O(y^4).
$$

在 $\mu=-1$ 处，三次项通常不能被消去，因为它对应的分母为 $\mu^3-\mu=0$。它是共振的（resonant），并包含决定翻转分岔的首要非线性信息。若 $c(0)\neq0$，对 $y$ 进行缩放即可将其归一化为符号 $\pm1$。

强调共轭关系（conjugacy）至关重要：这一计算在保留相同局部动力学的同时简化了公式，而不是用一个无关的映射取代原系统。

### 为什么小而非零的 $q$ 以及尺度 $y\sim\sqrt q$ 很重要（Why small nonzero q and the scale y ~ sqrt(q) matter）

分岔研究并非只考察 $q=0$ 时单独的临界映射，而是比较如下范围内的一个局部映射族：

$$
-\varepsilon<q<\varepsilon
$$

并研究 $q$ 穿过零时发生了什么变化。因此，小并不意味着等于零。

对于三次项取正号的翻转正规形

$$
f_q(y)=-(1+q)y+y^3,
$$

小的二周期点由 $f_q^2(y)=y$ 求得。在最低阶近似下，

$$
0=f_q^2(y)-y\approx2qy-2y^3=2y(q-y^2).
$$

因此，非零解满足的主导平衡关系为

$$
y^2\sim q, \qquad y_\pm\sim\pm\sqrt{q}.
$$

对于精确正规形，这两个点恰为 $\pm\sqrt q$，且满足

$$
f_q(\sqrt q)=-\sqrt q, \qquad f_q(-\sqrt q)=\sqrt q.
$$

它们不是 $f_q$ 的不动点；二者共同构成一条二周期轨道，同时也是 $f_q^2$ 的不动点。它们属于局部分析的范围，因为 $\sqrt q\to0$（当 $q\to0^+$ 时）。更确切地说，对于任意状态空间邻域 $|y|<\delta$，只要选取 $0<q<\delta^2$，就可使这两个点都位于其中。

$f_q$ 本身的不动点方程为

$$
y=f_q(y)\quad\Longleftrightarrow\quad y\bigl(y^2-(2+q)\bigr)=0.
$$

因此，完整的多项式映射还具有不动点 $y=\pm\sqrt{2+q}$。它们被排除在**局部**翻转分析之外，因为

$$
\pm\sqrt{2+q}\to\pm\sqrt2\neq0
$$

当 $q\to0$ 时成立。它们与分岔点始终保持正距离。因此，准确的说法是：$y=0$ 是原点某个充分小邻域内的唯一不动点，而不是整条实数轴上的唯一不动点。

### 翻转分岔图与蛛网图的解读（Interpreting flip diagrams and cobweb plots）

对于三次项取正号的映射

$$
g_q(y)=-(1+q)y+y^3,
$$

原点处的乘子为 $g_q'(0)=-(1+q)$。对于小的 $q<0$，$|g_q'(0)|<1$，因此迭代点在交替位于原点两侧的同时趋近于零。在 $q=0$ 时，乘子等于 $-1$。对于小的 $q>0$，原点不稳定，并出现稳定的二周期轨道 $\{\sqrt q,-\sqrt q\}$。

该二周期轨道的乘子是二次迭代的导数：

$$
(g_q^2)'(\sqrt q)=g_q'(\sqrt q)g_q'(-\sqrt q)=(-1+2q)^2.
$$

对于充分小的 $q>0$，其模小于一，从而确认了稳定性。在分岔图（bifurcation diagram）中，两条曲线分支是同一条二周期轨道的两个点，而不是两条新的不动点分支。按照惯例，实线分支表示稳定，虚线分支表示不稳定。竖直箭头描述固定 $q$ 时的吸引或排斥；参数在迭代过程中并不演化。

对于三次项取负号的映射

$$
g_q(y)=-(1+q)y-y^3,
$$

对称二周期轨道满足 $g_q(y)=-y$，因此

$$
y_\pm=\pm\sqrt{-q}, \qquad q<0.
$$

它的乘子仍为 $(-1+2q)^2$，当 $q<0$ 时该值大于一，所以该轨道不稳定。它在局部把被稳定原点吸引的点与向外运动的点分隔开来。当 $q>0$ 时，原点不稳定，并且不再存在小的实二周期轨道。

蛛网图（cobweb plot）的读法是：先从 $y_n$ 竖直移动到图像 $y_{n+1}=g_q(y_n)$，再水平移动到对角线 $y_{n+1}=y_n$。重复这两个步骤即可显示实际的迭代过程。一个闭合矩形表示一条二周期轨道；交替收缩的步长表示趋向原点，而不断扩大的矩形表示交替向外逃逸。

因此，三次项取正号时会产生超临界翻转分岔（supercritical flip），并在转变后出现稳定的周期轨道；三次项取负号时则会产生次临界翻转分岔（subcritical flip），并在转变前出现不稳定的周期轨道。

## 15. 奈马克–萨克分岔（Neimark-Sacker bifurcation）与埃农映射（Henon map）示例

对于二维映射，一对共轭复乘子（complex-conjugate multiplier）可能在 $|\mu_{1,2}|=1$ 且 $\mu_1\neq\mu_2$ 时穿越单位圆（unit circle）。由此产生的 **奈马克–萨克分岔（Neimark-Sacker bifurcation）** 是霍普夫分岔（Hopf bifurcation）的离散时间对应形式。此时通常不是周期轨道（periodic orbit），而是一条闭不变曲线（closed invariant curve）从不动点处分岔出来；该曲线上的动力学可能比霍普夫正规形（Hopf normal form）中的简单匀速旋转更复杂。

讲义用埃农映射（Henon map）说明如何检测分岔：

$$
\begin{pmatrix}x_1\\x_2\end{pmatrix}\mapsto\begin{pmatrix}x_2\\p_1-p_2x_1-x_2^2\end{pmatrix}.
$$

不动点（fixed point）满足 $x_1^*=x_2^*$ 以及

$$
x_2^*=p_1-p_2x_2^*-(x_2^*)^2,
$$

因此至多存在两个不动点。其雅可比矩阵（Jacobian）为

$$
A(p)=\begin{pmatrix}0&1\\-p_2&-2x_2^*\end{pmatrix}.
$$

与其显式求解乘子，不如定义标量 **分岔函数（bifurcation functions）**。必要的谱条件（spectral conditions）依次对应折叠分岔（fold）、翻转分岔（flip）和候选奈马克–萨克条件（candidate Neimark-Sacker condition）：

$$
\det(A-I)=0 \quad\text{(fold)},
$$

$$
\det(A+I)=0 \quad\text{(flip)},
$$

以及对于二维系统中的一对复乘子，

$$
\det(A)-1=0 \quad\text{(candidate Neimark-Sacker condition)}.
$$

这些是必要的谱检验（spectral tests）。仍须验证相应的乘子确实穿越了边界，并且所有泛型性条件（genericity conditions）均成立。

---

### 行列式如何生成分岔函数（bifurcation functions）

设 $x^*(p)$ 是光滑映射的一个不动点，并令

$$
A(p)=D_xg(x^*(p),p).
$$

乘子 $\lambda$ 是 $A$ 的特征值，当且仅当

$$
\det(A(p)-\lambda I)=0.
$$

因此，代入一个临界乘子（critical multiplier）即可将谱条件转化为参数的标量方程。

**不动点的折叠分岔（fold of fixed points）。** 临界乘子为 $+1$，因此定义

$$
\beta_{\mathrm{fold}}(p)=\det(A(p)-I).
$$

这一条件也可由隐函数定理（Implicit Function Theorem）直接解释。不动点满足 $g(x,p)-x=0$，其关于状态变量的导数为 $A-I$。若 $A-I$ 可逆，则该不动点可唯一且光滑地延拓；只有当这种可逆性失效时，才可能发生折叠分岔。

**翻转分岔（flip）。** 临界乘子为 $-1$，从而得到

$$
\beta_{\mathrm{flip}}(p)=\det(A(p)+I).
$$

**二维系统中的奈马克–萨克分岔（Neimark-Sacker in two dimensions）。** 若乘子是一对非实共轭复数 $re^{\pm i\theta}$，则

$$
\det A=\lambda_1\lambda_2=r^2.
$$

它们到达单位圆意味着 $\det A=1$，这提示可定义

$$
\beta_{\mathrm{NS}}(p)=\det A(p)-1.
$$

然而，仅有 $\det A=1$ 并不能保证乘子是一对非实数。在二维系统中还需要附加条件

$$
(\operatorname{tr}A)^2-4\det A<0
$$

当 $\det A=1$ 时，该条件即为 $|\operatorname{tr}A|<2$。

这些标量函数非常有用，因为单个方程 $\beta(p_1,p_2)=0$ 通常在二维参数平面内刻画一条余维一（codimension-one）曲线。它们的零点集可定位候选分岔，但不能单独证明泛型分岔（generic bifurcation）确实发生。

---

### 行列式检验（determinant tests）的适用范围与局限

对于任意有限维光滑映射，都可以用特征行列式（characteristic determinant）

$$
\chi_A(\lambda,p)=\det(A(p)-\lambda I)
$$

来施加选定的临界乘子条件。因此，基于行列式的分岔函数用途广泛，但它只构成分析中的谱部分。

一个泛型局部分岔（generic local bifurcation）通常要求以下所有条件均成立：

1. **谱条件（spectral condition）：** 相应的乘子位于单位圆上。
2. **单重性（simplicity）：** 临界乘子或共轭对具有预期的重数，且所有其他乘子均远离单位圆。
3. **横截性（transversality）：** 临界乘子以非零速度穿越稳定性边界，而非仅与之相切。
4. **非退化性（non-degeneracy）：** 相应的正规形系数（normal-form coefficient）不为零。

对于一个单重实临界乘子，适当分岔函数的非零导数往往表达了横截性。对于奈马克–萨克乘子对（Neimark-Sacker pair）$\lambda(p)=r(p)e^{i\theta(p)}$，穿越条件为 $r'(p_*)\neq0$。

在高于二维的系统中，$\det A$ 是所有乘子的乘积，通常不足以检测单位圆上的某一对共轭乘子。例如，在三维系统中，

$$
\det A=r^2\rho
$$

其中 $re^{\pm i\theta}$ 为一对乘子，$\rho$ 为第三个乘子。此时，更适合采用直接特征值分析、特征多项式判据（characteristic-polynomial criteria）或数值延拓（numerical continuation）。

同一原理也可推广到周期 $k$ 轨道（period-k orbit）。将 $A$ 替换为迭代映射的单周期矩阵（monodromy matrix）

$$
M=Dg^k(x_0)=Dg(x_{k-1})\cdots Dg(x_1)Dg(x_0).
$$

于是，$\det(M-I)=0$ 是该周期轨道的候选折叠条件，而 $\det(M+I)=0$ 是候选倍周期条件（period-doubling condition）。

行列式检验是局部线性工具。仅靠它们无法检测同宿连接（homoclinic connections）、异宿连接（heteroclinic connections）、危机（crises）或大尺度吸引子碰撞（large-scale attractor collisions）等全局分岔（global bifurcations）。

### 埃农示例的显式分岔函数（explicit bifurcation functions）

对于讲义中的埃农映射（Henon map）

$$
g(x_1,x_2)=\bigl(x_2,\ p_1-p_2x_1-x_2^2\bigr),
$$

不动点满足 $x_1^*=x_2^*=s$ 以及

$$
s^2+(1+p_2)s-p_1=0.
$$

该点处的雅可比矩阵为

$$
A=\begin{pmatrix}0&1\\-p_2&-2s\end{pmatrix},
$$

并且

$$
\operatorname{tr}A=-2s, \qquad \det A=p_2.
$$

候选折叠函数为

$$
\beta_{\mathrm{fold}}=\det(A-I)=2s+1+p_2.
$$

因此，候选折叠分岔满足 $2s+1+p_2=0$。候选翻转函数为

$$
\beta_{\mathrm{flip}}=\det(A+I)=1-2s+p_2,
$$

所以候选翻转分岔满足 $1-2s+p_2=0$。

对于候选奈马克–萨克分岔（Neimark-Sacker bifurcation），

$$
\det A=1 \quad\Longleftrightarrow\quad p_2=1.
$$

为了保证乘子是一对非实数，而不是两个互为倒数的实数乘子，还需要

$$
|\operatorname{tr}A|<2 \quad\Longleftrightarrow\quad |s|<1.
$$

这些方程可定位参数空间中的候选曲线或线段。完整证明仍须检验穿越性、单重性、共振排除条件（resonance exclusions）以及相应的非线性系数。奈马克–萨克不变曲线（Neimark-Sacker invariant curve）是霍普夫分岔（Hopf bifurcation）中新生周期轨道的离散时间对应形式，但不变曲线上的动力学可以包含准周期旋转（quasiperiodic rotation）和共振效应（resonance effects）。

## 16. 主要局部分岔（local bifurcations）的比较

| 系统（System） | 分岔（Bifurcation） | 临界谱（Critical spectrum） | 正规形（Normal form） | 主要局部结果（Main local outcome） |
|---|---|---|---|---|
| 一维流（1D flow） | 折叠／鞍结分岔（Fold / saddle-node） | $\lambda=0$ | $y'=q\pm y^2$ | 一对稳定—不稳定平衡点被产生或消灭 |
| 一维对称流（1D symmetric flow） | 叉形分岔（Pitchfork） | $\lambda=0$ 加对称性 | 对于超临界情形（supercritical case），$y'=qy-y^3$ | 一个对称分支失去稳定性，并出现两个稳定分支 |
| 一维流（1D flow） | 跨临界分岔（Transcritical） | 具有特殊结构的 $\lambda=0$ | $y'=y(q-y)$ | 两条平衡点分支相交并交换稳定性 |
| 二维流（2D flow） | 霍普夫分岔（Hopf bifurcation） | $\lambda_{1,2}=\pm i\omega$ | $z'=(q+i)z+l_1z|z|^2$ | 一个小振幅周期轨道出现或消失 |
| 一维映射（1D map） | 折叠分岔（Fold） | $\mu=+1$ | $y\mapsto q+y\pm y^2$ | 一对不动点被产生或消灭 |
| 一维映射（1D map） | 翻转分岔（Flip） | $\mu=-1$ | $y\mapsto-(1+q)y\pm y^3$ | 一个二周期轨道出现或消失 |
| 二维映射（2D map） | 奈马克–萨克分岔（Neimark-Sacker bifurcation） | $|\mu_{1,2}|=1$ | 本节选未推导 | 通常会分岔出一条闭不变曲线 |

该表突出了一项有用的类比：虚轴（imaginary axis）是流的特征值稳定性边界，而单位圆（unit circle）是映射的乘子稳定性边界。

## 17. 实用的分岔识别流程（bifurcation-identification workflow）

对于一个具体的含参系统，讲义建议采用以下流程。

1. **求平衡点或不动点（equilibria or fixed points）。** 对流求解 $f(x,p)=0$，对映射求解 $g(x,p)=x$。
2. **线性化（linearize）。** 计算 $D_xf(x^*,p)$ 或 $Dg(x^*,p)$。
3. **定位双曲性丧失（loss of hyperbolicity）。** 对于流，寻找实部为零的特征值；对于映射，寻找位于单位圆上的乘子。
4. **分类临界谱（critical spectrum）。** 一个单重零特征值（simple zero eigenvalue）提示可能发生折叠、叉形或跨临界分岔；一对纯虚共轭特征值提示霍普夫分岔（Hopf bifurcation）；乘子 $+1$、$-1$ 或一对单位模复乘子则分别提示折叠、翻转或奈马克–萨克分岔（Neimark-Sacker bifurcation）。
5. **检验横截性（transversality）。** 确认参数变化时，特征值或乘子以非零速度穿越稳定性边界。
6. **检验非退化性（non-degeneracy）。** 确认首个相关的非线性系数不为零：折叠分岔对应二次项，霍普夫分岔对应三次项／第一李雅普诺夫系数（first Lyapunov coefficient），翻转分岔对应有效三次项系数（effective cubic coefficient）。
7. **约化为正规形（normal form）。** 使用平移、重参数化（reparameterizations）、近恒等坐标变换（near-identity coordinate changes）和尺度变换（scalings）。
8. **读出动力学（dynamics）。** 根据正规形确定不变对象（invariant objects）的数量及稳定性。

谱条件只是起点。若不进行泛型性检验（genericity checks），系统可能呈现不同类型或更高余维（higher-codimension）的分岔。

### 最终要点（Final takeaways）

- 局部分岔（local bifurcations）发生于双曲性丧失（loss of hyperbolicity）处，但仅凭临界谱（critical spectrum）无法对其进行完整分类。
- 正规形（normal forms）保留控制临界点附近拓扑与稳定性的最低阶项。
- 横截性（transversality）和非退化性（non-degeneracy）使分岔具有泛型性和鲁棒性（generic and robust）。
- 折叠分岔（fold bifurcation）与霍普夫分岔（Hopf bifurcation）是低维流中主要的余维一（codimension-one）情形。
- 折叠分岔（fold）、翻转分岔（flip）和奈马克–萨克分岔（Neimark-Sacker bifurcation）是它们在映射中的重要对应形式。
- 正如讲义所述，下一步理论工作是系统化正规形变换（normal-form transformations），并将高维系统约化为低维中心动力学（center dynamics）。
