# Lecture 2 — Stability and Lyapunov Theory

## 1. Stability of equilibria

Let $x_*$ be an equilibrium of $(\mathcal{X},\mathcal{T},\phi_t)$. It is **Lyapunov stable** if, for every neighborhood $\mathcal{U}$ of $x_*$, there is a neighborhood $\mathcal{V}\subseteq\mathcal{U}$ of $x_*$ such that

$$
\phi_t(\mathcal{V})\subseteq\mathcal{U}
\qquad\text{for all }t\geq0.
$$

If this condition fails, $x_*$ is **unstable**. It is **asymptotically stable** if it is Lyapunov stable and there is a neighborhood $\mathcal{W}$ of $x_*$ for which

$$
\lim_{t\to\infty}\phi_t(x_0)=x_*
\qquad\text{for every }x_0\in\mathcal{W}.
$$

If the convergence holds for every $x_0\in\mathcal{X}$, then $x_*$ is **globally asymptotically stable**.

## 2. Lyapunov functions

Consider $\dot{x}=f(x)$ and let $x_*$ be an equilibrium. A continuously differentiable function $V:\mathcal{U}\to\mathbb{R}$ is a **Lyapunov function** near $x_*$ when

$$
V(x_*)=0,
\qquad
V(x)>0\quad\text{for }x\neq x_*,
$$

and its derivative along solutions is non-positive:

$$
\dot{V}(x)
=
\nabla V(x)\cdot f(x)
\leq0.
$$

Equivalently, $V(\phi_t(x_0))$ does not increase along a trajectory. If $\dot{V}(x)<0$ for every $x\neq x_*$, then $V$ is called a **strict Lyapunov function**. It is often interpreted as an energy-like quantity that dissipates over time.

## 3. Lyapunov stability criterion

Let $x_*$ be an equilibrium of $\dot{x}=f(x)$.

- If there is a positive-definite Lyapunov function $V$ with $\dot V\leq0$ in a neighborhood of $x_*$, then $x_*$ is Lyapunov stable.
- If, in addition, $\dot V<0$ for all $x\neq x_*$ in that neighborhood, then $x_*$ is locally asymptotically stable.

These statements are the standard Lyapunov direct method.

## 4. $\omega$-limit and $\alpha$-limit sets

For $x\in\mathcal{X}$, the **$\omega$-limit set** consists of all accumulation points of the forward orbit:

$$
\omega(x)
=
\left\{
 y\in\mathcal{X}:
 \exists\,t_j\to+\infty
 \text{ such that }
 \phi_{t_j}(x)\to y
\right\}.
$$

Similarly, the **$\alpha$-limit set** consists of accumulation points in backward time:

$$
\alpha(x)
=
\left\{
 y\in\mathcal{X}:
 \exists\,t_j\to-\infty
 \text{ such that }
 \phi_{t_j}(x)\to y
\right\}.
$$

The $\alpha$-limit set is defined only when the solution exists for arbitrarily large negative times.

## 5. LaSalle’s invariance principle

Let $\Omega$ be a compact, positively invariant set for $\dot{x}=f(x)$, and let $V\in C^1(\Omega)$ satisfy

$$
\dot V(x)=\nabla V(x)\cdot f(x)\leq0
\qquad\text{for all }x\in\Omega.
$$

Define

$$
E=\left\{x\in\Omega:\dot V(x)=0\right\},
$$

and let $M$ be the largest invariant subset of $E$. Then every trajectory starting in $\Omega$ approaches $M$ as $t\to\infty$.

In particular, if $M=\{x_*\}$ for an equilibrium $x_*$, then $x_*$ is asymptotically stable (under the usual local positive-definiteness assumptions on $V$). This is the precise form of the principle used when $\dot V$ is only non-positive rather than strictly negative.
