# Lecture 10: Center Manifolds

## Learning Objectives

After this lecture, you should be able to:

- distinguish the linear center subspace $E^c$ from the nonlinear center manifold $W^c$;
- state the Center Manifold Theorem and the Center Manifold Reduction Theorem;
- understand the non-uniqueness and regularity limitations of center manifolds;

## 1. Motivation and Spectral Decomposition

Consider a smooth ODE near an equilibrium at the origin:

$$
x'
=
Ax+f^{(2)}(x)+f^{(3)}(x)+\cdots
=
f(x),
\qquad
x\in\mathbb R^d,
\qquad
f\in C^k,
\tag{10.1}
$$

where

$$
A=Df(0),
\qquad
x_*=0.
$$

Previous local bifurcation calculations often implicitly assumed that the system had already been written in its minimal critical dimension. Center-manifold theory justifies this reduction when additional stable or unstable variables are present.

The center spectral subspace is the direct sum of the generalized eigenspaces associated with eigenvalues on the imaginary axis:

$
E^c(0)
=
\bigoplus_{\operatorname{Re}\lambda=0}
E_\lambda^{\mathrm{gen}}.
$

For complex-conjugate eigenvalues, $E^c$ is understood as the associated real invariant subspace.

The spectrum is divided into three groups:

$$
\begin{aligned}
d_c &: \quad \operatorname{Re}\lambda_c=0,\\
d_s &: \quad \operatorname{Re}\lambda_s<0,\\
d_u &: \quad \operatorname{Re}\lambda_u>0.
\end{aligned}
$$

Correspondingly,

$$
\mathbb R^d
=
E^s\oplus E^c\oplus E^u.
$$

If $d_c=0$, the equilibrium is hyperbolic and the Stable/Unstable Manifold Theorem applies. When $d_c>0$, the linearized system does not determine the behavior in the center directions, so nonlinear terms must be considered.

Figure 23 illustrates:

- the basic geometry of a center manifold $W^c$;
- the relationship between $E^c$, $E^s$, and $W^c$;
- the possible non-uniqueness and limited regularity of center manifolds.

The pictures represent a local enlargement near the equilibrium.


## 2. The Center Manifold Theorem

### Theorem 10.1: Center Manifold Theorem

Suppose the vector field in (10.1) is $C^k$. Then there exists a locally invariant $d_c$-dimensional $C^k$ manifold

$$
W_{\mathrm{loc}}^c(0)
$$

such that

$$
T_0W_{\mathrm{loc}}^c(0)=E^c(0).
$$

Moreover:

- every trajectory that remains in a sufficiently small neighborhood for all $t\geq0$ approaches $W_{\mathrm{loc}}^c(0)$ as $t\to+\infty$;
- every trajectory that remains in that neighborhood for all $t\leq0$ approaches $W_{\mathrm{loc}}^c(0)$ as $t\to-\infty$.

Here, convergence to the manifold means

$$
\operatorname{dist}
\left(
\phi_t(z),W_{\mathrm{loc}}^c(0)
\right)
\longrightarrow0.
$$

It does not necessarily mean that the trajectory converges to one particular point.

### Definition 10.2

The manifold $W_{\mathrm{loc}}^c(0)$ is called a local center manifold. It is often abbreviated as

$$
W^c(0)
\qquad\text{or}\qquad
W^c.
$$

The proof uses a fixed-point construction related to, but more subtle than, the proof of the Stable/Unstable Manifold Theorem.


## 3. Geometric Interpretation of Figure 23(a)

Figure 23(a) shows the simplest situation in which the phase space splits as

$$
\mathbb R^2=E^c\oplus E^s.
$$

The horizontal direction $E^c$ corresponds to eigenvalues with zero real part, while the vertical direction $E^s$ corresponds to eigenvalues with negative real part. Thus the linearized flow neither exponentially contracts nor expands along $E^c$, but it contracts exponentially along $E^s$.

### 3.1 From the center subspace to the center manifold

For the linear system, the center dynamics lies on the straight subspace $E^c$. Nonlinear terms generally bend this invariant set into a local center manifold $W^c$. In graph form,

$$
W^c_{\mathrm{loc}}
=
\{(x,h(x)):|x|\text{ small}\},
$$

with

$$
h(0)=0,
\qquad
Dh(0)=0.
$$

Equivalently,

$$
T_0W^c=E^c.
$$

The curved shape in the figure is schematic: it expresses nonlinear bending and does not imply that the manifold must be a parabola. In general $W^c\neq E^c$; they coincide only in special cases where the nonlinear coupling does not bend the center directions.

### 3.2 Nearby trajectories and reduced dynamics

A nearby point may be decomposed into center and stable components,

$$
z=z_c+z_s,
\qquad
z_c\in E^c,
\quad
z_s\in E^s.
$$

For the linearized system, the stable component satisfies

$$
z_s(t)=e^{A_st}z_s(0),
$$

so $z_s(t)\to0$ as $t\to+\infty$. For the nonlinear system, the relevant transverse quantity is the displacement $y-h(x)$, or equivalently the distance to $W^c$, rather than the stable coordinate $y$ itself. Every trajectory that remains in the chosen neighborhood for all forward time approaches $W^c$ in this transverse sense. Thus the arrows above the manifold point downward and those below it point upward. Once a trajectory is on $W^c$, local invariance keeps it there as long as it remains in the neighborhood.

If the system is written as

$$
\begin{aligned}
x'&=A_cx+F(x,y),\\
y'&=A_sy+G(x,y),
\end{aligned}
$$

then its restriction to $y=h(x)$ is

$$
x'=A_cx+F(x,h(x)).
$$

Hence $E^c$ is the linear tangent space, whereas $W^c$ is the nonlinear invariant manifold carrying the critical center dynamics. The manifold is not a set of equilibria: after approaching it, a trajectory may still move toward the origin, away from it, or otherwise evolve along it, depending on the reduced equation.


## 4. Limitations of Center Manifolds

### Theorem 10.3: Center Manifold Limitations

Center manifolds have two important limitations:

1. They need not be unique.
2. They need not be analytic, even when the vector field is analytic.

These statements are illustrated by the following example.


## 5. System (10.2): Non-Uniqueness and Non-Analyticity

Consider

$$
\begin{aligned}
x'&=x^2,\\
y'&=-y.
\end{aligned}
\tag{10.2}
$$

At the origin,

$$
A=
\begin{pmatrix}
0&0\\
0&-1
\end{pmatrix},
\qquad
E^c=\{(x,0):x\in\mathbb R\},
\qquad
E^s=\{(0,y):y\in\mathbb R\}.
$$

Thus a local center manifold can be written as $y=h(x)$ with $h(0)=h'(0)=0$.

### 5.1 Explicit solutions

For $x_0\neq0$, separation of variables gives

$$
\frac{dx}{x^2}=dt,
\qquad
-\frac1{x(t)}=t-\frac1{x_0},
$$

and therefore

$$
x(t)=\frac{1}{\frac1{x_0}-t}
=\frac{x_0}{1-x_0t}.
$$

A direct check gives

$$
x'(t)=\frac{x_0^2}{(1-x_0t)^2}=x(t)^2,
\qquad
x(0)=x_0.
$$

The other component is

$$
y(t)=y_0e^{-t}.
$$

When $x_0=0$, the solution is $x(t)\equiv0$. If $x_0>0$, the solution blows up as $t\to(1/x_0)^-$; the center-manifold discussion is local and uses only the part of the orbit inside a small neighborhood.

### 5.2 The orbit family

If $y=h(x)$ is invariant, then

$$
y'=h'(x)x',
$$

so

$$
x^2h'(x)=-h(x).
$$

For a nonzero branch,

$$
\frac{h'(x)}{h(x)}=-\frac1{x^2},
$$

which integrates to

$$
h(x)=p e^{1/x},
\qquad
p\in\mathbb R.
$$

Equivalently, one may divide the two differential equations for $x\neq0$:

$$
\frac{dy}{dx}
=
\frac{y'}{x'}
=
-\frac{y}{x^2},
$$

which integrates to the same family $y=p e^{1/x}$.

The same relation follows directly from the explicit solutions. Since

$$
\frac1{x(t)}=\frac1{x_0}-t,
$$

we obtain

$$
y(t)
=
y_0\exp\left(\frac1{x(t)}-\frac1{x_0}\right)
=
p e^{1/x(t)},
\qquad
p=y_0e^{-1/x_0}.
$$

Thus $p$ is constant along an orbit; together with the sign of $x$, it identifies the corresponding orbit branch.

### 5.3 A one-parameter family of center manifolds

For every $p\in\mathbb R$, define

$$
h_p(x)=
\begin{cases}
p e^{1/x},&x<0,\\
0,&x\ge0.
\end{cases}
$$

Then

$$
W_p^c(0)
=
\{(x,y)\in\mathbb R^2:y=h_p(x)\}
$$

is a local center manifold.

The left and right sides are different because

$$
e^{1/x}\to0
\quad\text{as }x\to0^-,
\qquad
e^{1/x}\to+\infty
\quad\text{as }x\to0^+.
$$

Hence every finite $p$ gives a left-hand curve that extends smoothly to, and dynamically approaches, the origin, whereas on the right only $p=0$, namely $y=0$, can do so. The left side is not required to be nonzero: $p=0$ gives the $x$-axis there as well, while $p=1$ and $p=-1$ give upper and lower nonzero left branches.

The time orientation gives the same explanation. If $x_0<0$, then

$$
x(t)\to0^-,
\qquad
y(t)\to0
\qquad
(t\to+\infty),
$$

so an entire family of left-hand orbits approaches the origin in forward time. If $x_0>0$, then $x(t)\to0^+$ only as $t\to-\infty$, but a nonzero $y_0e^{-t}$ grows without bound backward in time. Therefore the only right-hand orbit that approaches the origin is the positive $x$-axis. The underlying asymmetry is that $x'=x^2>0$ for $x\neq0$, so motion is always to the right, while $y'=-y$ contracts forward in time and expands backward in time.

More precisely, for $x_0>0$,

$$
\gamma(t)
=
\left(
\frac{x_0}{1-x_0t},0
\right),
\qquad
-\infty<t<\frac1{x_0},
$$

has geometric image $\{(x,0):x>0\}$. It approaches the origin only as $t\to-\infty$ and never crosses it at finite time. Consequently,

$$
W_p^c
=
\{(x,p e^{1/x}):x<0\}
\cup\{(0,0)\}
\cup\{(x,0):x>0\}
$$

is the union of a left orbit, the equilibrium orbit, and a right orbit; these three pieces form one smooth invariant curve.

### 5.4 Smoothness, non-uniqueness, and non-analyticity

For every integer $m\ge0$,

$$
\lim_{x\to0^-}\frac{e^{1/x}}{|x|^m}=0.
$$

In particular,

$$
h_p'(x)=-\frac{p}{x^2}e^{1/x}\to0
\qquad(x\to0^-).
$$

More generally, every derivative of $e^{1/x}$ for $x<0$ has the form

$$
P_n\left(\frac1x\right)e^{1/x},
$$

where $P_n$ is a polynomial. Since the exponential term decays faster than every power of $|x|$ as $x\to0^-$,

$$
h_p^{(n)}(0)=0
\qquad
\text{for every }n\ge0.
$$

Thus $h_p\in C^\infty$, $h_p(0)=h_p'(0)=0$, and $T_0W_p^c=E^c$. Different values of $p$ give different center manifolds, so the center manifold is not unique.

If $p\neq0$, however, the Taylor series of $h_p$ at the origin is identically zero although $h_p(x)\neq0$ for $x<0$. Hence $h_p$ is not analytic. The differences between these manifolds are flat at the origin, which is why their finite-order Taylor expansions—and therefore finite-order bifurcation and normal-form calculations—agree.

A stronger regularity warning is also possible: even for an analytic vector field, a $C^\infty$ center manifold need not exist. One standard counterexample is

$$
x_1'=-x_2x_1-x_1^3,
\qquad
x_2'=0,
\qquad
y'=-y+x_1^2.
$$

Verifying this stronger claim is a nontrivial exercise.


## 6. Center-Hyperbolic Splitting

After a suitable linear coordinate transformation, the system can be written as

$$
\begin{aligned}
x'&=Bx+F(x,y),
&
F&=\mathcal O(x^2,xy,y^2),\\
y'&=Cy+G(x,y),
&
G&=\mathcal O(x^2,xy,y^2),
\end{aligned}
\tag{10.3}
$$

where

$$
x\in\mathbb R^{d_c},
\qquad
y\in\mathbb R^{d_s+d_u}.
$$

The matrices satisfy

$$
B\in\mathbb R^{d_c\times d_c},
\qquad
\operatorname{spec}(B)\subset i\mathbb R,
$$

and

$$
C
\in
\mathbb R^{(d_s+d_u)\times(d_s+d_u)},
\qquad
\operatorname{spec}(C)\cap i\mathbb R=\varnothing.
$$

Thus:

- $x$ contains the center or critical variables;
- $y$ contains the stable and unstable, hence noncritical, variables;
- all eigenvalues of $B$ lie on the imaginary axis;
- $C$ has no eigenvalues on the imaginary axis;
- $F$ and $G$ contain only terms of total degree at least two.

The notation

$$
\mathcal O(x^2,xy,y^2)
$$

is shorthand for

$$
\mathcal O\left(\|(x,y)\|^2\right).
$$


## 7. Center Manifold Reduction

### Theorem 10.4: Center Manifold Reduction

The system (10.3) is locally topologically equivalent near the origin to

$$
\begin{aligned}
x'&=Bx+F(x,H(x)),\\
y'&=Cy,
\end{aligned}
\tag{10.4}
$$

where

$$
W_{\mathrm{loc}}^c(0)
=
\left\{
(x,y)
\in
\mathbb R^{d_c}
\times
\mathbb R^{d_s+d_u}
:
y=H(x)
\right\},
$$

and

$$
H(x)=\mathcal O(\|x\|^2).
$$

Thus, the original nonlinear system is locally separated into:

$$
\boxed{
\text{nonlinear center dynamics}
+
\text{linear hyperbolic dynamics}.
}
$$

The essential non-hyperbolic dynamics is governed by

$$
x'=Bx+F(x,H(x)).
$$

For the remainder of the lecture, the most important case is assumed:

$$
d_u=0.
$$

Then $y$ contains only stable variables.


## 8. Meaning of the Graph Map $H$

Writing

$$
W^c_{\mathrm{loc}}
=
\{(x,H(x)):x\in E^c\}
$$

does not mean that the dynamics maps the center subspace into the stable subspace. Rather, $x$ is used as a coordinate on the manifold, and $H(x)$ gives the stable or hyperbolic displacement of the manifold above the tangent space $E^c$.

When $d_u=0$,

$$
H:E^c\longrightarrow E^s.
$$

If unstable directions are also present, the graph map instead takes values in $E^s\oplus E^u$. Because the center manifold passes through the origin and is tangent to $E^c$,

$$
H(0)=0,
\qquad
DH(0)=0,
$$

and, when $H$ is at least $C^2$,

$$
H(x)=\mathcal O(\|x\|^2).
$$

Thus $E^c=\{(x,0)\}$ is only the linear tangent space; the nonlinear center manifold is generally curved. The exact function $H$ satisfies a nonlinear invariance equation and is rarely available in closed form. In applications one therefore computes a polynomial approximation $h(x)$ to the order required by the reduced dynamics.


## 9. Why Center Manifolds Capture Local Bifurcations

Consider a parameter-dependent system

$$
z'=f(z,\mu),
\qquad
f(0,0)=0.
$$

If every eigenvalue of $D_zf(0,0)$ has nonzero real part, the equilibrium is hyperbolic. Its local phase portrait is structurally stable under small perturbations, so a local bifurcation requires a loss of hyperbolicity:

$$
\operatorname{Re}\lambda=0.
$$

A zero eigenvalue produces the critical direction for steady-state bifurcations such as saddle-node, transcritical, or pitchfork bifurcations. A pair $\lambda_{1,2}=\pm i\omega$ produces the two critical directions of a Hopf bifurcation. These critical directions form the center subspace.

After the splitting into center and hyperbolic variables,

$$
\begin{aligned}
x'&=Bx+F(x,y,\mu),\\
y'&=Cy+G(x,y,\mu),
\end{aligned}
$$

the center manifold is $y=H(x,\mu)$, and the reduced system is

$$
x'=Bx+F(x,H(x,\mu),\mu).
$$

The hyperbolic directions have a behavior already determined by their linear part: stable components contract and unstable components expand. In the stable case, nearby trajectories rapidly approach $y=H(x,\mu)$, after which the slower center dynamics controls their local behavior. This is often called the slaving principle.

The reduction preserves the local invariant objects relevant to bifurcation. A nearby equilibrium is a bounded complete orbit, so it lies on the center manifold and satisfies the reduced equilibrium equation. A sufficiently small periodic orbit is also a bounded complete orbit and likewise lies on the center manifold. Consequently, steady-state and Hopf bifurcations can be detected in the reduced system.

For example, at a Hopf point the two-dimensional reduced equation can be written in complex normal form as

$$
z'
=
\bigl(\alpha(\mu)+i\omega(\mu)\bigr)z
+\ell_1z|z|^2
+\mathcal O(|z|^4).
$$

With $z=re^{i\theta}$,

$$
\begin{aligned}
r'&=\alpha(\mu)r+\operatorname{Re}(\ell_1)r^3+\cdots,\\
\theta'&=\omega(\mu)+\operatorname{Im}(\ell_1)r^2+\cdots.
\end{aligned}
$$

A small periodic orbit corresponds to a nonzero equilibrium of the radial equation. Hence its existence, stability, and supercritical or subcritical character are determined on the center manifold.


## 10. Center Manifolds and Normal Forms

Center-manifold reduction and normal-form reduction are consecutive but distinct operations:

$$
\text{full system}
\longrightarrow
\text{center manifold}
\longrightarrow
\text{reduced center dynamics}
\longrightarrow
\text{normal form}.
$$

The center-manifold step removes the independent hyperbolic variables by imposing $y=H(x)$. The normal-form step then applies near-identity coordinate changes to the reduced equation,

$$
x'=Bx+g_2(x)+g_3(x)+\cdots,
$$

and eliminates nonessential nonresonant terms. What remains contains the coefficients that determine the local bifurcation type.

One cannot generally replace the center manifold by the center subspace and simply set $y=0$. Stable variables are not independent critical modes, but their nonlinear response can change coefficients in the reduced equation. A block-form example is

$$
\begin{aligned}
x'&=a x^3+bxy+\cdots,\\
y'&=-\lambda y+c x^2+\cdots,
\qquad
\lambda>0.
\end{aligned}
$$

If

$$
y=H(x)=\alpha x^2+\cdots,
$$

the invariance equation gives, at quadratic order,

$$
\alpha=\frac{c}{\lambda}.
$$

Substitution into the center equation yields

$$
x'
=
\left(
a+\frac{bc}{\lambda}
\right)x^3+\cdots.
$$

Thus the stable variable does not create the bifurcation direction, but it changes the effective cubic coefficient. This is why $H$, or a sufficiently accurate approximation of it, must be computed before the normal form is interpreted.

Only a finite-order approximation is usually needed. Saddle-node and transcritical behavior is commonly decided by quadratic terms, while pitchfork and Hopf calculations generally require cubic terms. Although a center manifold may be nonunique, sufficiently smooth center manifolds have the same finite-order expansion needed for these calculations; in Example 10.2 their differences are flat at the origin. Therefore the relevant finite-order normal-form coefficients are well defined.


## 11. The Center-Manifold Invariance Equation

Suppose

$$
W^c=\{(x,y):y=h(x)\}.
$$

Along a trajectory on this graph,

$$
y'=Dh(x)x'.
$$

Using

$$
x'=Bx+F(x,h(x)),
$$

we obtain

$$
y'
=
Dh(x)
\left[
Bx+F(x,h(x))
\right].
$$

The original $y$-equation gives

$$
y'=Ch(x)+G(x,h(x)).
$$

Equating the two expressions yields

$$
Dh(x)
\left[
Bx+F(x,h(x))
\right]
=
Ch(x)+G(x,h(x)).
$$

### Definition 10.5

Define the invariance residual by

$$
\boxed{
\mathcal N[h](x)
=
Dh(x)
\left[
Bx+F(x,h(x))
\right]
-
Ch(x)
-
G(x,h(x)).
}
$$

The center-manifold invariance equation is

$$
\boxed{
\mathcal N[h](x)=0,
}
\tag{10.5}
$$

together with

$$
h(0)=0,
\qquad
Dh(0)=0.
$$

When $d_c>1$, this is generally a nonlinear first-order PDE. Even in one center dimension, it is rarely solvable in closed form.


## 12. A Short Coefficient-Matching Illustration

Consider

$$
\begin{aligned}
x'&=x^2,\\
y'&=-y+x^2.
\end{aligned}
$$

The center variable is $x$, while $y$ is stable. Let

$$
y=h(x)
=
ax^2+bx^3+\mathcal O(x^4).
$$

The invariance equation is

$$
h'(x)x^2=-h(x)+x^2.
$$

Since

$$
h'(x)
=
2ax+3bx^2+\mathcal O(x^3),
$$

the left-hand side is

$$
h'(x)x^2
=
2ax^3+\mathcal O(x^4).
$$

The right-hand side is

$$
-h(x)+x^2
=
(1-a)x^2-bx^3+\mathcal O(x^4).
$$

Matching quadratic terms gives

$$
a=1.
$$

Matching cubic terms gives

$$
2a=-b,
$$

so

$$
b=-2.
$$

Therefore,

$$
\boxed{
h(x)
=
x^2-2x^3+\mathcal O(x^4).
}
$$

This approximation describes how the center manifold is displaced in the stable direction. It does not describe a dynamical map that sends the center subspace into the stable subspace.


## 13. Approximation of Center Manifolds

### Theorem 10.6: Approximation Theorem

Suppose $h$ satisfies

$$
h(0)=0,
\qquad
Dh(0)=0,
$$

and, for some $r>1$,

$$
\mathcal N[h](x)
=
\mathcal O(\|x\|^r)
\qquad
(x\to0).
$$

Then an exact center-manifold graph $y=H(x)$ satisfies

$$
\boxed{
H(x)
=
h(x)+\mathcal O(\|x\|^r).
}
\tag{10.6}
$$

Thus:

$$
\boxed{
\text{an $r$-th order invariance residual produces an $r$-th order graph error}.
}
$$

This theorem makes formal coefficient matching rigorous.


## 14. Detailed Proof of the Approximation Theorem

The proof turns the center-manifold problem into a fixed-point problem and then shows that a small invariance residual requires only a correction of the same order. We assume $d_u=0$, so

$$
\sigma(C)
\subset
\{\lambda\in\mathbb C:\operatorname{Re}\lambda<0\}.
$$

As usual, the local vector field is smoothly cut off outside a small neighborhood so that the required trajectories and Lipschitz estimates are globally defined without changing the local problem.

### 14.1 Lyapunov--Perron representation

Suppose $y=H(x)$ is a center manifold and let $x(t,x_0,H)$ solve

$$
x'=Bx+F(x,H(x)),
\qquad
x(0,x_0,H)=x_0.
$$

Along the manifold,

$$
y(t)=H(x(t,x_0,H))
$$

satisfies $y'=Cy+G(x,y)$. Variation of constants, integrated from $s<0$ to $0$, gives

$$
y(0)-e^{-Cs}y(s)
=
\int_s^0
e^{-C\tau}G(x(\tau),y(\tau))\,d\tau.
$$

Because $C$ is stable and the cut-off orbit is bounded,

$$
e^{-Cs}y(s)\to0
\qquad
(s\to-\infty).
$$

Hence

$$
H(x_0)
=
\int_{-\infty}^0
e^{-Cs}
G\left(
x(s,x_0,H),
H(x(s,x_0,H))
\right)\,ds.
\tag{10.7}
$$

For a candidate graph $\Phi$, define the right-hand side as $(K\Phi)(x_0)$. A center manifold is therefore a fixed point:

$$
KH=H.
$$

The lower limit $-\infty$ removes the freely chosen stable homogeneous mode; only the stable response forced by the center motion remains.

### 14.2 Extending an approximate graph

Let $h$ satisfy

$$
h(0)=0,
\qquad
Dh(0)=0,
\qquad
\mathcal N(h(x))=\mathcal O(\|x\|^p),
\quad p>1.
$$

Choose a compactly supported $C^1$ extension $\theta$ that equals $h$ near the origin. Its invariance residual is

$$
N(x)
=
D\theta(x)
\bigl[
Bx+F(x,\theta(x))
\bigr]
-C\theta(x)
-G(x,\theta(x)).
$$

Near the origin,

$$
N(x)=\mathcal O(\|x\|^p).
$$

The cutoff is technical: it makes $\theta$ and its derivative bounded while preserving the original local approximation.

### 14.3 From invariance residual to fixed-point residual

Let $x_\theta(t)$ solve

$$
x_\theta'
=
Bx_\theta+F(x_\theta,\theta(x_\theta)),
\qquad
x_\theta(0)=x_0.
$$

By the chain and product rules,

$$
\begin{aligned}
\frac{d}{dt}
\left[
e^{-Ct}\theta(x_\theta(t))
\right]
&=
e^{-Ct}
\left[
D\theta(x_\theta(t))x_\theta'(t)
-C\theta(x_\theta(t))
\right]\\
&=
e^{-Ct}
\left[
G(x_\theta(t),\theta(x_\theta(t)))
+N(x_\theta(t))
\right].
\end{aligned}
$$

Integrating from $T<0$ to $0$ gives

$$
\begin{aligned}
\theta(x_0)
-e^{-CT}\theta(x_\theta(T))
={}&
\int_T^0
e^{-Cs}
G(x_\theta(s),\theta(x_\theta(s)))\,ds\\
&+
\int_T^0
e^{-Cs}N(x_\theta(s))\,ds.
\end{aligned}
$$

Since $\theta$ is bounded and $C$ is stable, the boundary term tends to zero as $T\to-\infty$. The first integral becomes $K\theta(x_0)$, so the exact identity is

$$
K\theta(x_0)-\theta(x_0)
=
-
\int_{-\infty}^0
e^{-Cs}N(x_\theta(s))\,ds.
\tag{14.1}
$$

Thus the failure of $\theta$ to be a fixed point is precisely the stable propagation of its invariance defect.

### 14.4 The order estimate

There are constants $M_C,\alpha>0$ such that, for $s\le0$,

$$
\|e^{-Cs}\|
\le
M_Ce^{\alpha s}.
$$

After taking the neighborhood sufficiently small, the center trajectory has at most an arbitrarily weak backward exponential growth: for sufficiently small $\eta>0$,

$$
\|x_\theta(s)\|
\le
M_xe^{-\eta s}\|x_0\|,
\qquad
s\le0.
$$

Since $N(x)=\mathcal O(\|x\|^p)$,

$$
\|N(x_\theta(s))\|
\le
M_NM_x^p e^{-p\eta s}\|x_0\|^p.
$$

Choose $\eta$ so that $p\eta<\alpha$. From (14.1),

$$
\begin{aligned}
\|K\theta(x_0)-\theta(x_0)\|
&\le
M_CM_NM_x^p\|x_0\|^p
\int_{-\infty}^0
e^{(\alpha-p\eta)s}\,ds\\
&=
\frac{M_CM_NM_x^p}{\alpha-p\eta}
\|x_0\|^p.
\end{aligned}
$$

Therefore

$$
K\theta-\theta
=
\mathcal O(\|x\|^p).
$$

The stable integral changes only the constant, not the power of $\|x\|$.

### 14.5 Correcting the approximate graph

Write the desired fixed point as

$$
H=\theta+w.
$$

Since $H=KH$, the correction must satisfy

$$
w=K(\theta+w)-\theta.
$$

Define

$$
Lw:=K(\theta+w)-\theta.
$$

On the weighted set

$$
\mathcal L
=
\left\{
w:\|w(x)\|\le c\|x\|^p
\right\},
$$

the cutoff, the exponential estimate, and the local Lipschitz bounds for $F$ and $G$ imply that, for a sufficiently small neighborhood, $L$ maps $\mathcal L$ into itself and is a contraction. The contraction-mapping theorem gives a fixed point $w_*$ with

$$
w_*(x)=\mathcal O(\|x\|^p).
$$

Then

$$
H:=\theta+w_*
$$

satisfies $KH=H$ and hence defines a genuine center manifold. Since $\theta=h$ near the origin,

$$
H(x)=h(x)+\mathcal O(\|x\|^p).
$$

This justifies the practical coefficient-matching method: canceling the invariance residual through order $p-1$ produces a center-manifold approximation with an error of order $p$.


## 15. Practical Finite-Order Computation

To compute a center-manifold approximation:

1. Translate the equilibrium to the origin.
2. Compute the Jacobian.
3. Identify the center, stable, and unstable eigenvalues.
4. Apply a linear transformation separating the center and hyperbolic variables.
5. Write the center manifold as $y=h(x)$.
6. Use the tangency conditions

   $$
   h(0)=0,
   \qquad
   Dh(0)=0.
   $$

7. Choose a polynomial ansatz:

   $$
   h(x)
   =
   h^{(2)}(x)+h^{(3)}(x)+\cdots+h^{(m)}(x).
   $$

8. Substitute the ansatz into the invariance equation.
9. Match coefficients of equal degree.
10. Substitute the approximation into the center equation.
11. Analyze the first nonzero term of the reduced dynamics.

### Restrictions on the Truncation Order

The truncation order may be selected according to the problem, but it is not completely arbitrary.

First, the expansion begins at quadratic order because

$$
h(0)=0,
\qquad
Dh(0)=0.
$$

Second, the vector field and the selected center manifold must have enough differentiability. For example, $C^3$ regularity does not justify a fifth-order Taylor expansion.

Third, the necessary order is determined by the reduced dynamics. One normally computes only far enough to find the first nonzero term that controls stability or bifurcation. If the expected leading term vanishes, the calculation must be continued.

For polynomial or analytic vector fields, formal coefficient calculations can often be continued to higher finite orders. However, an analytic vector field does not necessarily possess an analytic or $C^\infty$ center manifold. A Taylor approximation is justified only to an order supported by the available center-manifold regularity.


## 16. Example 10.7

Consider

$$
\begin{aligned}
x'&=xy,\\
y'&=-y+\mu x^2.
\end{aligned}
$$

The eigenvalues at the origin are $0$ and $-1$, so $x$ is the center variable and $y$ is stable. Write the center manifold as $y=h(x)$. Its invariance equation is

$$
h'(x)\,x\,h(x)
=
-h(x)+\mu x^2,
$$

or equivalently,

$$
h'(x)\,x\,h(x)+h(x)-\mu x^2=0.
\tag{16.1}
$$

### 16.1 Why the expansion starts at second order

The conditions

$$
h(0)=0,
\qquad
h'(0)=0
$$

remove the constant and linear Taylor terms. Hence the first possible term is quadratic.

To compute through third order, take

$$
h(x)
=
a_2x^2+a_3x^3+\mathcal O(x^4).
$$

Then

$$
h'(x)
=
2a_2x+3a_3x^2+\mathcal O(x^3),
$$

and

$$
h'(x)\,x\,h(x)
=
2a_2^2x^4+\mathcal O(x^5).
$$

Thus this product contributes no quadratic or cubic term. Substitution into (16.1) gives

$$
(a_2-\mu)x^2+a_3x^3+\mathcal O(x^4)=0,
$$

so

$$
a_2=\mu,
\qquad
a_3=0.
$$

Therefore

$$
h(x)=\mu x^2+\mathcal O(x^4).
$$

A second-order calculation would already give $a_2=\mu$, but only the weaker statement $h(x)=\mu x^2+\mathcal O(x^3)$. The third-order ansatz is useful because it confirms that the cubic coefficient is zero.

### 16.2 Why not stop at, or continue beyond, third order?

The truncation order is chosen according to the desired reduced dynamics; it is not intrinsic to the manifold. A fourth-order ansatz is also valid:

$$
h(x)
=
a_2x^2+a_3x^3+a_4x^4+\mathcal O(x^5).
$$

Using $a_2=\mu$ and $a_3=0$, the fourth-order terms in (16.1) give

$$
2\mu^2+a_4=0,
$$

and hence

$$
h(x)
=
\mu x^2-2\mu^2x^4+\mathcal O(x^5).
$$

On the center manifold,

$$
x'=xh(x).
$$

The third-order calculation is already enough to obtain the first decisive term:

$$
x'
=
\mu x^3+\mathcal O(x^5).
$$

Thus the origin is locally asymptotically stable in the center direction when $\mu<0$ and unstable when $\mu>0$. Using the fourth-order graph gives the more accurate equation

$$
x'
=
\mu x^3-2\mu^2x^5+\mathcal O(x^6).
$$

In this example a term $a_kx^k$ in $h$ becomes a term $a_kx^{k+1}$ in the reduced equation because of the prefactor $x$. Therefore the order should be high enough to determine the first nonzero coefficient needed for stability or bifurcation analysis, but it need not be higher. The only additional restriction is smoothness: a Taylor expansion can be justified only to an order supported by the regularity of the vector field and the chosen center manifold.


## 17. Example 10.8: Preparing Standard Coordinates

Consider

$$
\begin{aligned}
x'&=y-1,\\
y'&=-(y-1)+\mu x^2+x(y-1).
\end{aligned}
\tag{10.9}
$$

The equilibrium is

$$
(x,y)=(0,1).
$$

Translate it to the origin by setting

$$
\xi=x,
\qquad
\eta=y-1.
$$

Then

$$
\begin{aligned}
\xi'&=\eta,\\
\eta'&=-\eta+\mu\xi^2+\xi\eta.
\end{aligned}
\tag{10.10}
$$

The Jacobian is

$$
A=
\begin{pmatrix}
0&1\\
0&-1
\end{pmatrix},
$$

with eigenvalues $0$ and $-1$.

Use the transformation

$$
\begin{pmatrix}
\xi\\
\eta
\end{pmatrix}
=
T
\begin{pmatrix}
u\\
v
\end{pmatrix},
\qquad
T=
\begin{pmatrix}
1&1\\
0&-1
\end{pmatrix}
=
T^{-1}.
$$

Thus,

$$
\xi=u+v,
\qquad
\eta=-v.
$$

Define

$$
Q(u,v)
=
\mu(u+v)^2-(u+v)v.
$$

The transformed system is

$$
\begin{aligned}
u'&=Q(u,v),\\
v'&=-v-Q(u,v).
\end{aligned}
$$

It is now in standard center-stable form. One may write

$$
v=h(u)
$$

and apply the invariance equation exactly as in Example 10.7. The lecture leaves this final center-manifold calculation as an exercise.


## 18. Discrete-Time Systems

Analogous center-manifold results hold for maps. Center directions correspond to multipliers on the unit circle.

Consider a split map

$$
\begin{aligned}
x_{n+1}&=Bx_n+F(x_n,y_n),\\
y_{n+1}&=Cy_n+G(x_n,y_n).
\end{aligned}
$$

If the center manifold is

$$
y=h(x),
$$

then graph invariance requires

$$
\boxed{
h\left(
Bx+F(x,h(x))
\right)
=
Ch(x)+G(x,h(x)).
}
$$

Unlike the continuous-time invariance equation, this condition contains composition rather than a derivative.

## 20. Main Logical Chain

The center-manifold method can be summarized as

$$
\boxed{
\text{spectral decomposition}
\longrightarrow
\text{center manifold}
\longrightarrow
\text{invariance equation}
\longrightarrow
\text{finite-order approximation}
\longrightarrow
\text{reduced dynamics}
\longrightarrow
\text{normal form}
\longrightarrow
\text{stability and bifurcation}.
}
$$

The key principles are:

$$
\boxed{
\begin{aligned}
&\text{Hyperbolic directions are governed primarily by their linear dynamics;}\\
&\text{loss of hyperbolicity occurs in the center directions;}\\
&\text{the center manifold contains the critical local dynamics;}\\
&\text{stable variables affect the coefficients through the graph }y=H(x);\\
&\text{finite-order approximations are sufficient for local bifurcation analysis;}\\
&\text{normal forms retain only the essential resonant nonlinearities.}
\end{aligned}
}
$$
