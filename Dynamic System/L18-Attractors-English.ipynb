## 1. General Setting

We consider a $C^0$ dynamical system

$$
(\mathcal X,\mathcal T,\phi_t)
$$

defined for all forward times $t\ge 0$. Typical cases are:

- $\mathcal T=[0,\infty)$ for a continuous-time flow;
- $\mathcal T=\mathbb N_0$ for a discrete-time map.

The aim is to identify subsets of phase space that contain the long-term behavior of trajectories.


## 2. Trapping Regions, Attracting Sets, and Attractors

### Definition 18.1 — Trapping Region

A compact set $\mathcal U\subset\mathcal X$ is a **trapping region** if

$$
\phi_t(\mathcal U)\subset\operatorname{int}(\mathcal U),
\qquad t>0.
$$

Thus every point in $\mathcal U$ is mapped strictly inside $\mathcal U$. The forward images are nested:

$$
t_2\ge t_1
\quad\Longrightarrow\quad
\phi_{t_2}(\mathcal U)\subseteq\phi_{t_1}(\mathcal U).
$$

### Definition 18.3 — Attracting Set

A set $\mathcal A$ is an **attracting set** if there exists a trapping region $\mathcal U$ such that

$$
\mathcal A=\bigcap_{t\ge0}\phi_t(\mathcal U).
$$

Hence $\mathcal A$ is the part of $\mathcal U$ that survives arbitrarily long forward evolution.

### Definition 18.3 — Attractor

An attracting set $\mathcal A$ is an **attractor** if it contains no proper nonempty attracting subset:

$$
\widetilde{\mathcal A}\ne\varnothing,\quad
\widetilde{\mathcal A}\subseteq\mathcal A,\quad
\widetilde{\mathcal A}\text{ attracting}
\quad\Longrightarrow\quad
\widetilde{\mathcal A}=\mathcal A.
$$

This is the precise meaning of **indecomposability** in the definition. A trapping region for $\mathcal A$ is also called an **isolating neighborhood**. Repelling sets and repellers are defined by reversing time.

### Lorenz Example

The Lorenz system is

$$
\begin{aligned}
x_1'&=\sigma(x_2-x_1),\\
x_2'&=\rho x_1-x_2-x_1x_3,\\
x_3'&=x_1x_2-\beta x_3,
\end{aligned}
\qquad \sigma,\rho,\beta>0.
$$

A trapping ellipsoid is normally verified by differentiating its defining function along the Lorenz vector field and showing that the field points inward on the boundary.

## 3. Basic Properties of Attracting Sets

### Theorem 18.4

Let $\mathcal V$ be a trapping region and let

$$
\mathcal A=\bigcap_{t\ge0}\phi_t(\mathcal V).
$$

Then:

1. $\mathcal A$ is closed. If the system is defined for positive and negative times, then $\mathcal A$ is invariant.
2. If $x_0\in\mathcal V$, then $\omega(x_0)\subseteq\mathcal A$.
3. If $x_*\in\mathcal A$ is a hyperbolic equilibrium and backward evolution exists, then

   $$
   W^u(x_*)\subseteq\mathcal A.
   $$

Here

$$
\omega(x_0)
=
\left\{
y:\exists t_n\to\infty,\ \phi_{t_n}(x_0)\to y
\right\}.
$$

### Main Proof Ideas

Each $\phi_t(\mathcal V)$ is compact and hence closed, so their intersection is closed. For a two-sided system, $\phi_s$ is invertible; re-indexing the nested intersection gives $\phi_s(\mathcal A)=\mathcal A$.

If $\phi_{t_n}(x_0)\to y$ with $t_n\to\infty$, then for every fixed $s\ge0$ and sufficiently large $n$,

$$
\phi_{t_n}(x_0)
=
\phi_s\bigl(\phi_{t_n-s}(x_0)\bigr)
\in\phi_s(\mathcal V).
$$

Since $\phi_s(\mathcal V)$ is closed, $y\in\phi_s(\mathcal V)$. This holds for every $s$, so $y\in\mathcal A$.

### Why Does $W^u(x_*)$ Appear?

The unstable manifold comes from the hyperbolicity of $x_*$, not from the attracting property of $\mathcal A$. For a hyperbolic equilibrium,

$$
T_{x_*}\mathcal X=E^s\oplus E^u,
$$

and the stable–unstable manifold theorem gives

$$
W^u(x_*)
=
\left\{
y:\phi_t(y)\to x_*
\text{ as }t\to-\infty
\right\}.
$$

Take $y\in W^u(x_*)$. Since $\phi_{-t}(y)\to x_*\in\operatorname{int}(\mathcal V)$, the backward orbit eventually lies in $\mathcal V$. Positive invariance then shows that its entire backward history lies in $\mathcal V$. Hence, for every $t\ge0$,

$$
y=\phi_t(\phi_{-t}(y))\in\phi_t(\mathcal V).
$$

Therefore

$$
\boxed{W^u(x_*)\subseteq\mathcal A}.
$$

If $E^u=\{0\}$, then $W^u(x_*)=\{x_*\}$. If $x_*$ is a saddle, the attracting set must also contain the unstable trajectories leaving it.


## 4. Basins and Milnor Attractors

### A Polar-Coordinate Example

Consider

$$
r'=1-r,
\qquad
\theta'
=
-\left(\theta-\frac14\right)
 \left(\theta-\frac34\right).
$$

Let

$$
\mathcal U=\left[\frac12,\frac32\right]\times S^1.
$$

The radial solution is

$$
r(t)=1+(r_0-1)e^{-t}.
$$

If $r_0\in[1/2,3/2]$, then

$$
1-\frac12e^{-t}\le r(t)\le1+\frac12e^{-t}.
$$

Thus $\phi_t(\mathcal U)\subset\operatorname{int}(\mathcal U)$ for every $t>0$, and

$$
\bigcap_{t\ge0}\phi_t(\mathcal U)
=
\{(r,\theta):r=1\}.
$$

The corresponding attracting set is the unit circle.

### Definition 18.6 — Basin of Attraction

For a closed invariant nonempty set $\mathcal A$, define

$$
\operatorname{BA}(\mathcal A)
=
\left\{
x_0\in\mathcal X:\omega(x_0)\subseteq\mathcal A
\right\}.
$$

The basin contains all initial points whose asymptotic limit points belong to $\mathcal A$. This does not require convergence to one point. If $\mathcal A$ is a periodic orbit, one may have $\omega(x_0)=\mathcal A$.

Under boundedness or precompactness assumptions, the condition is closely related to

$$
\operatorname{dist}\bigl(\phi_t(x_0),\mathcal A\bigr)\to0.
$$

A trapping region is generally only part of the basin:

$$
\mathcal U\subseteq\operatorname{BA}(\mathcal A).
$$

In the radial example, $r(t)\to1$ for every $r_0>0$, so the basin of the unit circle is the entire punctured plane.

### Definition 18.7 — Milnor Attractor

Let $\mu$ be a measure on $\mathcal X$. A closed invariant set $\mathcal A$ is a **Milnor attractor** if

$$
\mu\bigl(\operatorname{BA}(\mathcal A)\bigr)>0
$$

and there is no proper subset $\widetilde{\mathcal A}\subsetneq\mathcal A$ such that

$$
\mu\left(
\operatorname{BA}(\mathcal A)
\setminus
\operatorname{BA}(\widetilde{\mathcal A})
\right)=0.
$$

The first condition requires a non-negligible basin. The second is a measure-theoretic minimality condition:

> $\mathcal A$ cannot be replaced by a proper subset while retaining almost all of its basin.

It does not say that no smaller basin can exist, and it does not imply that limit points are uniformly distributed over $\mathcal A$. Uniform distribution concerns time averages or empirical measures such as

$$
\frac1T\int_0^T\delta_{\phi_t(x_0)}\,dt,
$$

which are not part of the Milnor-attractor definition.

### Example: $x'=-x^2$

The solution is

$$
x(t)=\frac{x_0}{1+x_0t}.
$$

For $x_0>0$, $x(t)\to0$. Thus $\{0\}$ attracts a positive-measure set of initial conditions. However, every neighborhood of $0$ contains negative initial conditions that move away from $0$, so $0$ is not Lyapunov stable.

$$
\boxed{\{0\}\text{ is a Milnor attractor but not a stable attractor.}}
$$

> **Source note.** In the polar-coordinate example, the printed minus sign makes $\theta=3/4$ stable and $\theta=1/4$ unstable, while the prose states the reverse. Furthermore, the stable equilibrium is itself a proper attracting set, so the full unit circle should be called an attracting set rather than an indecomposable attractor.


## 5. The Smale–Williams Solenoid

Let

$$
g(\theta,x,y)
=
\left(
2\theta,\,
\lambda x+\frac12\cos(2\pi\theta),\,
\lambda y+\frac12\sin(2\pi\theta)
\right),
\qquad 0<\lambda<\frac12.
$$

This is a discrete-time map:

$$
z_{n+1}=g(z_n).
$$

Its orbit is $z_0,g(z_0),g^2(z_0),\ldots$, and its attracting set is

$$
\mathcal A
=
\bigcap_{k=0}^{\infty}g^k(S^1\times\mathbb D^2).
$$

It is not a flow. In particular,

$$
g(S^1\times\mathbb D^2)
\subsetneq
S^1\times\mathbb D^2,
$$

so $g$ is not a surjective two-sided flow map on the same space.

Because $S^1=[0,1]/(0\sim1)$, the first component means

$$
\theta\mapsto2\theta\pmod1.
$$

Locally, angular distances are doubled; globally, the stretched torus wraps twice around the original torus. The transverse directions are contracted by $\lambda$. Since

$$
\lambda+\frac12<1,
$$

the image lies strictly inside the original solid torus.

Repeated iteration produces $2^k$ thinner transverse pieces. Infinite binary choices generate a Cantor-set direction, so the attractor locally resembles

$$
\text{Cantor set}\times\text{interval}.
$$

This attractor is called the **Smale–Williams solenoid**.


## 6. Chaotic, Hyperbolic, and Strange Attractors

An attractor $\mathcal A$ is:

- a **chaotic attractor** if it is also a chaotic set;
- a **hyperbolic attractor** if it is also a hyperbolic set;
- a **strange attractor** if it is chaotic but not hyperbolic.

The Smale–Williams solenoid is both hyperbolic and chaotic. Therefore it is not strange under this definition.

A complicated geometric appearance alone does not prove that an attractor is strange. One must establish both chaos and nonhyperbolicity.


## 7. The Lorenz Attractor Has Zero Volume

### Theorem 18.12

Let

$$
\mathcal E(t)=\phi_t(\mathcal E),
\qquad
V(t)=\operatorname{Vol}(\mathcal E(t)).
$$

A boundary element moves with normal velocity $\mathbf n\cdot f$. Therefore

$$
V'(t)
=
\int_{\partial\mathcal E(t)}
\mathbf n\cdot f\,dA.
$$

By the Divergence Theorem,

$$
V'(t)
=
\int_{\mathcal E(t)}
\nabla\cdot f\,dx.
$$

For the Lorenz vector field,

$$
\nabla\cdot f=-\sigma-1-\beta.
$$

Hence

$$
V'(t)=-(\sigma+1+\beta)V(t),
$$

so

$$
V(t)=V(0)e^{-(\sigma+1+\beta)t}\longrightarrow0.
$$

Since $\mathcal A\subseteq\mathcal E(t)$ for every $t$,

$$
0\le\operatorname{Vol}(\mathcal A)\le V(t)\to0.
$$

Therefore

$$
\boxed{\operatorname{Vol}(\mathcal A)=0}.
$$

Negative divergence means net volume contraction. It does not mean that every direction contracts: expansion in some directions may coexist with stronger contraction in other directions.


## 8. Box-Counting Dimension

For $S\subset\mathbb R^d$, let $N(r,S)$ be the minimum number of boxes of side length $r$ required to cover $S$. Define

$$
\dim_B(S)
=
\lim_{r\to0}
\frac{\ln N(r,S)}{\ln(1/r)},
$$

provided that the limit exists.

A $d$-dimensional set typically satisfies

$$
N(r,S)\approx Cr^{-d}.
$$

Taking logarithms gives

$$
\ln N(r,S)\approx\ln C+d\ln(1/r),
$$

so the quotient converges to $d$. Thus box-counting dimension measures the power-law growth of the number of boxes required as the observation scale becomes finer.

For the middle-third Cantor set, at stage $j$ there are $2^j$ intervals, each of length $3^{-j}$. Taking $r=3^{-j}$ gives

$$
N(r,\mathcal K)=2^j
$$

and therefore

$$
\dim_B(\mathcal K)
=
\frac{\ln2}{\ln3}
\approx0.6309.
$$


## 9. Main Conceptual Structure

$$
\boxed{
\text{trapping region}
\longrightarrow
\text{attracting set}
\longrightarrow
\text{indecomposable attractor}
}
$$

The concepts answer different questions:

- A **trapping region** asks whether trajectories remain confined.
- An **attracting set** describes what survives all forward evolution.
- An **attractor** asks whether a smaller attracting component remains.
- A **basin** identifies the initial conditions approaching the set.
- A **Milnor attractor** describes attraction for a positive-measure collection of initial conditions.
- **Hyperbolicity** and **chaos** describe internal dynamics.
- **Volume** and **box-counting dimension** describe geometric thinness and complexity.
