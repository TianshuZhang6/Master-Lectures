# 第 12 讲：全局分岔

## 学习目标

学习本讲后，你应当能够：

- 区分异宿轨道、同宿轨道、同宿环与极限环；
- 解释分裂函数的含义以及为什么 $\xi'(0)\neq0$ 是一般性破裂条件；
- 推导 Andronov-Leontovich 定理中的局部映射与全局映射；
- 通过返回映射判断分岔极限环的存在性与稳定性；
- 解释非共振条件在 Sternberg 线性化中的作用；
- 描述 Shilnikov 鞍焦点附近矩形变为螺旋带的过程；
- 解释 FitzHugh-Nagumo PDE 中同宿与周期行波剖面的含义。


## 1. 全局连接

考虑光滑一参数系统

$$
\dot{x}=f(x,p),
\qquad x\in\mathbb R^d,
\quad p\in\mathbb R,
\quad d\in\{2,3\},
$$

其中 $f(0,p)=0$，并假设 $x_*=0$ 是双曲鞍点。局部分岔研究平衡点或周期轨道附近的动力学；全局分岔则涉及稳定流形、不稳定流形及其全局连接。

若轨道 $\gamma(t)$ 满足

$$
\lim_{t\to-\infty}\gamma(t)=x_*,
\qquad
\lim_{t\to+\infty}\gamma(t)=y_*,
\qquad x_*\neq y_*,
$$

则称其为从 $x_*$ 到 $y_*$ 的**异宿轨道**。等价地，

$$
\gamma\subset W^u(x_*)\cap W^s(y_*).
$$

若非平凡轨道 $\gamma_0(t)$ 满足

$$
\lim_{t\to-\infty}\gamma_0(t)
=
\lim_{t\to+\infty}\gamma_0(t)
=0,
$$

则称其为连接到原点的**同宿轨道**。因此 $\gamma_0\subset W^u(0)\cap W^s(0)$。在二维系统中，这类鞍点连接通常是结构不稳定的：小扰动一般会使相应不变流形分开。

### 为什么明确写出 $\gamma_0\cup\{0\}$

若轨道定义为

$$
\gamma_0=\{\gamma_0(t):t\in\mathbb R\},
$$

则平衡点并不属于 $\gamma_0$。如果在某个有限时刻满足 $\gamma_0(t_0)=0$，由 ODE 解的唯一性可得 $\gamma_0(t)\equiv0$，这与轨道非平凡矛盾。因此，

$$
0\notin\gamma_0,
\qquad
\overline{\gamma_0}=\gamma_0\cup\{0\}.
$$

闭集 $\gamma_0\cup\{0\}$ 才是完整的几何**同宿环**。

同宿轨道不是极限环。极限环 $\Gamma_p$ 是一条具有有限周期 $T_p>0$ 的真正周期轨道：

$$
x_p(t+T_p)=x_p(t).
$$

极限环不能包含平衡点。当极限环从同宿环分岔出来时，沿存在极限环的参数一侧有 $\Gamma_p$ 在几何上趋近 $\gamma_0\cup\{0\}$。由于极限环越来越靠近鞍点，

$$
T_p\longrightarrow+\infty.
$$

所以同宿环是周期趋于无穷大的极限对象，但其自身并不是周期轨道。


## 2. 分裂函数

假设 $p=0$ 时存在同宿轨道。选择穿过同宿环上一个正则点的横截面 $\Sigma$，并把稳定流形与截面的交点规定为坐标零点。设返回的不稳定流形与 $\Sigma$ 的交点坐标为

$$
q=\xi(p).
$$

函数 $\xi$ 称为**分裂函数**，它测量稳定流形和不稳定流形之间的有符号距离：

$$
\xi(p)=0
\quad\Longleftrightarrow\quad
\text{同宿连接存在},
$$

而 $\xi(p)\neq0$ 表示连接已经破裂。一般性破裂条件是

$$
\xi(0)=0,
\qquad
\xi'(0)\neq0.
$$

局部有

$$
q=\xi(p)=\xi'(0)p+o(p).
$$

条件不是 $\xi'(0)=0$。导数非零表示参数穿过零点时，两条不变流形以一阶速度分开。

必须区分以下三件事：

- $\xi(p)=0$ 检测同宿连接；
- $\xi'(0)\neq0$ 保证参数展开非退化；
- 极限环由 Poincaré 返回映射的非零不动点给出：

  $$
  P_p(z)=z,
  \qquad z>0.
  $$

周期轨道并不是 $W^u$ 的一部分返回到 $W^u$。它由 $W^s$ 和 $W^u$ 附近的普通轨道组成，并通过截面上的一个普通点返回自身来确定。

$\xi$ 的符号依赖于截面坐标的方向。因此，在明确截面正方向以前，不能判断对应的是参数正向还是负向，也不能判断是同宿环内部还是外部。


## 3. Andronov-Leontovich 定理

假设 $d=2$，鞍点特征值满足

$$
\lambda_s(0)<0<\lambda_u(0).
$$

定义**鞍点量**

$$
\sigma(p)=\lambda_s(p)+\lambda_u(p).
$$

在二维系统中，$\sigma(p)=\operatorname{tr}D_xf(0,p)$。假设 $p=0$ 时存在同宿轨道 $\gamma_0$，并且

$$
\sigma(0)\neq0,
\qquad
\xi'(0)\neq0.
$$

那么，在 $p=0$ 的某一侧，存在唯一一条靠近 $\gamma_0\cup\{0\}$ 的极限环。其稳定性由 $\sigma(0)$ 决定：

- 若 $\sigma(0)<0$，极限环稳定；
- 若 $\sigma(0)>0$，极限环不稳定。

条件 $\sigma(0)\neq0$ 排除了稳定收缩和不稳定扩张具有相同主导强度的情形。

### 证明：拉直不变流形

在鞍点附近写成

$$
\begin{aligned}
\dot{x}_1&=\lambda_sx_1+f_1(x_1,x_2),\\
\dot{x}_2&=\lambda_ux_2+f_2(x_1,x_2).
\end{aligned}
$$

局部不变流形可表示为

$$
W^s_{\mathrm{loc}}(0)=\{x_2=S(x_1)\},
\qquad
W^u_{\mathrm{loc}}(0)=\{x_1=U(x_2)\},
$$

其中 $S(0)=S'(0)=U(0)=U'(0)=0$。坐标变换

$$
y_1=x_1-U(x_2),
\qquad
y_2=x_2-S(x_1)
$$

把两条流形拉直为坐标轴。变换后的系统为

$$
\begin{aligned}
\dot y_1&=y_1\bigl(\lambda_s+g_1(y)\bigr),\\
\dot y_2&=y_2\bigl(\lambda_u+g_2(y)\bigr),
\end{aligned}
$$

其中 $g(y)=O(\lVert y\rVert)$。之所以能提出因子 $y_1$ 和 $y_2$，是因为两条坐标轴是不变流形。

### 退出时间线性化

通过匹配非线性轨道与线性轨道离开小矩形的正向和反向退出时间，可以构造局部 $C^1$ 坐标变换。若 $\tau_s$ 是沿反向时间到达稳定边界的时间，$\tau_u$ 是沿正向时间到达不稳定边界的时间，则相应线性坐标可写为

$$
X=e^{\lambda_s\tau_s},
\qquad
Y=e^{-\lambda_u\tau_u}.
$$

在新坐标中，

$$
\dot X=\lambda_sX,
\qquad
\dot Y=\lambda_uY.
$$

变常数公式和 $g(y)=O(\lVert y\rVert)$ 表明非线性修正是高阶小量，并得到 $D\Phi(0)=I$。因此该变换是局部 $C^1$ 微分同胚。这个二维构造不需要 $C^\infty$ 线性化所要求的全阶非共振条件。

### 局部映射与全局映射

选取进入截面和离开截面

$$
\Sigma_{\mathrm{in}}=\{X=1,\ 0<Y\leq1\},
\qquad
\Sigma_{\mathrm{out}}=\{0<X\leq1,\ Y=1\}.
$$

从 $(1,Y)$ 出发，

$$
X(t)=e^{\lambda_st},
\qquad
Y(t)=Ye^{\lambda_ut}.
$$

轨道到达离开截面时 $Ye^{\lambda_uT}=1$，所以

$$
T=-\frac{1}{\lambda_u}\log Y,
\qquad
X(T)=Y^{-\lambda_s/\lambda_u}.
$$

令

$$
\rho=-\frac{\lambda_s}{\lambda_u}>0.
$$

局部映射为 $L(Y)=Y^\rho$。沿原同宿轨道外部部分的全局运动定义光滑映射

$$
G(X,p)=q+aX+O(X^2),
\qquad
q=\xi(p),
$$

其中在选取相容的截面方向后可令 $a>0$。完整返回映射为

$$
P_p(Y)=G\circ L(Y)=q+aY^\rho+\text{高阶项}.
$$

真正的极限环对应

$$
P_p(Y_*)=Y_*,
\qquad Y_*>0.
$$

也就是说，它是位移函数 $D(Y,p)=P_p(Y)-Y$ 的正根。边界值 $Y=0$ 对应分离轨道，其穿过鞍点邻域所需时间为无穷，因此不是 Poincaré 映射的普通点。边界处 $q=0$ 对应的是同宿连接。

### 存在性、唯一性与稳定性

指数与鞍点量的关系为

$$
\sigma=\lambda_s+\lambda_u=\lambda_u(1-\rho).
$$

当 $\sigma<0$ 时，$\rho>1$ 且 $Y^\rho=o(Y)$。在证明采用的坐标方向下，$q>0$ 时存在唯一小正根：

$$
Y_*=q+o(q),
\qquad
P_p'(Y_*)\sim a\rho Y_*^{\rho-1}\longrightarrow0.
$$

因此该极限环稳定。

当 $\sigma>0$ 时，$0<\rho<1$ 且 $Y^\rho$ 在原点附近主导 $Y$。在同一证明坐标中，$q<0$ 时存在唯一小正根：

$$
Y_*\sim\left(\frac{-q}{a}\right)^{1/\rho},
\qquad
P_p'(Y_*)\sim a\rho Y_*^{\rho-1}\longrightarrow+\infty.
$$

所以极限环不稳定。

这些 $q$ 的符号属于证明所选定的有向坐标。若反转截面方向，$q$ 的符号也会反转。若要判断原参数的正负，必须先确定实际分裂坐标与证明坐标的对应关系，然后再使用 $q=\xi'(0)p+o(p)$。极限环是在原同宿环内部还是外部，也必须通过截面正方向与几何内外的关系确定。

例如，在练习 E12.1 中，$x$ 轴坐标的正方向朝向同宿环外部，而返回映射证明采用的正方向朝向内部。由于

$$
\sigma=-2,
\qquad
\xi'(0)<0,
$$

增加练习中的参数 $\alpha$ 会使分离轨道向内部破裂，并在 $\alpha>0$ 时产生位于原同宿环内部的稳定极限环。


## 4. Sternberg 线性化与共振

设光滑映射 $g:\mathbb R^d\to\mathbb R^d$ 在原点有不动点，并且

$$
Dg(0)=\Lambda=\operatorname{diag}(\lambda_1,\ldots,\lambda_d).
$$

对多重指标 $k=(k_1,\ldots,k_d)$，定义

$$
|k|=k_1+\cdots+k_d,
\qquad
\lambda^k=\lambda_1^{k_1}\cdots\lambda_d^{k_d}.
$$

映射的非共振条件为

$$
\lambda_j\neq\lambda^k,
\qquad |k|\geq2.
$$

在 Sternberg 定理的假设下，该条件保证存在局部 $C^\infty$ 坐标变换 $H$，使得

$$
H\circ g=\Lambda\circ H.
$$

若要消去非线性单项式 $cx^ke_j$，相应同调方程需要除以 $\lambda^k-\lambda_j$。若该量非零，单项式可以被消去；若

$$
\lambda^k=\lambda_j,
$$

则发生共振，该项通常必须保留在正规形中。从动力学角度看，这意味着该非线性单项式与第 $j$ 个线性方向具有相同的缩放率。

Hartman-Grobman 定理在双曲不动点附近给出拓扑共轭；Sternberg 非共振条件给出更强的光滑共轭，因此能够保留导数和精确渐近信息。

对于特征值为 $\mu_j$ 的连续时间系统，共振条件采用加法形式：

$$
\mu_j=k_1\mu_1+\cdots+k_d\mu_d.
$$

前面的二维证明只需要特殊构造的 $C^1$ 线性化，因此与更强的 Sternberg 定理并不矛盾。


## 5. Shilnikov 同宿轨道

假设三维鞍焦点的特征值为

$$
\lambda_1=\lambda>0,
\qquad
\lambda_{2,3}=\alpha\pm i\omega,
\quad \alpha<0,
\quad \omega>0.
$$

这里

$$
\alpha=\operatorname{Re}\lambda_2=\operatorname{Re}\lambda_3.
$$

$\alpha$ 是每个复特征值的共同实部，不是两个实部之和。两个实部之和是 $2\alpha$，而线性化矩阵的迹为 $\lambda+2\alpha$。Shilnikov 鞍点量则为

$$
\sigma=\lambda+\alpha,
$$

这是因为稳定平面内的半径满足 $\dot r=\alpha r$。Shilnikov 条件为

$$
\lambda+\alpha>0
\quad\Longleftrightarrow\quad
\lambda>|\alpha|.
$$

在同宿参数处，该条件会在同宿环附近产生马蹄型不变集和无穷多个鞍型周期轨道。鞍型周期轨道有一个等于 $1$ 的平凡 Floquet 乘子，另外两个非平凡乘子满足

$$
|\mu_s|<1<|\mu_u|.
$$

固定的紧致马蹄不变集在足够小的扰动下可以持续存在，但这不表示同宿连接本身会对每个邻近参数继续存在。

### 矩形如何变成螺旋带

在适当坐标中，

$$
\begin{aligned}
\dot\zeta&=\lambda\zeta,\\
\dot\xi_1&=\alpha\xi_1-\omega\xi_2,\\
\dot\xi_2&=\omega\xi_1+\alpha\xi_2.
\end{aligned}
$$

令 $z=\xi_1+i\xi_2$，则

$$
\zeta(t)=\zeta_0e^{\lambda t},
\qquad
z(t)=z_0e^{(\alpha+i\omega)t}.
$$

在进入截面上固定 $\xi_1=c_h$，并用

$$
(\zeta,\xi_1,\xi_2)=(s,c_h,u),
\qquad
0<s<\varepsilon,
\quad |u|<\delta
$$

参数化一个小矩形。取离开截面 $\Sigma_v=\{\zeta=c_v\}$。飞行时间由 $se^{\lambda T}=c_v$ 确定，因此

$$
T(s)=\frac1\lambda\log\frac{c_v}{s}.
$$

当 $s\to0^+$ 时，$T(s)\to+\infty$。越靠近稳定流形的初始点，在鞍焦点附近停留越久，旋转的圈数也越多。令

$$
\nu=-\frac{\alpha}{\lambda}>0.
$$

稳定输出坐标具有形式

$$
z_{\mathrm{out}}
=
\left(\frac{s}{c_v}\right)^\nu
e^{i\theta(s)}(c_h+iu),
\qquad
\theta(s)=\frac{\omega}{\lambda}\log\frac{c_v}{s}.
$$

固定 $s$ 并改变 $u$ 会得到一条短线段。这条线段被旋转 $\theta(s)$，同时按 $s^\nu$ 缩小。当 $s$ 减小时：

- 旋转角趋于无穷；
- 线段到螺旋中心的距离按 $s^\nu$ 减小；
- 线段的绝对长度也按 $s^\nu$ 减小。

这些旋转短线段的并集是一条二维螺旋带，而不是一条单独轨道。螺旋带积聚于

$$
W^u_{\mathrm{loc}}(0)\cap\Sigma_v=(c_v,0,0),
$$

而不是三维相空间中的平衡点 $(0,0,0)$。$s=0$ 位于稳定流形上，不会到达离开截面。因此螺旋中心是一族不同轨道出口位置的极限，而不是某条轨道在 $\Sigma_v$ 上最终趋近的点。

螺旋带的绝对宽度满足 $\operatorname{width}\sim\delta s^\nu$。其中心半径也是 $s^\nu$ 阶，因此绝对宽度趋于零，但相对厚度未必趋于零。

### 拉伸来自哪里

螺旋带不是沿不稳定轴直接被拉长，因为在离开截面上 $\zeta=c_v$ 已固定。不稳定坐标的作用是控制飞行时间。由

$$
\theta'(s)=-\frac{\omega}{\lambda s}
$$

可见，很小的 $s$ 变化会产生很大的旋转角变化。局部映射满足

$$
\left\lVert\frac{\partial L}{\partial u}\right\rVert\asymp s^\nu,
\qquad
\left\lVert\frac{\partial L}{\partial s}\right\rVert\asymp s^{\nu-1}.
$$

Shilnikov 条件给出 $0<\nu<1$。所以 $s^\nu\to0$，而 $s^{\nu-1}\to+\infty$。矩形沿螺旋带横向被压缩，而沿螺旋方向在导数意义下被强烈拉伸。其来源是退出时间对初始位置的高度敏感性，而不是沿不稳定轴的直接几何拉长。

局部映射和全局映射满足

$$
L:\Sigma_h\to\Sigma_v,
\qquad
G:\Sigma_v\to\Sigma_h.
$$

所以以 $\Sigma_h$ 为基准的返回映射是 $P=G\circ L:\Sigma_h\to\Sigma_h$。图 31 的图注若把 $G\circ L$ 写成 $\Sigma_v$ 到自身的映射，则与正文不一致；若以 $\Sigma_v$ 为基准，返回映射应为 $L\circ G$。

全局映射只会对螺旋带产生有限的光滑变形，不会消除其无穷多圈。螺旋带与原矩形反复相交，从而产生马蹄动力学和无穷多个鞍型周期轨道。


## 6. FitzHugh-Nagumo 行波

考虑

$$
\begin{aligned}
u_t&=u_{xx}+f(u)-v,\\
v_t&=\varepsilon(u-v),
\end{aligned}
$$

其中

$$
f(u)=u(1-u)(u-p),
\qquad
0<p<\frac12.
$$

寻找行波

$$
u(x,t)=U(s),
\qquad
v(x,t)=V(s),
\qquad
s=x-ct.
$$

于是 $u_t=-cU'$、$u_{xx}=U''$、$v_t=-cV'$。令 $W=U'$，在 $c\neq0$ 时得到

$$
\begin{aligned}
U'&=W,\\
W'&=V-cW-f(U),\\
V'&=\frac{\varepsilon}{c}(V-U).
\end{aligned}
$$

此处的导数是对行波坐标 $s$ 求导，而不是对原时间 $t$ 求导。

### ODE 轨道的含义

若行波剖面系统有同宿轨道

$$
(U(s),W(s),V(s))\longrightarrow(0,0,0)
\qquad(s\to\pm\infty),
$$

则在原 PDE 中，它对应一个前后都回到静息状态的局部行波脉冲。剖面 ODE 的周期轨道则对应空间周期行波，即波列。空间 ODE 中的鞍型不稳定性并不能单独证明原 PDE 解在时间上的稳定性或谱稳定性。

### 平衡点

当 $\varepsilon\neq0$ 且 $c\neq0$ 时，平衡点满足 $W=0$ 和 $V=U=f(U)$。若 $U\neq0$，则要求

$$
(1-U)(U-p)=1.
$$

然而，

$$
\max_{U\in\mathbb R}(1-U)(U-p)=\frac{(1-p)^2}{4}<1.
$$

因此唯一平衡点是 $(U,W,V)=(0,0,0)$。

当 $\varepsilon=0$ 时，唯一性失效。完整三维系统具有平衡点曲线

$$
\{(U,0,f(U)):U\in\mathbb R\},
$$

而不变平面 $V=0$ 中包含 $(0,0,0)$、$(p,0,0)$ 和 $(1,0,0)$。

### Hamilton 奇异极限

当 $\varepsilon=0$ 时，$V$ 为常数。限制在 $V=0$ 上得到

$$
U'=W,
\qquad
W'=-cW-f(U).
$$

只有完成这一降维后才能令 $c=0$。不能直接在完整三维公式中代入 $c=0$，因为该公式是通过除以 $c$ 得到的。

当 $c=0$ 时，

$$
U'=W,
\qquad
W'=-f(U).
$$

Hamilton 函数为

$$
H(U,W)=\frac12W^2+F(U),
$$

其中

$$
F(U)=-\frac14U^4+\frac{1+p}{3}U^3-\frac p2U^2.
$$

沿轨道 $dH/ds=0$。由于 $f'(0)=-p$，原点是特征值为 $\pm\sqrt p$ 的鞍点。零能量曲线 $H(U,W)=0$ 包含一条连接到原点的同宿环。在 PDE 中，它对应一个静止的局部脉冲。

当 $\varepsilon=0$ 但 $c\neq0$ 时，

$$
\frac{dH}{ds}=-cW^2,
$$

因此原 Hamilton 同宿环通常会被破坏。在完整三维系统中，

$$
\frac{dH}{ds}=VW-cW^2.
$$

恢复变量 $V$ 可以在轨道的不同部分注入或移除能量，从而可能实现全局能量平衡。但要得到精确的三维同宿连接，仍需稳定流形和不稳定流形恰好匹配，因此证明困难得多。

对适当参数，原点可能具有一个正实特征值以及一对稳定复特征值 $\alpha\pm i\omega$。若存在同宿轨道并满足 $\lambda+\alpha>0$，则 Shilnikov 定理可以应用于该行波 ODE。

原文最后使用参数 $(c,a)$，与前文记号不一致。按这里的方程，相关参数应理解为 $c$、$p$ 和 $\varepsilon$。
