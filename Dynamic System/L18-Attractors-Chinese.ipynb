## 1. 一般设定

我们考虑一个对所有正向时间 $t\ge0$ 都有定义的 $C^0$ 动力系统（dynamical system）

$$
(\mathcal X,\mathcal T,\phi_t).
$$

典型情形包括：

- 对连续时间流（continuous-time flow），$\mathcal T=[0,\infty)$；
- 对离散时间映射（discrete-time map），$\mathcal T=\mathbb N_0$。

本讲的目标是找出相空间（phase space）中能够包含轨道长期行为的集合。


## 2. 捕获区域、吸引集与吸引子

### 定义 18.1——捕获区域（Trapping Region）

如果紧集 $\mathcal U\subset\mathcal X$ 满足

$$
\phi_t(\mathcal U)\subset\operatorname{int}(\mathcal U),
\qquad t>0,
$$

那么称 $\mathcal U$ 为捕获区域（trapping region）。这表示 $\mathcal U$ 中的点经过任意正时间后都会被映到其内部。

正向像形成嵌套集合族（nested family）：

$$
t_2\ge t_1
\quad\Longrightarrow\quad
\phi_{t_2}(\mathcal U)\subseteq\phi_{t_1}(\mathcal U).
$$

### 定义 18.3——吸引集（Attracting Set）

如果存在捕获区域 $\mathcal U$，使得

$$
\mathcal A=\bigcap_{t\ge0}\phi_t(\mathcal U),
$$

那么称 $\mathcal A$ 为吸引集（attracting set）。它表示将 $\mathcal U$ 不断向前演化后始终没有被排除的核心集合。

### 定义 18.3——吸引子（Attractor）

如果吸引集 $\mathcal A$ 不包含任何更小的非空吸引集，即

$$
\widetilde{\mathcal A}\ne\varnothing,\quad
\widetilde{\mathcal A}\subseteq\mathcal A,\quad
\widetilde{\mathcal A}\text{ 是吸引集}
\quad\Longrightarrow\quad
\widetilde{\mathcal A}=\mathcal A,
$$

那么称 $\mathcal A$ 为吸引子（attractor）。这就是不可分解性（indecomposability）的含义。

捕获区域也称为隔离邻域（isolating neighborhood）。通过反转时间可以类似定义排斥集（repelling set）和排斥子（repeller）。

### Lorenz 系统例子

Lorenz 系统（Lorenz system）为

$$
\begin{aligned}
x_1'&=\sigma(x_2-x_1),\\
x_2'&=\rho x_1-x_2-x_1x_3,\\
x_3'&=x_1x_2-\beta x_3.
\end{aligned}
$$

验证某个椭球是捕获区域时，通常沿 Lorenz 向量场（Lorenz vector field）计算其定义函数的导数，并证明向量场在边界上指向内部。

> **原文提示：**原文给出的椭球公式将平衡点 $0$ 放在边界上，而 $0$ 始终不动。因此该椭球无法严格满足捕获条件；原公式可能需要修正，或者需要将椭球略微扩大。


## 3. 吸引集的基本性质

### 定理 18.4

设 $\mathcal V$ 是捕获区域，并令

$$
\mathcal A=\bigcap_{t\ge0}\phi_t(\mathcal V).
$$

则：

1. $\mathcal A$ 是闭集。如果系统可以向正负两个时间方向演化，则 $\mathcal A$ 是不变集（invariant set）。
2. 如果 $x_0\in\mathcal V$，则 $\omega(x_0)\subseteq\mathcal A$。
3. 如果 $x_*\in\mathcal A$ 是双曲平衡点（hyperbolic equilibrium），并且系统可以向负时间演化，则

   $$
   W^u(x_*)\subseteq\mathcal A.
   $$

其中，$\omega$-极限集（$\omega$-limit set）定义为

$$
\omega(x_0)
=
\left\{
y:\exists t_n\to\infty,\ \phi_{t_n}(x_0)\to y
\right\}.
$$

### 证明的关键思路

每个 $\phi_t(\mathcal V)$ 都是紧集，因而是闭集，所以它们的交 $\mathcal A$ 也是闭集。对于双向时间系统（two-sided system），$\phi_s$ 可逆；重新编号嵌套集合的交可得 $\phi_s(\mathcal A)=\mathcal A$。

若 $\phi_{t_n}(x_0)\to y$ 且 $t_n\to\infty$，那么对任意固定的 $s\ge0$，当 $n$ 足够大时，

$$
\phi_{t_n}(x_0)
=
\phi_s\bigl(\phi_{t_n-s}(x_0)\bigr)
\in\phi_s(\mathcal V).
$$

由于 $\phi_s(\mathcal V)$ 是闭集，所以 $y\in\phi_s(\mathcal V)$。该结论对所有 $s$ 都成立，因此 $y\in\mathcal A$。

### 为什么会出现 $W^u(x_*)$？

不稳定流形（unstable manifold）来自 $x_*$ 的双曲性（hyperbolicity），并不是因为 $\mathcal A$ 是吸引集才出现。

对于双曲平衡点，

$$
T_{x_*}\mathcal X=E^s\oplus E^u.
$$

由稳定—不稳定流形定理（stable–unstable manifold theorem），

$$
W^u(x_*)
=
\left\{
y:\phi_t(y)\to x_*
\text{ 当 }t\to-\infty
\right\}.
$$

取 $y\in W^u(x_*)$。因为 $\phi_{-t}(y)\to x_*\in\operatorname{int}(\mathcal V)$，负向轨道最终会进入 $\mathcal V$。再利用 $\mathcal V$ 的正向不变性可知其整个负向历史都位于 $\mathcal V$ 中。因此，对任意 $t\ge0$，

$$
y=\phi_t(\phi_{-t}(y))\in\phi_t(\mathcal V).
$$

最终得到

$$
\boxed{W^u(x_*)\subseteq\mathcal A}.
$$

如果 $E^u=\{0\}$，则 $W^u(x_*)=\{x_*\}$。如果 $x_*$ 是鞍点（saddle），吸引集还必须包含从它向外展开的不稳定轨道。


## 4. 吸引域与 Milnor 吸引子

### 极坐标例子

考虑

$$
r'=1-r,
\qquad
\theta'
=
-\left(\theta-\frac14\right)
 \left(\theta-\frac34\right).
$$

取

$$
\mathcal U=\left[\frac12,\frac32\right]\times S^1.
$$

径向方程的解为

$$
r(t)=1+(r_0-1)e^{-t}.
$$

如果 $r_0\in[1/2,3/2]$，那么

$$
1-\frac12e^{-t}\le r(t)\le1+\frac12e^{-t}.
$$

因此对任意 $t>0$，都有 $\phi_t(\mathcal U)\subset\operatorname{int}(\mathcal U)$，并且

$$
\bigcap_{t\ge0}\phi_t(\mathcal U)
=
\{(r,\theta):r=1\}.
$$

相应的吸引集是单位圆。

### 定义 18.6——吸引域（Basin of Attraction）

对于非空、闭且不变的集合 $\mathcal A$，定义

$$
\operatorname{BA}(\mathcal A)
=
\left\{
x_0\in\mathcal X:\omega(x_0)\subseteq\mathcal A
\right\}.
$$

吸引域由所有长期渐近行为落在 $\mathcal A$ 中的初始点组成。这不要求轨道收敛到一个点。例如，如果 $\mathcal A$ 是周期轨道（periodic orbit），那么可能有 $\omega(x_0)=\mathcal A$。

在轨道有界或正半轨道预紧（precompact）的条件下，该条件通常与

$$
\operatorname{dist}\bigl(\phi_t(x_0),\mathcal A\bigr)\to0
$$

密切相关。

捕获区域通常只是吸引域的一部分：

$$
\mathcal U\subseteq\operatorname{BA}(\mathcal A).
$$

在上述径向例子中，任意 $r_0>0$ 都满足 $r(t)\to1$，所以单位圆的吸引域是整个穿孔平面（punctured plane）。

### 定义 18.7——Milnor 吸引子（Milnor Attractor）

设 $\mu$ 是 $\mathcal X$ 上的测度。如果闭不变集 $\mathcal A$ 满足

$$
\mu\bigl(\operatorname{BA}(\mathcal A)\bigr)>0
$$

并且不存在真子集 $\widetilde{\mathcal A}\subsetneq\mathcal A$ 使得

$$
\mu\left(
\operatorname{BA}(\mathcal A)
\setminus
\operatorname{BA}(\widetilde{\mathcal A})
\right)=0,
$$

那么称 $\mathcal A$ 为 Milnor 吸引子（Milnor attractor）。

第一条表示 $\mathcal A$ 吸引了一批具有正测度（positive measure）的初始条件。第二条是测度意义下的最小性（measure-theoretic minimality）：

> $\mathcal A$ 不能在保留几乎全部吸引域的同时继续缩小。

这里的“更小”指集合 $\widetilde{\mathcal A}\subsetneq\mathcal A$，而不是说不存在测度更小的吸引域。

该定义也不意味着极限点在 $\mathcal A$ 上均匀分布（uniform distribution）。均匀分布需要研究时间平均或经验测度（empirical measure），例如

$$
\frac1T\int_0^T\delta_{\phi_t(x_0)}\,dt,
$$

这些统计性质不属于 Milnor 吸引子的定义。

### 例子：$x'=-x^2$

方程的解为

$$
x(t)=\frac{x_0}{1+x_0t}.
$$

如果 $x_0>0$，那么 $x(t)\to0$，所以 $\{0\}$ 吸引了正测度的一批初始条件。但是，任意靠近 $0$ 的邻域中都存在负初值，这些轨道会向左远离 $0$。因此 $0$ 不是 Lyapunov 稳定的（Lyapunov stable）。

$$
\boxed{\{0\}\text{ 是 Milnor 吸引子，但不是稳定吸引子。}}
$$

> **原文提示：**极坐标例子中，按照原文打印的负号，$\theta=3/4$ 稳定，而 $\theta=1/4$ 不稳定，与后续文字相反。此外，稳定平衡点本身是单位圆内更小的吸引集，所以单位圆严格来说应称为吸引集，而不是不可分解吸引子。


## 5. Smale–Williams 螺线管吸引子

令

$$
g(\theta,x,y)
=
\left(
2\theta,\,
\lambda x+\frac12\cos(2\pi\theta),\,
\lambda y+\frac12\sin(2\pi\theta)
\right),
\qquad 0<\lambda<\frac12.
$$

这是离散时间映射（discrete-time map）：

$$
z_{n+1}=g(z_n).
$$

其轨道为 $z_0,g(z_0),g^2(z_0),\ldots$，相应的吸引集为

$$
\mathcal A
=
\bigcap_{k=0}^{\infty}g^k(S^1\times\mathbb D^2).
$$

这里的 $g$ 不是流（flow）。特别地，

$$
g(S^1\times\mathbb D^2)
\subsetneq
S^1\times\mathbb D^2,
$$

所以它不是同一空间上双向流的满射时间映射。

因为 $S^1=[0,1]/(0\sim1)$，第一分量表示

$$
\theta\mapsto2\theta\pmod1.
$$

局部上，圆周方向的距离被扩大两倍；整体上，拉长后的实心环面会绕原环面两圈。横截面方向按照比例 $\lambda$ 压缩。由于

$$
\lambda+\frac12<1,
$$

压缩后的横截面圆盘严格位于原圆盘内部。

第 $k$ 次迭代会产生 $2^k$ 个更细的横截面分支。无限二进制选择形成 Cantor 集（Cantor set）方向，因此吸引子局部具有

$$
\text{Cantor set}\times\text{interval}
$$

的乘积结构。这个吸引子称为 Smale–Williams 螺线管吸引子（Smale–Williams solenoid）。


## 6. 混沌、双曲与奇异吸引子

吸引子 $\mathcal A$ 可以分为：

- 混沌吸引子（chaotic attractor）：$\mathcal A$ 同时是混沌集（chaotic set）；
- 双曲吸引子（hyperbolic attractor）：$\mathcal A$ 同时是双曲集（hyperbolic set）；
- 奇异吸引子（strange attractor）：$\mathcal A$ 是混沌的，但不是双曲的。

Smale–Williams 螺线管既是双曲的，又是混沌的。因此按照本节定义，它不是奇异吸引子。

复杂的几何外观本身不能证明吸引子是奇异的；还必须证明混沌性（chaoticity）和非双曲性（nonhyperbolicity）。


## 7. Lorenz 吸引子的体积为零

### 定理 18.12

定义

$$
\mathcal E(t)=\phi_t(\mathcal E),
\qquad
V(t)=\operatorname{Vol}(\mathcal E(t)).
$$

边界上的小面积元以向量场速度 $f$ 运动，只有法向速度 $\mathbf n\cdot f$ 会改变区域体积。因此

$$
V'(t)
=
\int_{\partial\mathcal E(t)}
\mathbf n\cdot f\,dA.
$$

该积分表示边界上的净向外通量（outward flux）。由散度定理（Divergence Theorem），

$$
V'(t)
=
\int_{\mathcal E(t)}
\nabla\cdot f\,dx.
$$

对于 Lorenz 向量场，

$$
\nabla\cdot f=-\sigma-1-\beta.
$$

因此

$$
V'(t)=-(\sigma+1+\beta)V(t),
$$

从而

$$
V(t)=V(0)e^{-(\sigma+1+\beta)t}\longrightarrow0.
$$

因为对任意 $t$ 都有 $\mathcal A\subseteq\mathcal E(t)$，所以

$$
0\le\operatorname{Vol}(\mathcal A)\le V(t)\to0.
$$

最终得到

$$
\boxed{\operatorname{Vol}(\mathcal A)=0}.
$$

负散度表示三维体积整体收缩（volume contraction），但不表示所有方向都在收缩。某些方向可以发生拉伸，只要其他方向上的压缩更强。


## 8. 箱维数（Box-Counting Dimension）

设 $N(r,S)$ 是覆盖集合 $S$ 所需的、边长为 $r$ 的最少盒子数。定义

$$
\dim_B(S)
=
\lim_{r\to0}
\frac{\ln N(r,S)}{\ln(1/r)},
$$

前提是该极限存在。

一个 $d$ 维集合通常满足

$$
N(r,S)\approx Cr^{-d}.
$$

两边取对数得到

$$
\ln N(r,S)\approx\ln C+d\ln(1/r),
$$

因此上述商趋于 $d$。所以箱维数测量的是：

> 当观察尺度变细时，覆盖集合所需盒子数增长的幂指数。

对于三分 Cantor 集（middle-third Cantor set），第 $j$ 步具有 $2^j$ 个区间，每个区间长度为 $3^{-j}$。取 $r=3^{-j}$，则

$$
N(r,\mathcal K)=2^j,
$$

因此

$$
\dim_B(\mathcal K)
=
\frac{\ln2}{\ln3}
\approx0.6309.
$$


## 9. 核心概念结构

$$
\boxed{
\text{捕获区域}
\longrightarrow
\text{吸引集}
\longrightarrow
\text{不可分解吸引子}
}
$$

不同概念回答不同问题：

- 捕获区域（trapping region）：轨道是否会被限制在某一区域中？
- 吸引集（attracting set）：不断正向演化后最终留下什么集合？
- 吸引子（attractor）：该吸引集内部是否还存在更小的吸引集？
- 吸引域（basin of attraction）：哪些初始点会渐近趋向该集合？
- Milnor 吸引子（Milnor attractor）：该集合是否吸引正测度的初始条件，并且在模零意义下不可继续缩小？
- 双曲性（hyperbolicity）与混沌性（chaoticity）：吸引子内部的动力学结构是什么？
- 体积（volume）与箱维数（box-counting dimension）：吸引子在几何上有多薄、多复杂？
