# L7 Hyperbolic Sets

基于讲义第 7 节 *Hyperbolic Sets*。本讲说明双曲性为何在 $C^1$ 小扰动下保持，并由此得到阴影、拓扑共轭与 Anosov 微分同胚的结构稳定性。

## 1. 几何背景：黎曼流形与切映射

令 $M$ 为 $C^1$ 黎曼流形。黎曼度量在每个切空间 $T_xM$ 上给出向量长度 $\lvert v\rvert$，再由曲线长度诱导距离 $d(x,y)$。

取一个 $C^1$ 微分同胚

$$
g:M\to M.
$$

它在点 $x$ 处的导数（切映射）是

$$
Dg_x:T_xM\to T_{g(x)}M.
$$

双曲性研究的对象是切向量在反复作用 $Dg_x$ 后的收缩与扩张，而不是只研究轨道点的位置。

## 2. 双曲集：定义 7.1

紧的 $g$-不变集 $\Lambda\subset M$ 称为双曲集，若存在 $C>0$、$\lambda\in(0,1)$，以及每个 $x\in\Lambda$ 处的子空间 $E^s(x),E^u(x)\subset T_xM$，满足：

- **不变性**

  $$
  Dg_xE^s(x)=E^s(g(x)),\qquad Dg_xE^u(x)=E^u(g(x)).
  $$

**对Es 上的向量做映射， 相当于映射后 的Es 方向**

- **稳定方向的正向收缩**

  $$
  \left\lvert Dg_x^jv^s\right\rvert\le C\lambda^j\left\lvert v^s\right\rvert,
  \qquad v^s\in E^s(x),\quad j\ge0.
  $$

- **非稳定方向的反向收缩**

  $$
  \left\lvert Dg_x^{-j}v^u\right\rvert\le C\lambda^j\left\lvert v^u\right\rvert,
  \qquad v^u\in E^u(x),\quad j\ge0.
  $$

- **直和分裂**

  $$
  T_xM=E^s(x)\oplus E^u(x).
  $$

因此，$E^s$ 是正向指数收缩方向，$E^u$ 是反向指数收缩、等价于正向指数扩张方向。

## 3. 切丛：定义 7.2

切丛把所有点处的切空间收集起来：

$$
TM=\bigcup_{x\in M}T_xM.
$$

若 $M$ 的维数为 $d$，则 $TM$ 的维数为 $2d$。其中一部分坐标记录基点 $x$，另一部分坐标记录向量 $v\in T_xM$。

稳定与非稳定方向不是固定的单个线性空间，而是切丛中的连续子丛：

$$
x\longmapsto E^s(x),\qquad x\longmapsto E^u(x).
$$

## 4. 例子与 Anosov 微分同胚

讲义考虑环面 $\mathbb{T}^2$ 上的线性自同构

$$
g(x)=Ax,
\qquad
A=
\begin{pmatrix}
2 & 1\\
1 & 1
\end{pmatrix}.
$$

矩阵 $A$ 有一个特征值的绝对值小于 $1$，另一个大于 $1$；相应特征方向分别给出稳定与非稳定方向。因此整个环面是双曲集。

若整个流形都是双曲集，即

$$
\Lambda=M,
$$

则称 $g$ 为 **Anosov 微分同胚**。

## 5. 定理 7.5：将近似交换图修正为精确交换图

设 $\Lambda$ 是 $g$ 的双曲集。若 $f$ 足够接近 $g$，并且 $h:Y\to Y$ 与连续映射 $\varphi:Y\to U$ 满足近似交换关系

$$
\varphi\circ h\approx f\circ\varphi,
$$

则存在**唯一**且接近 $\varphi$ 的连续映射 $\psi:Y\to U$，使图精确交换：

$$
\psi\circ h=f\circ\psi.
$$

含义：在双曲集附近，近似轨道关系可以被修正为精确轨道关系。证明使用 Banach 不动点定理。

若 $Y=\{y\}$ 且 $h(y)=y$，则该定理说明双曲不动点在小扰动后仍延续为不动点。

## 6. 伪轨道与阴影定理：定义 7.7 与定理 7.8

序列 $(x_k)$ 是 $g$ 的 $\delta$-伪轨道，若

$$
d\bigl(g(x_k),x_{k+1}\bigr)\le\delta
\qquad\text{对所有 }k.
$$

**shadow theorem**：给定 $\varepsilon>0$，存在 $\delta>0$，使得任意始终靠近 $\Lambda$ 的 $\delta$-伪轨道，都被某条真实轨道阴影：存在 $x$ 使

$$
d\bigl(g^k(x),x_k\bigr)<\varepsilon
\qquad\text{对所有 }k.
$$

证明将伪轨道点集赋予离散拓扑，并定义 $h(x_k)=x_{k+1}$。通常嵌入 $\varphi(x_k)=x_k$ 满足

$$
\varphi\circ h\approx g\circ\varphi,
$$

再对 $f=g$ 应用定理 7.5，即得精确轨道关系。

**实际上找一个关于几次映射一直存在于整个轨道领域的点，（因为映射是连续的,且收缩（hyperbolic）所以一定存在这样的初始点**

## 7. 定理 7.9 (T1)：双曲性的持续性

若 $\Lambda$ 是 $g$ 的双曲集，则存在邻域 $\mathcal{U}(\Lambda)$ 和 $\varepsilon_1>0$，使得：若 $K\subset\mathcal{U}(\Lambda)$ 是 $f$ 的紧不变集，且

$$
d_1(f,g)<\varepsilon_1,
$$

则 $K$ 是 $f$ 的双曲集。

### 锥场构造

将 $E_g^s,E_g^u$ 延拓到 $\mathcal{U}(\Lambda)$。对

$$
v=v^s+v^u,
$$

定义稳定锥与非稳定锥：

$$
K_\alpha^s(x)=\left\{v\in T_xM:\lvert v^u\rvert\le\alpha\lvert v^s\rvert\right\},
$$

$$
K_\alpha^u(x)=\left\{v\in T_xM:\lvert v^s\rvert\le\alpha\lvert v^u\rvert\right\}.
$$

当 $\alpha$ 足够小时，稳定锥中的向量正向收缩，非稳定锥中的向量反向收缩。这些严格估计在 $C^1$ 小扰动下仍成立。

### 讨论中的澄清：无限交筛选了什么？

**问：$E_f^s$ 是否就是正向迭代后始终留在稳定锥中的、更强的部分？**

是。定义

$$
E_f^s(x)=\bigcap_{k\ge0}(Df^{-k})_{f^k(x)}K_\alpha^s\bigl(f^k(x)\bigr).
$$

等价地，

$$
v\in E_f^s(x)
\quad\Longleftrightarrow\quad
(Df^k)_xv\in K_\alpha^s\bigl(f^k(x)\bigr)
\quad\text{对所有 }k\ge0.
$$

$E_f^u$ 对称地由反向迭代定义：

$$
E_f^u(x)=\bigcap_{k\ge0}(Df^k)_{f^{-k}(x)}K_\alpha^u\bigl(f^{-k}(x)\bigr).
$$

它筛掉“当前看起来主要稳定、但带有少量非稳定分量”的向量；这种向量通常最终会离开稳定锥。

**问：仅写出 $E_f^s,E_f^u$ 是否已经证明 $K$ 双曲？**

还没有。还需验证：

1. $E_f^s(x),E_f^u(x)$ 是连续线性子空间，且

   $$
   T_xM=E_f^s(x)\oplus E_f^u(x).
   $$

2. 它们对 $Df$ 不变：

   $$
   Df_xE_f^s(x)=E_f^s(f(x)),
   \qquad
   Df_xE_f^u(x)=E_f^u(f(x)).
   $$

3. 它们满足统一收缩估计：

   $$
   \left\lvert (Df^n)_xv^s\right\rvert\le\lambda^n\lvert v^s\rvert,
   \qquad
   \left\lvert (Df^{-n})_xv^u\right\rvert\le\lambda^n\lvert v^u\rvert.
   $$

结合 $K$ 已紧且 $f$-不变，便得到 $K$ 是 $f$ 的双曲集。

## 8. 定理 7.9 (T2)：双曲动力学的拓扑共轭

给定 $\Lambda$ 的任意邻域 $\mathcal{V}$ 和任意 $\varepsilon>0$，若 $f$ 与 $g$ 足够 $C^1$ 接近，则存在 $f$ 的双曲集 $K\subset\mathcal{V}$ 和同胚

$$
\psi:K\to\Lambda,
$$

满足

$$
\psi\circ f|_K=g|_\Lambda\circ\psi,
\qquad
d_0(\psi,\operatorname{Id})<\varepsilon.
$$

### 第一次构造：从 $\Lambda$ 到 $K$

取

$$
Y=\Lambda,
\qquad
h=g|_\Lambda,
$$

并令 $\varphi:\Lambda\hookrightarrow M$ 为通常嵌入。定理 7.5 给出 $\psi_1:\Lambda\to M$，满足

$$
\psi_1\circ g|_\Lambda=f\circ\psi_1.
$$

令

$$
K=\psi_1(\Lambda).
$$

则

$$
f(K)=f\bigl(\psi_1(\Lambda)\bigr)
=\psi_1\bigl(g(\Lambda)\bigr)
=\psi_1(\Lambda)=K.
$$

$K$ 是紧集，且 $\psi_1$ 充分接近恒等嵌入时有 $K\subset\mathcal{U}(\Lambda)$；由 (T1)，$K$ 是 $f$ 的双曲集。

### 讨论中的澄清：为什么要第二次构造？

**问：第一次已经得到 $\psi_1$，为什么不直接称其为共轭？**

第一次只得到半共轭

$$
\psi_1\circ g=f\circ\psi_1.
$$

它没有自动保证 $\psi_1$ 为一一对应，也没有给出逆映射。

**问：第二次构造如何得到？**

交换角色：把 $f$ 视为原映射，把 $g$ 视为 $f$ 的小扰动。取

$$
Y=K,
\qquad
h=f|_K,
$$

并令 $\varphi:K\hookrightarrow M$ 为通常嵌入。因为

$$
\varphi\circ f|_K=f\circ\varphi,
$$

定理 7.5 给出 $\psi_2:K\to M$，使

$$
\psi_2\circ f|_K=g\circ\psi_2.
$$

令

$$
\theta=\psi_2\circ\psi_1.
$$

则

$$
\theta\circ g
=\psi_2\circ f\circ\psi_1
=g\circ\theta.
$$

$\theta$ 接近恒等嵌入，且与 $g$ 交换。由定理 7.5 的唯一性（等价地，由双曲集的扩张性），

$$
\theta=\operatorname{Id}_\Lambda.
$$

所以

$$
\psi_2\circ\psi_1=\operatorname{Id}_\Lambda,
\qquad
\psi_2=\psi_1^{-1}.
$$

第二次构造的作用正是产生逆映射，将半共轭提升为拓扑共轭。


**remark: the whole process is because the $\Lambda$ can be directe decomposed and the decomposed part Es and Eu can be extended into the neighbourhood field, and that neiberhood field also is can be marked by f and Decomposed to f's Es and Eu. finally use the homeomorphism to get the $\mathcal{K}$**

## 9. 推论 7.10：Anosov 微分同胚的开性与结构稳定性

若 $g$ 是 Anosov 微分同胚，则

$$
\Lambda=M.
$$

- 对 (T1) 取 $\Lambda=M$。当 $f$ 足够 $C^1$ 接近 $g$ 时，$M$ 仍是 $f$ 的双曲集，因此 $f$ 仍是 Anosov。故 Anosov 微分同胚在 $C^1$ 拓扑下构成开集。

- 对 (T2) 取 $\Lambda=M$ 和 $\mathcal{V}=M$。得到接近恒等映射的同胚

  $$
  \psi:M\to M,
  $$

  使

  $$
  \psi\circ f=g\circ\psi.
  $$

  因而扰动前后的全局动力学拓扑共轭，Anosov 微分同胚结构稳定。

(T1) 保证双曲分裂持续存在；(T2) 进一步保证轨道的拓扑结构也保持不变。
