# Lecture 9: Normal Forms

These notes develop the Poincare normal-form method for smooth ODEs near an equilibrium. The focus is on near-identity coordinate changes, the homological equation, resonances, the planar Hopf normal form, and the Bogdanov-Takens normal form.

## Learning objectives

- Derive the homological equation for a polynomial coordinate change.
- Separate removable nonlinear terms from resonant terms.
- Use kernels and adjoint kernels to choose normal-form complements.
- Explain the cubic Hopf normal form.
- Recognize the Bogdanov-Takens normal form.


## 1. Setup

Consider a sufficiently smooth autonomous ODE

$$x'=f(x),\qquad x\in\mathbb R^d,$$

with an equilibrium at $x=0$. Let $A=Df(0)$. Its Taylor expansion is

$$x'=Ax+f^{(2)}(x)+f^{(3)}(x)+\cdots,$$

where $f^{(k)}\in\mathcal H_k$ and $\mathcal H_k$ is the space of $d$-component homogeneous polynomial vector fields of degree $k$. Thus, for $h\in\mathcal H_k$,

$$h(\alpha x)=\alpha^k h(x),\qquad h(x)=O(|x|^k),\qquad Dh(x)=O(|x|^{k-1}).$$

The dimension is

$$\dim\mathcal H_k=d\binom{d+k-1}{k}.$$


## 2. One normal-form step at degree $k$

Fix $k\geq2$ and make a near-identity polynomial coordinate change

$$x=y+h^{(k)}(y),\qquad h^{(k)}\in\mathcal H_k.$$

For short, write $h=h^{(k)}$. Its inverse, to the required order, is

$$y=x-h(x)+O(|x|^{k+1}).$$

Indeed, the next correction is of order $Dh(y)h(y)=O(|y|^{2k-1})$, and $2k-1\geq k+1$ for $k\geq2$. Differentiation gives

$$y'=x'-Dh(x)x'+O(|x|^{k+1}).$$

Substituting the Taylor expansion and retaining only terms through degree $k$ yields

$$y'=Ax+\sum_{j=2}^{k}f^{(j)}(x)-Dh(x)Ax+O(|x|^{k+1}).$$

The discarded product $Dh(x)f^{(j)}(x)$ has degree $k+j-1\geq k+1$ for every $j\geq2$. Replacing $x$ by $y+h(y)$ then gives

$$\begin{aligned}
y'={}&Ay+\sum_{j=2}^{k-1}f^{(j)}(y)\\
&+\left[f^{(k)}(y)-\bigl(Dh^{(k)}(y)Ay-Ah^{(k)}(y)\bigr)\right]+O(|y|^{k+1}).
\end{aligned}$$


## 3. Homological equation and resonances

Define the homological operator

$$\boxed{\mathcal L_Ah(y)=Dh(y)Ay-Ah(y).}$$

Hence a degree-$k$ transformation changes the degree-$k$ term according to

$$f^{(k)}\longmapsto f^{(k)}-\mathcal L_Ah^{(k)}.$$

All terms of degree below $k$ remain unchanged. To eliminate a degree-$k$ term, solve the **homological equation**

$$\boxed{\mathcal L_Ah^{(k)}=f^{(k)}.}$$

Choose a complement $\mathcal R_k$ such that

$$\mathcal H_k=\operatorname{range}(\mathcal L_A)\oplus\mathcal R_k.$$

Writing $f^{(k)}=g^{(k)}+r^{(k)}$, with $g^{(k)}$ in the range and $r^{(k)}\in\mathcal R_k$, choose $h^{(k)}$ so that $\mathcal L_Ah^{(k)}=g^{(k)}$. The surviving term is $r^{(k)}$.

- Elements of $\operatorname{range}(\mathcal L_A)$ are **removable**.
- Elements in the chosen complement $\mathcal R_k$ are **resonant terms**.
- The choice of complement is not unique, so normal forms are generally not unique.

The operator also has the Lie-bracket form

$$\mathcal L_Ah=[h,Ax]=(Dh)Ax-(D(Ax))h.$$


## 4. Poincare normal form and the linear-algebra viewpoint

For any fixed order $m$, there is a polynomial near-identity transformation

$$x=y+h^{(2)}(y)+\cdots+h^{(m)}(y)+O(|y|^{m+1})$$

that transforms the system into

$$y'=Ay+r^{(2)}(y)+\cdots+r^{(m)}(y)+O(|y|^{m+1}),\qquad r^{(j)}\in\mathcal R_j.$$

This is the **Poincare normal-form theorem**. The construction is recursive: normalize degree $2$, then degree $3$, and so on.

Given an inner product on $\mathcal H_k$, a natural complement is

$$\boxed{\mathcal R_k=\ker(\mathcal L_A^*)=(\operatorname{range}\mathcal L_A)^\perp.}$$

Equivalently, the equation $\mathcal L_Ah=b$ is solvable exactly when $b$ is orthogonal to every vector in $\ker(\mathcal L_A^*)$. If zero is a semisimple eigenvalue of $\mathcal L_A$, then

$$\mathcal H_k=\ker(\mathcal L_A)\oplus\operatorname{range}(\mathcal L_A),$$

so one may choose $\mathcal R_k=\ker(\mathcal L_A)$.

If $A=\operatorname{diag}(\lambda_1,\ldots,\lambda_d)$ over $\mathbb C$, then for $h(y)=y^\alpha e_i$,

$$\mathcal L_A(y^\alpha e_i)=\left(\sum_{j=1}^d\alpha_j\lambda_j-\lambda_i\right)y^\alpha e_i.$$

Thus the resonance condition is

$$\boxed{\sum_{j=1}^d\alpha_j\lambda_j=\lambda_i.}$$


## 5. Hopf normal form

At a planar Hopf point, the eigenvalues are $\pm i\omega$, with $\omega>0$. After a linear change of state variables and the time rescaling $\tau=\omega t$, the linear part becomes

$$J=\begin{pmatrix}0&-1\\1&0\end{pmatrix}.$$

The Poincare normal form through cubic order is

$$\boxed{y'=Jy+(y_1^2+y_2^2)\begin{pmatrix}p_1&-p_2\\p_2&p_1\end{pmatrix}y+O(|y|^4).}$$

### Quadratic order

Using the ordered basis

$$(x^2\partial_x,xy\partial_x,y^2\partial_x,x^2\partial_y,xy\partial_y,y^2\partial_y),$$

the source's Hopf calculation uses $\widehat{\mathcal L}_J=-\mathcal L_J$, represented by

$$M^{(2)}=\begin{pmatrix}
0&-1&0&-1&0&0\\
2&0&-2&0&-1&0\\
0&1&0&0&0&-1\\
1&0&0&0&-1&0\\
0&1&0&2&0&-2\\
0&0&1&0&1&0
\end{pmatrix}.$$

Since $\det M^{(2)}=9\neq0$, every quadratic term is removable.


### Cubic order

For the basis

$$(x^3\partial_x,x^2y\partial_x,xy^2\partial_x,y^3\partial_x,x^3\partial_y,x^2y\partial_y,xy^2\partial_y,y^3\partial_y),$$

the cubic matrix has rank $6$. Hence the cubic resonant space has dimension $8-6=2$. A convenient kernel basis is

$$v=(1,0,1,0,0,1,0,1)^\top,$$

$$w=(0,-1,0,-1,1,0,1,0)^\top.$$

These correspond to the vector fields

$$v\longleftrightarrow (x^2+y^2)\begin{pmatrix}x\\y\end{pmatrix},\qquad w\longleftrightarrow (x^2+y^2)\begin{pmatrix}-y\\x\end{pmatrix}.$$

Therefore the cubic normal form is precisely

$$p_1(x^2+y^2)\begin{pmatrix}x\\y\end{pmatrix}+p_2(x^2+y^2)\begin{pmatrix}-y\\x\end{pmatrix}.$$

In complex notation $z=x+iy$,

$$z'=iz+(p_1+ip_2)z|z|^2+O(|z|^4).$$

The resonance condition for a monomial $z^p\bar z^q$ in the $z$ equation is $p-q=1$. There is no quadratic solution, while at cubic order the resonant monomial is $z^2\bar z=z|z|^2$.

In polar coordinates,

$$r'=p_1r^3+O(r^4),\qquad \theta'=1+p_2r^2+O(r^3).$$

Thus $p_1$ gives the leading radial effect and $p_2$ gives the leading correction to angular velocity.


## 6. Parameter-dependent systems and Bogdanov-Takens

A parameter-dependent system

$$x'=f(x,p)$$

is treated as an autonomous system by adjoining

$$p'=0.$$

For a generic planar two-parameter system with a double-zero eigenvalue, the Bogdanov-Takens normal form is

$$\boxed{\begin{pmatrix}y_1'\\y_2'\end{pmatrix}=\begin{pmatrix}0\\q_1\end{pmatrix}+\begin{pmatrix}0&1\\q_2&0\end{pmatrix}\begin{pmatrix}y_1\\y_2\end{pmatrix}+\begin{pmatrix}0\\y_1^2\pm y_1y_2\end{pmatrix}.}$$

Equivalently,

$$y_1'=y_2,\qquad y_2'=q_1+q_2y_1+y_1^2\pm y_1y_2.$$

The calculation starts from the nilpotent matrix

$$A=\begin{pmatrix}0&1\\0&0\end{pmatrix}.$$

Two further questions arise: when higher-order terms can be dropped, and how a high-dimensional system can first be reduced. The latter motivates center-manifold theory.


## 7. Practical workflow and exercises

For each degree $k=2,3,\ldots,m$:

1. Compute $A=Df(0)$ and the homogeneous Taylor terms.
2. Choose a monomial basis of $\mathcal H_k$.
3. Construct the matrix of $\mathcal L_A$.
4. Compute its range, kernel, and adjoint kernel.
5. Decompose $f^{(k)}$ into removable and resonant parts.
6. Solve the homological equation for the removable part.
7. Apply the coordinate change and continue to degree $k+1$.

Show that

$$x'=\begin{pmatrix}0&1\\0&0\end{pmatrix}x+O(|x|^2)$$

can be transformed into

$$y'=\begin{pmatrix}0&1\\0&0\end{pmatrix}y+\begin{pmatrix}0\\p_1y_1^2+p_2y_1y_2\end{pmatrix}+O(|y|^3).$$

### Exercise E9.2

For maps, if $x^+=Ax+f^{(k)}(x)+O(|x|^{k+1})$, the degree-$k$ map homological operator can be written as

$$\mathcal L_A^{\mathrm{map}}h=h\circ A-Ah,$$

so that $\widetilde f^{(k)}=f^{(k)}-\mathcal L_A^{\mathrm{map}}h$. Apply this to

$$g\begin{pmatrix}x\\y\end{pmatrix}=\begin{pmatrix}x+(x-y)^2\\3y+x^2+2y^2\end{pmatrix}.$$

## Key takeaway

$$\boxed{\widetilde f^{(k)}=f^{(k)}-\mathcal L_Ah^{(k)}}$$

is the mechanism that removes nonresonant nonlinear terms degree by degree.
