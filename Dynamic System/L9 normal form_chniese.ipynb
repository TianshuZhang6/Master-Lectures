# 第 9 讲：正规形（Normal Forms）

本讲介绍如何在平衡点附近，通过近恒等坐标变换逐阶消除非线性项，从而得到 Poincare 正规形。主要内容包括同调方程、共振项、Hopf 正规形和 Bogdanov-Takens 正规形。

## 学习目标

- 推导多项式坐标变换对应的同调方程。
- 区分可消去项与共振项。
- 利用核与伴随核选择正规形的补空间。
- 理解平面 Hopf 分岔中的三次正规形。
- 识别 Bogdanov-Takens 正规形。


## 1. 问题设定

考虑充分光滑的自治常微分方程

$$x'=f(x),\qquad x\in\mathbb R^d,$$

并设 $x=0$ 是一个平衡点。记线性化为

$$A=Df(0).$$

在原点处作 Taylor 展开：

$$x'=Ax+f^{(2)}(x)+f^{(3)}(x)+\cdots,$$

其中 $f^{(k)}\in\mathcal H_k$。这里 $\mathcal H_k$ 是所有次数为 $k$ 的 $d$ 分量齐次多项式向量场构成的空间。若 $h\in\mathcal H_k$，则

$$h(\alpha x)=\alpha^k h(x),\qquad h(x)=O(|x|^k),\qquad Dh(x)=O(|x|^{k-1}).$$

其维数为

$$\dim\mathcal H_k=d\binom{d+k-1}{k}.$$


## 2. 第 $k$ 阶的正规化步骤

固定 $k\geq2$，作近恒等多项式坐标变换

$$x=y+h^{(k)}(y),\qquad h^{(k)}\in\mathcal H_k.$$

以下简记为 $h=h^{(k)}$。其逆变换在所需精度下为

$$y=x-h(x)+O(|x|^{k+1}).$$

原因是下一项的阶数为

$$Dh(y)h(y)=O(|y|^{k-1})O(|y|^k)=O(|y|^{2k-1}),$$

而当 $k\geq2$ 时，$2k-1\geq k+1$。对时间求导可得

$$y'=x'-Dh(x)x'+O(|x|^{k+1}).$$

代入原方程并仅保留不超过 $k$ 阶的项：

$$y'=Ax+\sum_{j=2}^{k}f^{(j)}(x)-Dh(x)Ax+O(|x|^{k+1}).$$

因为当 $j\geq2$ 时，$Dh(x)f^{(j)}(x)$ 的次数为 $k+j-1\geq k+1$，所以它被纳入余项。最后代入 $x=y+h(y)$，得到

$$\begin{aligned}
y'={}&Ay+\sum_{j=2}^{k-1}f^{(j)}(y)\\
&+\left[f^{(k)}(y)-\bigl(Dh^{(k)}(y)Ay-Ah^{(k)}(y)\bigr)\right]+O(|y|^{k+1}).
\end{aligned}$$


## 3. 同调算子、同调方程与共振

定义同调算子

$$\boxed{\mathcal L_Ah(y)=Dh(y)Ay-Ah(y).}$$

因此，第 $k$ 阶项在变换后满足

$$f^{(k)}\longmapsto f^{(k)}-\mathcal L_Ah^{(k)}.$$

特别地，次数低于 $k$ 的项保持不变。若要消去第 $k$ 阶项，需要解 **同调方程**

$$\boxed{\mathcal L_Ah^{(k)}=f^{(k)}.}$$

选择一个补空间 $\mathcal R_k$，使得

$$\mathcal H_k=\operatorname{range}(\mathcal L_A)\oplus\mathcal R_k.$$

将 $f^{(k)}$ 写为

$$f^{(k)}=g^{(k)}+r^{(k)},$$

其中 $g^{(k)}\in\operatorname{range}(\mathcal L_A)$，而 $r^{(k)}\in\mathcal R_k$。解 $\mathcal L_Ah^{(k)}=g^{(k)}$ 后，剩余项就是 $r^{(k)}$。

- $\operatorname{range}(\mathcal L_A)$ 中的项可以消去。
- 补空间 $\mathcal R_k$ 中的项称为共振项。
- 补空间的选择不唯一，因此正规形通常也不唯一。

同调算子也可用 Lie 括号写成

$$\mathcal L_Ah=[h,Ax]=(Dh)Ax-(D(Ax))h.$$


## 4. Poincare 正规形与线性代数观点

对任意固定截断阶数 $m$，存在一个多项式近恒等变换

$$x=y+h^{(2)}(y)+\cdots+h^{(m)}(y)+O(|y|^{m+1}),$$

使原系统变为

$$y'=Ay+r^{(2)}(y)+\cdots+r^{(m)}(y)+O(|y|^{m+1}),\qquad r^{(j)}\in\mathcal R_j.$$

这就是 **Poincare 正规形定理**。其证明思想是递归地先正规化二次项，再正规化三次项，依此类推。

在 $\mathcal H_k$ 上给定内积后，可自然选择

$$\boxed{\mathcal R_k=\ker(\mathcal L_A^*)=(\operatorname{range}\mathcal L_A)^\perp.}$$

换言之，方程 $\mathcal L_Ah=b$ 有解，当且仅当 $b$ 与 $\ker(\mathcal L_A^*)$ 中的所有向量都正交。若 $\mathcal L_A$ 的零特征值是半单的，则

$$\mathcal H_k=\ker(\mathcal L_A)\oplus\operatorname{range}(\mathcal L_A),$$

此时也可选择 $\mathcal R_k=\ker(\mathcal L_A)$。

若 $A=\operatorname{diag}(\lambda_1,\ldots,\lambda_d)$，且 $h(y)=y^\alpha e_i$，则

$$\mathcal L_A(y^\alpha e_i)=\left(\sum_{j=1}^d\alpha_j\lambda_j-\lambda_i\right)y^\alpha e_i.$$

所以共振条件为

$$\boxed{\sum_{j=1}^d\alpha_j\lambda_j=\lambda_i.}$$


## 5. Hopf 正规形

在平面 Hopf 分岔点，线性化的特征值为 $\pm i\omega$，其中 $\omega>0$。经过线性状态变换以及时间缩放 $\tau=\omega t$ 后，线性部分可写为

$$J=\begin{pmatrix}0&-1\\1&0\end{pmatrix}.$$

三阶 Poincare 正规形为

$$\boxed{y'=Jy+(y_1^2+y_2^2)\begin{pmatrix}p_1&-p_2\\p_2&p_1\end{pmatrix}y+O(|y|^4).}$$

### 二次项

取基

$$(x^2\partial_x,xy\partial_x,y^2\partial_x,x^2\partial_y,xy\partial_y,y^2\partial_y).$$

为了和原讲义中的矩阵一致，下面临时使用 $\widehat{\mathcal L}_J=-\mathcal L_J$。其矩阵为

$$M^{(2)}=\begin{pmatrix}
0&-1&0&-1&0&0\\
2&0&-2&0&-1&0\\
0&1&0&0&0&-1\\
1&0&0&0&-1&0\\
0&1&0&2&0&-2\\
0&0&1&0&1&0
\end{pmatrix}.$$

因为

$$\det M^{(2)}=9\neq0,$$

所有二次项都可被消去。


### 三次项

在三次齐次多项式空间中，同调矩阵的秩为 $6$，而 $\dim\mathcal H_3=8$，因此共振空间为二维。可选择 $\ker M^{(3)}$ 的一组基

$$v=(1,0,1,0,0,1,0,1)^\top,$$

$$w=(0,-1,0,-1,1,0,1,0)^\top.$$

它们对应于

$$v\longleftrightarrow (x^2+y^2)\begin{pmatrix}x\\y\end{pmatrix},\qquad w\longleftrightarrow (x^2+y^2)\begin{pmatrix}-y\\x\end{pmatrix}. $$

因此保留下来的三次项正是

$$p_1(x^2+y^2)\begin{pmatrix}x\\y\end{pmatrix}+p_2(x^2+y^2)\begin{pmatrix}-y\\x\end{pmatrix}. $$

令 $z=x+iy$，则可写为

$$z'=iz+(p_1+ip_2)z|z|^2+O(|z|^4).$$

在 $z$ 方程中，单项式 $z^p\bar z^q$ 的共振条件为 $p-q=1$。二次阶不存在满足条件的单项式；三次阶中唯一的共振形式是

$$z^2\bar z=z|z|^2.$$

在极坐标下，

$$r'=p_1r^3+O(r^4),\qquad \theta'=1+p_2r^2+O(r^3).$$

所以 $p_1$ 控制主导的径向增长或衰减，$p_2$ 控制角速度的主导非线性修正。


## 6. 含参数系统与 Bogdanov-Takens 正规形

含参数系统

$$x'=f(x,p)$$

可通过加入

$$p'=0$$

而视为自治系统。

对具有双零特征值的通用平面双参数系统，Bogdanov-Takens 正规形为

$$\boxed{\begin{pmatrix}y_1'\\y_2'\end{pmatrix}=\begin{pmatrix}0\\q_1\end{pmatrix}+\begin{pmatrix}0&1\\q_2&0\end{pmatrix}\begin{pmatrix}y_1\\y_2\end{pmatrix}+\begin{pmatrix}0\\y_1^2\pm y_1y_2\end{pmatrix}.}$$

即

$$y_1'=y_2,\qquad y_2'=q_1+q_2y_1+y_1^2\pm y_1y_2.$$

其线性部分是幂零矩阵

$$A=\begin{pmatrix}0&1\\0&0\end{pmatrix}. $$

进一步还需解决两个问题：何时可以忽略高阶项，以及如何先把高维系统降维。后一个问题引向中心流形理论。


## 7. 实用流程与练习

对每一个次数 $k=2,3,\ldots,m$：

1. 计算 $A=Df(0)$ 和齐次 Taylor 项。
2. 为 $\mathcal H_k$ 选择单项式基。
3. 构造 $\mathcal L_A$ 的矩阵。
4. 求其像空间、核与伴随核。
5. 将 $f^{(k)}$ 分解为可消去部分和共振部分。
6. 对可消去部分解同调方程。
7. 应用坐标变换，并继续下一阶。

### 练习 E9.1

证明

$$x'=\begin{pmatrix}0&1\\0&0\end{pmatrix}x+O(|x|^2)$$

可局部变换为

$$y'=\begin{pmatrix}0&1\\0&0\end{pmatrix}y+\begin{pmatrix}0\\p_1y_1^2+p_2y_1y_2\end{pmatrix}+O(|y|^3).$$

### 练习 E9.2

对离散映射，若

$$x^+=Ax+f^{(k)}(x)+O(|x|^{k+1}),$$

则可取映射的同调算子为

$$\mathcal L_A^{\mathrm{map}}h=h\circ A-Ah,$$

从而

$$\widetilde f^{(k)}=f^{(k)}-\mathcal L_A^{\mathrm{map}}h.$$

将这一方法应用于

$$g\begin{pmatrix}x\\y\end{pmatrix}=\begin{pmatrix}x+(x-y)^2\\3y+x^2+2y^2\end{pmatrix}. $$

## 核心结论

$$\boxed{\widetilde f^{(k)}=f^{(k)}-\mathcal L_Ah^{(k)}}$$

它说明：正规形方法通过逐阶消去非共振项，把系统化简为只保留共振项的形式。
