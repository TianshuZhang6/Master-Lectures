## 1. Main idea and learning goals

A parameter-dependent ordinary differential equation has the form

$$
x' = f(x,p), \qquad x \in \mathbb{R}^d,\quad p \in \mathbb{R}^m.
$$

A **local bifurcation** occurs when a small change of the parameter produces a phase portrait that is not topologically equivalent to the previous one. The typical mechanism in this lecture is **loss of hyperbolicity**: an eigenvalue of the linearization reaches the imaginary axis for a flow, or a multiplier reaches the unit circle for a map.

The section develops four recurring ideas:

1. Detect where hyperbolicity is lost.
2. Impose non-degeneracy and transversality conditions so that the event is generic.
3. simplify the system by coordinate and parameter changes.
4. Compare the resulting system with a low-order **normal form** whose dynamics are easy to analyze.

For flows, the main codimension-one cases are the fold and Hopf bifurcations. For maps, the section treats fold, flip (period-doubling), and Neimark-Sacker bifurcations. Pitchfork and transcritical bifurcations appear as important symmetry-driven or non-generic alternatives.

---

### Clarification: when must a bifurcation have a zero eigenvalue?

The general requirement for a **local bifurcation of an equilibrium** is loss of hyperbolicity, not necessarily the presence of an eigenvalue equal to zero.

For a continuous-time system

$$
\dot{x}=f(x,p), \qquad A(p)=D_xf(x^*(p),p),
$$

an equilibrium is hyperbolic when every eigenvalue satisfies $\operatorname{Re}\lambda_j\neq0$. A necessary spectral condition for a local equilibrium bifurcation is therefore

$$
\operatorname{Re}\lambda_j=0
$$

for at least one critical eigenvalue. In a one-dimensional ODE the eigenvalue is real, so loss of hyperbolicity indeed means $\lambda=0$. This is the critical condition for fold, transcritical, and pitchfork bifurcations. In two or more dimensions, however, a Hopf bifurcation has

$$
\lambda_{1,2}=\pm i\omega, \qquad \omega>0,
$$

which are nonzero eigenvalues with zero real part.

For a discrete map $x_{k+1}=g(x_k,p)$, hyperbolicity is lost when a multiplier reaches the unit circle:

$$
|\mu|=1.
$$

The principal cases are $\mu=1$ for a fold, $\mu=-1$ for a flip, and $\mu=e^{\pm i\theta}$ for a Neimark-Sacker bifurcation. A multiplier $\mu=0$ is normally strongly contracting, not critical.

Finally, this criterion concerns local bifurcations attached to an equilibrium or fixed point. Global events such as homoclinic or heteroclinic bifurcations can occur while the equilibria involved remain hyperbolic, because the change is in the global arrangement of invariant manifolds rather than in a local spectrum.

## 2. Structural language: equivalence, strata, diagrams, unfoldings, and codimension

Consider two parameterized ODEs

$$
x'=f(x,p), \qquad y'=g(y,q).
$$

They are **fiber topologically equivalent** if there is a homeomorphism of parameter spaces, $q=\xi(p)$, and, for every fixed corresponding parameter pair, the two phase portraits are topologically equivalent. Thus, equivalence is checked fiber by fiber while allowing a relabeling of the parameter.

A **stratum** $S_{p_0}$ is the maximally connected parameter region containing $p_0$ on which all phase portraits are topologically equivalent. A **bifurcation diagram** partitions parameter space into these strata and attaches a representative phase portrait to each stratum.

An **unfolding** is a parameterized family that contains a bifurcation persistently**which mean then family of the parameters make the bifurcations happen**. Its **codimension** is the minimum number of parameters needed for such an unfolding. A codimension-one bifurcation is therefore encountered generically by varying one parameter.

In the lecture's bifurcation diagrams, solid equilibrium branches are stable and dashed branches are unstable.

## 3. Pitchfork bifurcation

The introductory example is

$$
x'=px-x^3=x(p-x^2).
$$

Equilibria satisfy $x(p-x^2)=0$. Therefore:

- For $p\leq 0$, the only equilibrium is $x^*=0$. It is globally asymptotically stable.
- For $p>0$, the origin remains an equilibrium but becomes unstable, while two stable equilibria appear:

$$
x_{\pm}=\pm\sqrt{p}.
$$

This is a **supercritical pitchfork bifurcation**. It is strongly associated with the reflection symmetry $x\mapsto -x$: if $x(t)$ is transformed to $-x(t)$, the equation keeps the same form. The symmetry forces the two nonzero equilibrium branches to appear as a pair.

The key qualitative event is not simply the appearance of equilibria. It is the change from one stable symmetric state to two stable symmetry-broken states, with the original state losing stability at $p=0$.

## 4. Fold (saddle-node) bifurcation: basic example

The standard example is

$$
x'=p-x^2.
$$

For $p>0$, the equilibria are

$$
x_{\pm}=\pm\sqrt{p}.
$$

The linearization at an equilibrium is

$$
X' = \partial_x f(x_{\pm},p)X=-2x_{\pm}X.
$$

Hence $x_+$ is stable and $x_-$ is unstable. At $p=0$, the two equilibria collide at the origin, the eigenvalue becomes zero, and hyperbolicity is lost. For $p<0$, no equilibrium remains.

The equation $x'=p+x^2$ has the same local bifurcation type with the orientation reversed. Both are called a **fold** or **saddle-node bifurcation**. The family is a codimension-one unfolding because sufficiently small smooth perturbations still display the same local creation or destruction of a stable-unstable equilibrium pair.

**remark: the pitchfork and the fold are different bifurcation, the pitchfork always has the equibirium point x=0, while in fold the equibirium points $x=+-\sqrt{q}$ will disappear and appear by changing of p**

## 5. Why the lowest-order terms determine the fold

Lemma 8.7 states that, near $(x,p)=(0,0)$,

$$
x'=p\pm x^2+O(x^3)
$$

is locally topologically equivalent to

$$
x'=p\pm x^2.
$$

The notation $a(x)=O(b(x))$ as $x\to0$ means that $|a(x)|/|b(x)|$ stays bounded. The remainder may also contain suitable higher-order parameter-dependent terms, such as $px^3$.

For the negative-sign case, write

$$
y'=g(y,p)=p-y^2+\psi(y,p), \qquad \psi=O(y^3).
$$

Because $\partial_p g(0,0)=1$, the Implicit Function Theorem gives a nearby equilibrium curve

$$
\mathcal{M}=\{(y,p):p=y^2+G(y)\}, \qquad G(y)=O(y^3).
$$

For $p>0$, its two equilibria remain close to $\pm\sqrt{p}$. A parameter-dependent homeomorphism can map the two equilibria of the normal form to those of the perturbed equation while preserving their order and local flow direction. Therefore the higher-order terms bend the diagram but do not change its local topological type.

$$

$$

Then, near the origin, the system is topologically equivalent to one of the two normal forms

$$

Interpretation of the conditions:

- $f(0,0)=0$: the critical point is an equilibrium.
- $\partial_x f(0,0)=0$: the linear eigenvalue is zero, so hyperbolicity is lost.
- $\partial_{xx}f(0,0)\neq0$: the first important nonlinear term is genuinely quadratic; this is the non-degeneracy condition.
- $\partial_p f(0,0)\neq0$: the parameter crosses the bifurcation set with nonzero speed; this is the transversality condition.

### Detailed derivation and the purpose of each transformation

The goal is to show that every system satisfying the fold conditions can be simplified, by locally invertible changes of state and parameter, to

$$
y'=q\pm y^2.
$$

Each transformation removes one inessential feature of the Taylor expansion while preserving the local phase portrait.

#### Step 0 - Taylor expansion near the nonhyperbolic equilibrium

Write

$$
f(x,p)=f_0(p)+f_1(p)x+f_2(p)x^2+O(x^3).
$$

The fold assumptions imply

$$
f_0(0)=0, \qquad f_1(0)=0, \qquad f_2(0)\neq0, \qquad f_0'(0)\neq0.
$$

Here $f_1(0)=0$ expresses the loss of hyperbolicity, $f_2(0)\neq0$ ensures that the quadratic term is the first effective state nonlinearity, and $f_0'(0)\neq0$ ensures that the parameter moves the vector field through the critical configuration.

#### Step 1 - translate the state variable and remove the linear term

Introduce the parameter-dependent translation

$$
y=x+\delta(p).
$$

Equivalently,

$$
\boxed{x=y-\delta(p).}
$$

The parameter is fixed along each trajectory, so $p'=0$ and $\delta(p)$ has no time dependence. Therefore

$$
y'=x'.
$$

Substitute $x=y-\delta$ into the Taylor expansion:

$$
\begin{aligned}
y'=x'
&=f_0(p)+f_1(p)(y-\delta)+f_2(p)(y-\delta)^2
+O\bigl((y-\delta)^3\bigr)\\
&=C(p,\delta)+L(p,\delta)y+Q(p,\delta)y^2+O(y^3).
\end{aligned}
$$

The three coefficients have the local forms

$$
\begin{aligned}
C(p,\delta)
&=f_0(p)-f_1(p)\delta+f_2(p)\delta^2+O(\delta^3),\\
L(p,\delta)
&=f_1(p)-2f_2(p)\delta+\delta^2\psi(p,\delta),\\
Q(p,\delta)
&=f_2(p)+O(\delta),
\end{aligned}
$$

where $\psi$ is smooth. To remove the linear term in $y$, choose $\delta$ so that

$$
F(p,\delta):=L(p,\delta)=0.
$$

At the bifurcation point,

$$
F(0,0)=0,
$$

and

$$
\partial_\delta F(0,0)=-2f_2(0)\neq0.
$$

The Implicit Function Theorem therefore gives a unique smooth function $\delta=\delta(p)$ with $\delta(0)=0$ and $F(p,\delta(p))=0$.

**Purpose of this step.** The translation centers the coordinate at the parameter-dependent point where the state derivative of the vector field vanishes. It removes the linear $y$ term and exposes the quadratic fold geometry. It does not mean that every nearby equilibrium is moved to $y=0$; for $q>0$ the normal form still has two equilibria away from the origin.

Because $\delta(p)=O(p)$, substituting the selected $\delta(p)$ into $C$ and $Q$ gives

$$
C(p,\delta(p))=f_0'(0)p+O(p^2)
$$

and

$$
Q(p,\delta(p))=f_2(0)+O(p).
$$

Hence the transformed equation is

$$
y'=\bigl[f_0'(0)p+O(p^2)\bigr]
+\bigl[f_2(0)+O(p)\bigr]y^2+O(y^3).
$$

#### Step 2 - use the constant term as the new parameter

Define

$$
\alpha(p):=C(p,\delta(p))=f_0'(0)p+p^2\mu(p),
$$

and set

$$
\boxed{q=\alpha(p).}
$$

Since

$$
\alpha(0)=0, \qquad \alpha'(0)=f_0'(0)=\partial_pf(0,0)\neq0,
$$

the Inverse Function Theorem gives a smooth local inverse $p=p(q)$.

**Purpose of this step.** The original parameter may enter the constant term through a complicated smooth function. Replacing $p$ by $q=\alpha(p)$ straightens this dependence so that the constant term becomes exactly $q$. Because the change is locally invertible, it only relabels the nearby family of systems and does not merge the two sides of the bifurcation.

After substituting $p=p(q)$, define

$$
b(q):=Q(p(q),\delta(p(q))).
$$

Then

$$
y'=q+b(q)y^2+O(y^3), \qquad b(0)=f_2(0)\neq0.
$$

#### Step 3 - normalize the magnitude of the quadratic coefficient

Because $b(0)\neq0$, continuity implies that $b(q)$ has a constant sign for sufficiently small $q$. Let

$$
B(q)=|b(q)|>0
$$

and introduce

$$
\widetilde y=B(q)y, \qquad \widetilde q=B(q)q.
$$

For a fixed parameter $q$, $B(q)$ is constant in time, so

$$
\widetilde y'=B(q)y'.
$$

Using $y=\widetilde y/B(q)$ gives

$$
\begin{aligned}
\widetilde y'
&=B(q)q+\frac{b(q)}{B(q)}\widetilde y^2+O(\widetilde y^3)\\
&=\widetilde q+s\widetilde y^2+O(\widetilde y^3),
\end{aligned}
$$

where

$$
s=\frac{b(q)}{|b(q)|}=\operatorname{sign}(b(0))\in\{-1,+1\}.
$$

**Purpose of this step.** The scaling removes the numerical magnitude of the quadratic coefficient. Only its sign remains, showing that there are two oriented versions of the fold normal form. The state scaling is invertible because $B(q)>0$, and the parameter scaling is locally invertible because its derivative at zero is $B(0)>0$.

Dropping the tildes gives

$$
y'=q+sy^2+O(y^3), \qquad s=\pm1.
$$

#### Step 4 - remove topologically irrelevant higher-order terms

Lemma 8.7 states that

$$
y'=q\pm y^2+O(y^3)
$$

is locally topologically equivalent to

$$
\boxed{y'=q\pm y^2.}
$$

This final step does not claim that the higher-order terms are algebraically zero. It says that they only bend the local equilibrium branches and can be removed by a local homeomorphism without changing the number, stability, or arrangement of the nearby equilibria.

The complete logic is therefore

$$
\text{general Taylor expansion}
\longrightarrow
\text{remove the linear state term}
\longrightarrow
\text{straighten the parameter}
\longrightarrow
\text{normalize the quadratic term}
\longrightarrow
\text{discard topologically inessential higher-order terms}.
$$

## 8. Transcritical bifurcation and exchange of stability

If the fold transversality condition is dropped, another bifurcation can occur. The lecture uses

$$
x'=x(p-x).
$$

The equilibrium branches are

$$
x^*_1=0, \qquad x^*_2=p.
$$

At $x=0$, the derivative is $\partial_x f(0,p)=p$, so the origin is stable for $p<0$ and unstable for $p>0$. At $x=p$,

$$
\partial_x f(p,p)=p-2p=-p,
$$

so this branch is unstable for $p<0$ and stable for $p>0$. The branches cross at the origin and exchange stability. This is a **transcritical bifurcation**.

The example shows why genericity conditions matter: loss of hyperbolicity alone does not determine the bifurcation type. The structure of the nonlinear terms and the way the parameter enters the equation decide whether equilibria collide and disappear, cross, or split symmetrically.

## 9. Hopf bifurcation theorem

In a generic two-dimensional, one-parameter flow, a complex-conjugate eigenvalue pair may cross the imaginary axis. Suppose

$$
x'=f(x,p), \qquad x\in\mathbb{R}^2,\quad p\in\mathbb{R},
$$

with $f(0,0)=0$ and

$$
\lambda_{1,2}(p)=\mu(p)\pm i\omega(p), \qquad \mu(0)=0,\quad \omega(0)>0.
$$

Under genericity assumptions, the system is locally topologically equivalent to

$$
\begin{pmatrix}y_1'\\y_2'\end{pmatrix}
=
\begin{pmatrix}q&-1\\1&q\end{pmatrix}
\begin{pmatrix}y_1\\y_2\end{pmatrix}
+l_1(y_1^2+y_2^2)\begin{pmatrix}y_1\\y_2\end{pmatrix},
$$

where the cubic coefficient $l_1\neq0$ is the **first Lyapunov coefficient**. The lecture writes the two topological normal forms using the signs $l_1=\pm1$.

The transversality condition is $\mu'(0)\neq0$: the real part of the eigenvalue pair crosses the imaginary axis at nonzero speed. The non-degeneracy condition $l_1\neq0$ ensures that cubic terms control the creation or destruction of the nearby periodic orbit. The Hopf bifurcation is codimension one.

## 10. Hopf dynamics in complex and polar form

Set $z=y_1+iy_2$. The Hopf normal form becomes

$$
z'=(q+i)z+l_1z|z|^2.
$$

Using $z=re^{i\varphi}$ gives the polar equations

$$
r'=r(q+l_1r^2), \qquad \varphi'=1.
$$

The angular equation describes uniform rotation. All bifurcation information is in the scalar radial equation. Besides $r=0$, a periodic orbit exists when

$$
r^2=-\frac{q}{l_1}>0.
$$

Therefore:

- **Supercritical Hopf ($l_1<0$):** the equilibrium is stable for $q<0$ and unstable for $q>0$. A small **stable** periodic orbit exists for $q>0$, with radius $r=\sqrt{-q/l_1}$.
- **Subcritical Hopf ($l_1>0$):** the equilibrium is stable for $q<0$ and unstable for $q>0$. A small **unstable** periodic orbit exists for $q<0$.

For a general vector field, $l_1$ can be computed from derivatives through third order. Its sign determines the criticality and stability of the bifurcating periodic orbit.

---

### Detailed derivation of the Hopf polar form

Start from the complex Hopf normal form

$$
z'=(q+i)z+l_1z|z|^2.
$$

Write

$$
z=re^{i\varphi}, \qquad r\geq0.
$$

Because $|z|=r$, the right-hand side becomes

$$
(q+i)re^{i\varphi}+l_1re^{i\varphi}r^2
=re^{i\varphi}(q+i+l_1r^2).
$$

The variables $r=r(t)$ and $\varphi=\varphi(t)$ both depend on time. Differentiating the polar representation gives

$$
\begin{aligned}
z'&=(re^{i\varphi})'\\
&=r'e^{i\varphi}+ir\varphi'e^{i\varphi}\\
&=e^{i\varphi}(r'+ir\varphi').
\end{aligned}
$$

Equating the two expressions for $z'$ and cancelling the nonzero factor $e^{i\varphi}$ yields

$$
r'+ir\varphi'=r(q+l_1r^2)+ir.
$$

Equality of complex numbers means equality of their real and imaginary parts. Hence

$$
r'=r(q+l_1r^2), \qquad r\varphi'=r.
$$

For $r>0$, division by $r$ gives $\varphi'=1$. At $r=0$ the angle is undefined, but the radial equation remains valid. Thus

$$
\boxed{r'=r(q+l_1r^2), \qquad \varphi'=1.}
$$

The same result follows directly from real variables. With $r^2=y_1^2+y_2^2$,

$$
rr'=y_1y_1'+y_2y_2',
$$

and the rotational cross terms cancel. The angular velocity can be computed from

$$
\varphi'=\frac{y_1y_2'-y_2y_1'}{y_1^2+y_2^2}=1.
$$

This decomposition separates the dynamics into radial growth or decay and a uniform counterclockwise rotation.

---

### Interpreting the Hopf radial equation

The complete local behavior is encoded by

$$
r'=F(r)=r(q+l_1r^2), \qquad r\geq0.
$$

The radial equilibrium $r=0$ represents the equilibrium $(y_1,y_2)=(0,0)$. A positive radial equilibrium satisfies

$$
r_*^2=-\frac{q}{l_1}, \qquad r_*=\sqrt{-\frac{q}{l_1}}.
$$

When $r(t)\equiv r_*$, the angular equation gives $\varphi(t)=t+\varphi_0$. Therefore

$$
y_1(t)=r_*\cos(t+\varphi_0), \qquad y_2(t)=r_*\sin(t+\varphi_0),
$$

which is a periodic orbit of period $2\pi$. Thus a positive equilibrium of the radial equation corresponds to a periodic orbit of the planar system.

Near the origin, $r'\approx qr$. Hence the origin is attracting for $q<0$ and repelling for $q>0$. The sign of $l_1$ determines where a nonzero orbit exists and whether it is stable.

**Supercritical case, $l_1<0$.** Write $l_1=-a$ with $a>0$. For $q<0$, all sufficiently small positive radii decrease to zero. At $q=0$, $r'=-ar^3$, so the nonhyperbolic origin is still asymptotically stable but convergence is slower than exponential. For $q>0$, the radius

$$
r_*=\sqrt{q/a}
$$

is attracting: radii below it grow, while nearby radii above it shrink. The equilibrium becomes unstable and a stable periodic orbit is born.

**Subcritical case, $l_1>0$.** For $q<0$, the origin is stable but is surrounded locally by the repelling periodic orbit $r_*=\sqrt{-q/l_1}$. At $q=0$, $r'=l_1r^3>0$ for $r>0$, so the nonhyperbolic origin is unstable. For $q>0$, all sufficiently small nonzero radii grow and no small periodic orbit exists.

The stability of the nonzero orbit can also be checked from

$$
F'(r)=q+3l_1r^2, \qquad F'(r_*)=-2q.
$$

Thus the orbit is stable when $q>0$ and unstable when $q<0$, combined with the existence condition $-q/l_1>0$.

## 11. From flows to maps: the unit-circle criterion

For a parameterized iterated map

$$
x_{k+1}=g(x_k,p),
$$

let $x^*$ be a fixed point and let

$$
A(p)=Dg(x^*,p)
$$

be the linearized map. Its eigenvalues are called **multipliers**. A fixed point of a map is hyperbolic when no multiplier lies on the unit circle. Consequently, loss of hyperbolicity occurs when

$$
|\mu|=1\quad\text{or}\quad|\mu_{1,2}|=1.
$$

This replaces the imaginary-axis condition for flows. The main codimension-one possibilities in low dimensions are:

- a real multiplier crosses $+1$: fold bifurcation of maps;
- a real multiplier crosses $-1$: flip or period-doubling bifurcation;
- a nonreal conjugate pair crosses the unit circle: Neimark-Sacker bifurcation.

## 12. Fold bifurcation of maps

For a generic one-dimensional map with $g(0,0)=0$ and critical multiplier $\mu=1$, Theorem 8.12 gives the normal form

$$
y\mapsto q+y\pm y^2.
$$

The genericity conditions are

$$
\partial_{xx}g(0,0)\neq0, \qquad \partial_p g(0,0)\neq0.
$$

Fixed points satisfy

$$
y=q+y\pm y^2,
$$

so the fixed-point equation reduces to the same quadratic balance that appears in the flow fold. Thus a stable-unstable fixed-point pair is created or destroyed. The proof parallels the fold theorem for flows: translate the fixed-point branch, reparameterize, scale the quadratic coefficient, and discard topologically irrelevant higher-order terms.


**careful in the map the fixed point should be $x=g(x,p)$ which induce $x-g(x,p)=0$**

## 13. Flip (period-doubling) bifurcation theorem

A flip bifurcation occurs when a real multiplier crosses $-1$. For a generic one-dimensional map with $g(0,0)=0$ and $\mu=-1$, Theorem 8.13 gives the normal form

$$
y\mapsto -(1+q)y\pm y^3.
$$

The transversality condition is

$$
\partial_{xp}g(0,0)\neq0,
$$

which ensures that the multiplier crosses $-1$ with nonzero parameter speed. A nonzero effective cubic coefficient supplies the non-degeneracy condition.

**Formula note.** The theorem statement in the excerpt prints

$$
\frac12(\partial_{xx}g(0,0))^2+\frac16\partial_{xxx}g(0,0)\neq0,
$$

while its subsequent derivation obtains

$$
c(0)=\left(\frac12\partial_{xx}g(0,0)\right)^2+\frac16\partial_{xxx}g(0,0)\neq0.
$$

These expressions differ by a factor in the squared term. The proof's expression follows directly from $a(0)=\tfrac12\partial_{xx}g(0,0)$ and $c(0)=a(0)^2+b(0)$, so the discrepancy should be checked against the full source or errata.

---

### Why write the flip multiplier as $-(1+\kappa(p))$?

After translating the fixed-point branch to the origin, write the map as

$$
g(x,p)=g_1(p)x+g_2(p)x^2+g_3(p)x^3+O(x^4).
$$

Since the multiplier at a flip point is $-1$,

$$
g_1(0)=\partial_xg(0,0)=-1.
$$

It is completely valid to write

$$
g_1(p)=-k(p), \qquad k(0)=1.
$$

The lecture instead writes

$$
g_1(p)=-(1+\kappa(p)), \qquad \kappa(0)=0,
$$

where $\kappa(p)=k(p)-1$. The two notations are mathematically equivalent, but the second is better adapted to local bifurcation analysis:

- $\kappa$ directly measures displacement of the multiplier from the critical value $-1$.
- The critical parameter is represented by the convenient zero value $\kappa=0$.
- Since $\kappa(0)=0$, it is a small quantity with expansion $\kappa(p)=\kappa'(0)p+O(p^2)$.
- If $\kappa'(0)\neq0$, one may immediately use $q=\kappa(p)$ as the new parameter, giving the normal-form multiplier $\mu(q)=-(1+q)$.

For sufficiently small $q$,

$$
q<0\Rightarrow |\mu(q)|<1, \qquad q=0\Rightarrow\mu=-1, \qquad q>0\Rightarrow|\mu(q)|>1.
$$

Thus the sign of $q$ directly distinguishes the stable and unstable sides.

There is also a sign detail worth recording. Since $g_1(p)=-(1+\kappa(p))$,

$$
g_1'(0)=\partial_{xp}g(0,0)=-\kappa'(0).
$$

Therefore the exact identity contains a minus sign, although the sign is irrelevant to the nonzero transversality condition.

## 14. Flip normal-form reduction and period-two orbit

Because $\partial_x g(0,0)=-1\neq1$, the Implicit Function Theorem first gives a unique nearby fixed-point branch. A preliminary coordinate change makes this branch $x=0$. Taylor expansion gives

$$
g(x,p)=g_1(p)x+g_2(p)x^2+g_3(p)x^3+O(x^4).
$$

Transversality permits a new parameter $q$ such that the linear coefficient is $\mu(q)=-(1+q)$. A near-identity coordinate change

$$
x=y+\delta(q)y^2
$$

removes the quadratic term because $\mu(0)^2-\mu(0)=2\neq0$. Scaling the remaining nonzero cubic coefficient leads to the normal form.

For the positive cubic sign, let

$$
f(y,q)=-(1+q)y+y^3.
$$

The fixed point $y^*=0$ has multiplier $-1-q$: it is stable for $q<0$ and unstable for $q>0$. Fixed points of the second iterate $f^2$ reveal a new period-two orbit. For small $q>0$, its two points are

$$
y_{\pm}=\pm\sqrt{q}\quad\text{(plus higher-order corrections in a general system)}.
$$

For the exact positive-sign normal form, these two points map into one another and form a stable period-two orbit. This is why the event is called period doubling: the original fixed point loses stability as a two-cycle is born.

---

### Parameter and state-space local invertibility

The flip proof uses two distinct locally invertible transformations, and they serve different purposes.

**1. Parameter reparameterization.** From

$$
\kappa(0)=0, \qquad \kappa'(0)\neq0,
$$

the Inverse Function Theorem gives a smooth local inverse between

$$
q=\kappa(p) \qquad\text{and}\qquad p=\kappa^{-1}(q).
$$

This means that $q$ is only a relabeling of the parameter axis near the bifurcation. No nearby parameter values are lost or merged. If $\kappa'(0)>0$, orientation is preserved; if $\kappa'(0)<0$, the parameter direction is reversed, but the bifurcation type is unchanged.

The condition is important. For example, $q=p^2$ merges $p$ and $-p$ and therefore cannot distinguish the two sides of the bifurcation. The map $q=p^3$ is one-to-one, but its inverse $p=q^{1/3}$ is not smooth at zero, so it is not a smooth local parameter coordinate.

**2. State-space coordinate change.** Later, the proof uses

$$
x=h_q(y)=y+\delta(q)y^2.
$$

Here local invertibility follows from $\partial_yh_q(0)=1$. Its purpose is not to relabel parameters but to simplify the Taylor expansion in state space. The transformed map is the local conjugate

$$
\widetilde g_q=h_q^{-1}\circ g_q\circ h_q.
$$

Because $h_q$ is one-to-one locally, fixed points and periodic orbits correspond without information loss. At corresponding fixed points, the linearizations are similar (equal in the one-dimensional case), so multipliers and local stability are preserved.

In short, $q=\kappa(p)$ standardizes the parameter, while $x=h_q(y)$ simplifies the state dynamics.

---

### Why the quadratic term can be removed

After reparameterization, suppose

$$
g_q(x)=\mu(q)x+a(q)x^2+b(q)x^3+O(x^4), \qquad \mu(q)=-(1+q).
$$

Use the near-identity transformation

$$
x=h_q(y)=y+\delta(q)y^2,
$$

whose inverse has the expansion

$$
h_q^{-1}(x)=x-\delta x^2+2\delta^2x^3+O(x^4).
$$

The conjugated map $h_q^{-1}\circ g_q\circ h_q$ has quadratic coefficient

$$
a+\delta\mu-\delta\mu^2.
$$

Setting this equal to zero gives

$$
\delta(q)=\frac{a(q)}{\mu(q)^2-\mu(q)}.
$$

At the critical point $\mu(0)=-1$,

$$
\mu(0)^2-\mu(0)=1-(-1)=2\neq0,
$$

so the quadratic term is nonresonant and can be removed smoothly. The result is

$$
\widetilde g_q(y)=-(1+q)y+c(q)y^3+O(y^4).
$$

The cubic term cannot generally be removed at $\mu=-1$, because its corresponding denominator is $\mu^3-\mu=0$. It is resonant and contains the leading nonlinear information that determines the flip bifurcation. If $c(0)\neq0$, a scaling of $y$ normalizes it to the sign $\pm1$.

The emphasis on conjugacy is essential: the calculation simplifies the formula while retaining the same local dynamics, rather than replacing the original system by an unrelated map.

---

### Why small nonzero $q$ and the scale $y\sim\sqrt q$ matter

A bifurcation study does not examine only the single critical map at $q=0$. It compares a local family for

$$
-\varepsilon<q<\varepsilon
$$

and asks what changes as $q$ crosses zero. Therefore small does not mean equal to zero.

For the positive-cubic flip normal form

$$
f_q(y)=-(1+q)y+y^3,
$$

the small period-two points are found from $f_q^2(y)=y$. At leading order,

$$
0=f_q^2(y)-y\approx2qy-2y^3=2y(q-y^2).
$$

The nonzero balance is therefore

$$
y^2\sim q, \qquad y_\pm\sim\pm\sqrt{q}.
$$

For the exact normal form these points are exactly $\pm\sqrt q$ and satisfy

$$
f_q(\sqrt q)=-\sqrt q, \qquad f_q(-\sqrt q)=\sqrt q.
$$

They are not fixed points of $f_q$; together they form one period-two orbit and are fixed points of $f_q^2$. They belong to the local analysis because $\sqrt q\to0$ as $q\to0^+$. More precisely, for any state-space neighborhood $|y|<\delta$, choosing $0<q<\delta^2$ places both points inside it.

The fixed-point equation of $f_q$ itself is

$$
y=f_q(y)\quad\Longleftrightarrow\quad y\bigl(y^2-(2+q)\bigr)=0.
$$

Thus the full polynomial map also has fixed points $y=\pm\sqrt{2+q}$. They are excluded from the **local** flip analysis because

$$
\pm\sqrt{2+q}\to\pm\sqrt2\neq0
$$

as $q\to0$. They remain a positive distance from the bifurcation point. The precise statement is therefore that $y=0$ is the unique fixed point in a sufficiently small neighborhood of the origin, not on the entire real line.

---

### Interpreting flip diagrams and cobweb plots

For the positive-cubic map

$$
g_q(y)=-(1+q)y+y^3,
$$

the origin has multiplier $g_q'(0)=-(1+q)$. For small $q<0$, $|g_q'(0)|<1$, so iterates approach zero while alternating sides because the multiplier is negative. At $q=0$ the multiplier equals $-1$. For small $q>0$, the origin is unstable and the stable two-cycle $\{\sqrt q,-\sqrt q\}$ appears.

The multiplier of this two-cycle is the derivative of the second iterate:

$$
(g_q^2)'(\sqrt q)=g_q'(\sqrt q)g_q'(-\sqrt q)=(-1+2q)^2.
$$

It has modulus less than one for sufficiently small $q>0$, confirming stability. In a bifurcation diagram, the two curved branches are the two points of the same two-cycle, not two new fixed-point branches. Solid branches conventionally indicate stability and dashed branches instability. Vertical arrows describe attraction or repulsion for fixed $q$; the parameter does not evolve during iteration.

For the negative-cubic map

$$
g_q(y)=-(1+q)y-y^3,
$$

the symmetric two-cycle satisfies $g_q(y)=-y$, hence

$$
y_\pm=\pm\sqrt{-q}, \qquad q<0.
$$

Its multiplier is again $(-1+2q)^2$, which is greater than one for $q<0$, so the cycle is unstable. It locally separates points attracted to the stable origin from points that move outward. When $q>0$, the origin is unstable and no small real two-cycle remains.

A cobweb plot is read by moving vertically from $y_n$ to the graph $y_{n+1}=g_q(y_n)$ and horizontally to the diagonal $y_{n+1}=y_n$. Repeating these two moves shows the actual iteration. A closed rectangle represents a two-cycle; shrinking alternating steps indicate attraction to the origin, while expanding rectangles indicate alternating escape.

Therefore the positive cubic sign gives a supercritical flip with a stable cycle after the transition, while the negative cubic sign gives a subcritical flip with an unstable cycle before the transition.

## 15. Neimark-Sacker bifurcation and the Henon-map example

For a two-dimensional map, a complex-conjugate multiplier pair may cross the unit circle with $|\mu_{1,2}|=1$ and $\mu_1\neq\mu_2$. The resulting **Neimark-Sacker bifurcation** is the discrete-time analogue of a Hopf bifurcation. Instead of a periodic orbit, a closed invariant curve typically bifurcates from the fixed point; the dynamics on that curve may be more complicated than the simple uniform rotation of the Hopf normal form.

The lecture illustrates bifurcation detection with the Henon map

$$
\begin{pmatrix}x_1\\x_2\end{pmatrix}\mapsto\begin{pmatrix}x_2\\p_1-p_2x_1-x_2^2\end{pmatrix}.
$$

A fixed point satisfies $x_1^*=x_2^*$ and

$$
x_2^*=p_1-p_2x_2^*-(x_2^*)^2,
$$

so there are at most two fixed points. Its Jacobian is

$$
A(p)=\begin{pmatrix}0&1\\-p_2&-2x_2^*\end{pmatrix}.
$$

Rather than solving for multipliers explicitly, one can define scalar **bifurcation functions**. Necessary spectral conditions are

$$
\det(A-I)=0 \quad\text{(fold)},
$$

$$
\det(A+I)=0 \quad\text{(flip)},
$$

and, for a complex pair in two dimensions,

$$
\det(A)-1=0 \quad\text{(candidate Neimark-Sacker condition)}.
$$

These are necessary spectral tests. One must still verify that the correct multiplier actually crosses the boundary and that all genericity conditions hold.

---

### How determinants generate bifurcation functions

Let $x^*(p)$ be a fixed point of a smooth map and let

$$
A(p)=D_xg(x^*(p),p).
$$

A multiplier $\lambda$ is an eigenvalue of $A$ precisely when

$$
\det(A(p)-\lambda I)=0.
$$

Substituting a critical multiplier therefore converts a spectral condition into a scalar equation in the parameters.

**Fold of fixed points.** The critical multiplier is $+1$, so define

$$
\beta_{\mathrm{fold}}(p)=\det(A(p)-I).
$$

This condition also has a direct Implicit Function Theorem interpretation. Fixed points solve $g(x,p)-x=0$, whose state derivative is $A-I$. If $A-I$ is invertible, the fixed point continues uniquely and smoothly; a fold can occur only when this invertibility fails.

**Flip.** The critical multiplier is $-1$, giving

$$
\beta_{\mathrm{flip}}(p)=\det(A(p)+I).
$$

**Neimark-Sacker in two dimensions.** If the multipliers are a nonreal conjugate pair $re^{\pm i\theta}$, then

$$
\det A=\lambda_1\lambda_2=r^2.
$$

Their arrival on the unit circle implies $\det A=1$, suggesting

$$
\beta_{\mathrm{NS}}(p)=\det A(p)-1.
$$

However, $\det A=1$ alone does not guarantee a nonreal pair. In two dimensions the additional condition

$$
(\operatorname{tr}A)^2-4\det A<0
$$

is required. When $\det A=1$, this is $|\operatorname{tr}A|<2$.

These scalar functions are useful because a single equation $\beta(p_1,p_2)=0$ typically describes a codimension-one curve in a two-parameter plane. Their zero sets locate candidate bifurcations; they do not by themselves prove that a generic bifurcation occurs.

---

### Scope and limitations of determinant tests

For any finite-dimensional smooth map, the characteristic determinant

$$
\chi_A(\lambda,p)=\det(A(p)-\lambda I)
$$

can be used to impose a chosen critical-multiplier condition. This makes determinant-based bifurcation functions broadly useful, but only as the spectral part of the analysis.

A generic local bifurcation normally requires all of the following:

1. **Spectral condition:** the appropriate multiplier lies on the unit circle.
2. **Simplicity:** the critical multiplier or conjugate pair has the expected multiplicity, and all other multipliers stay away from the unit circle.
3. **Transversality:** the critical multiplier crosses the stability boundary with nonzero speed rather than merely touching it.
4. **Non-degeneracy:** the relevant normal-form coefficient does not vanish.

For a simple real critical multiplier, a nonzero derivative of a suitable bifurcation function often expresses transversality. For a Neimark-Sacker pair $\lambda(p)=r(p)e^{i\theta(p)}$, the crossing condition is $r'(p_*)\neq0$.

In dimensions greater than two, $\det A$ is the product of all multipliers and is usually insufficient for detecting a single conjugate pair on the unit circle. For example, in three dimensions,

$$
\det A=r^2\rho
$$

for a pair $re^{\pm i\theta}$ and a third multiplier $\rho$. Direct eigenvalue analysis, characteristic-polynomial criteria, or numerical continuation is then more appropriate.

The same principle extends to a period-$k$ orbit. Replace $A$ by the monodromy matrix of the iterate

$$
M=Dg^k(x_0)=Dg(x_{k-1})\cdots Dg(x_1)Dg(x_0).
$$

Then $\det(M-I)=0$ is a candidate fold condition for the periodic orbit, and $\det(M+I)=0$ is a candidate period-doubling condition.

Determinant tests are local linear tools. They cannot by themselves detect global bifurcations such as homoclinic connections, heteroclinic connections, crises, or large-scale attractor collisions.

---

### Explicit bifurcation functions for the Henon example

For the lecture's Henon map

$$
g(x_1,x_2)=\bigl(x_2,\ p_1-p_2x_1-x_2^2\bigr),
$$

a fixed point has $x_1^*=x_2^*=s$ and satisfies

$$
s^2+(1+p_2)s-p_1=0.
$$

The Jacobian at this point is

$$
A=\begin{pmatrix}0&1\\-p_2&-2s\end{pmatrix},
$$

with

$$
\operatorname{tr}A=-2s, \qquad \det A=p_2.
$$

The candidate fold function is

$$
\beta_{\mathrm{fold}}=\det(A-I)=2s+1+p_2.
$$

Thus a fold candidate satisfies $2s+1+p_2=0$. The candidate flip function is

$$
\beta_{\mathrm{flip}}=\det(A+I)=1-2s+p_2,
$$

so a flip candidate satisfies $1-2s+p_2=0$.

For a Neimark-Sacker candidate,

$$
\det A=1 \quad\Longleftrightarrow\quad p_2=1.
$$

To ensure that the multipliers are nonreal rather than two real reciprocal multipliers, also require

$$
|\operatorname{tr}A|<2 \quad\Longleftrightarrow\quad |s|<1.
$$

These equations locate candidate curves or segments in parameter space. A complete proof must still check crossing, simplicity, resonance exclusions, and the relevant nonlinear coefficient. The Neimark-Sacker invariant curve is the discrete-time analogue of the periodic orbit born in a Hopf bifurcation, but the dynamics on the invariant curve can include quasiperiodic rotation and resonance effects.

## 16. Comparison of the principal local bifurcations

| System | Bifurcation | Critical spectrum | Normal form | Main local outcome |
|---|---|---|---|---|
| 1D flow | Fold / saddle-node | $\lambda=0$ | $y'=q\pm y^2$ | A stable-unstable equilibrium pair is created or destroyed |
| 1D symmetric flow | Pitchfork | $\lambda=0$ plus symmetry | $y'=qy-y^3$ for the supercritical case | One symmetric branch loses stability and two stable branches appear |
| 1D flow | Transcritical | $\lambda=0$ with special structure | $y'=y(q-y)$ | Two equilibrium branches cross and exchange stability |
| 2D flow | Hopf | $\lambda_{1,2}=\pm i\omega$ | $z'=(q+i)z+l_1z|z|^2$ | A small periodic orbit appears or disappears |
| 1D map | Fold | $\mu=+1$ | $y\mapsto q+y\pm y^2$ | A fixed-point pair is created or destroyed |
| 1D map | Flip | $\mu=-1$ | $y\mapsto-(1+q)y\pm y^3$ | A period-two orbit appears or disappears |
| 2D map | Neimark-Sacker | $|\mu_{1,2}|=1$ | not derived in this excerpt | A closed invariant curve typically bifurcates |

The table highlights a useful analogy: the imaginary axis is the stability boundary for flow eigenvalues, while the unit circle is the stability boundary for map multipliers.

## 17. Practical bifurcation-identification workflow

Given a concrete parameter-dependent system, the lecture suggests the following workflow.

1. **Find equilibria or fixed points.** Solve $f(x,p)=0$ for flows or $g(x,p)=x$ for maps.
2. **Linearize.** Compute $D_xf(x^*,p)$ or $Dg(x^*,p)$.
3. **Locate loss of hyperbolicity.** For flows, look for eigenvalues with zero real part. For maps, look for multipliers on the unit circle.
4. **Classify the critical spectrum.** A simple zero eigenvalue suggests a fold, pitchfork, or transcritical event; a conjugate imaginary pair suggests Hopf; multipliers $+1$, $-1$, or a complex unit pair suggest fold, flip, or Neimark-Sacker.
5. **Check transversality.** Confirm that the eigenvalue or multiplier crosses the stability boundary with nonzero speed as the parameter varies.
6. **Check non-degeneracy.** Confirm that the first relevant nonlinear coefficient does not vanish: quadratic for a fold, cubic/first Lyapunov coefficient for Hopf, and effective cubic coefficient for flip.
7. **Reduce to normal form.** Use translations, reparameterizations, near-identity coordinate changes, and scalings.
8. **Read off the dynamics.** Determine the number and stability of invariant objects from the normal form.

A spectral condition is only the starting point. Without the genericity checks, the system may exhibit a different or higher-codimension bifurcation.

### Final takeaways

- Local bifurcations occur at loss of hyperbolicity, but the critical spectrum alone does not completely classify them.
- Normal forms retain the lowest-order terms that control the topology and stability near the critical point.
- Transversality and non-degeneracy make a bifurcation generic and robust.
- Fold and Hopf bifurcations are the main codimension-one cases for low-dimensional flows.
- Fold, flip, and Neimark-Sacker are their important map counterparts.
- The next theoretical steps, as stated by the lecture, are to systematize normal-form transformations and to reduce higher-dimensional systems to low-dimensional center dynamics.
