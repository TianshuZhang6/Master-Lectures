# Lecture 13: The Smale Horseshoe

## Learning Objectives

After this lecture, you should be able to:

- explain how stretching, contraction, and folding create a horseshoe map;
- construct the maximal invariant set inside the unit square;
- encode horseshoe orbits by bi-infinite binary sequences;
- prove that the horseshoe is topologically conjugate to the full two-symbol shift;
- derive periodic, non-periodic, dense-orbit, and hyperbolicity properties from the conjugacy;
- formulate the generalized horseshoe theorem using curved strips and invariant cones;
- explain the geometric content of the $\lambda$-Lemma;
- understand why a transverse homoclinic point produces horseshoe dynamics.


## 1. Main Idea

The Smale horseshoe is a geometric mechanism that converts a two-dimensional map into symbolic dynamics:

$$
\text{stretching and folding}
\Longrightarrow
\text{an invariant Cantor set}
\Longrightarrow
\text{a shift on symbol sequences}.
$$

Its importance comes from four central facts:

- it contains infinitely many saddle-type periodic orbits;
- it contains uncountably many non-periodic orbits;
- it possesses a dense orbit;
- its invariant dynamics is hyperbolic and therefore robust under small perturbations.

The conceptual chain developed in this lecture is

$$
\text{stretch-compress-fold}
\Longrightarrow
\text{symbolic coding}
\Longrightarrow
\text{shift dynamics}
\Longrightarrow
\text{robust dynamical complexity}.
$$

The final part explains why a transverse homoclinic intersection naturally creates a horseshoe.


## 2. Construction of the Standard Horseshoe

Let

$$
\mathcal Y=[0,1]\times[0,1]
$$

be the unit square. Construct a map $g$ in three geometric steps:

1. Contract $\mathcal Y$ horizontally by a factor

   $$
   0<\lambda<\frac12.
   $$

2. Expand it vertically by a factor

   $$
   \mu>2.
   $$

3. Fold the resulting rectangle and place it across $\mathcal Y$.

The part of the image that returns to the square consists of two disjoint vertical strips:

$$
g(\mathcal Y)\cap\mathcal Y
=
\mathcal V_0\cup\mathcal V_1.
$$

Their preimages form two horizontal strips:

$$
\mathcal Y\cap g^{-1}(\mathcal Y)
=
\mathcal H_0\cup\mathcal H_1,
$$

with

$$
g(\mathcal H_i)=\mathcal V_i,
\qquad i\in\{0,1\}.
$$

The curved middle part lies outside $\mathcal Y$, so points entering it escape from the square. On the two horizontal strips,

$$
Dg=
\begin{pmatrix}
\pm\lambda & 0\\
0 & \pm\mu
\end{pmatrix}.
\tag{13.1}
$$

More explicitly,

$$
Dg|_{\mathcal H_0}
=
\begin{pmatrix}
\lambda&0\\
0&\mu
\end{pmatrix},
\qquad
Dg|_{\mathcal H_1}
=
\begin{pmatrix}
-\lambda&0\\
0&-\mu
\end{pmatrix}.
$$

The minus signs result from the folding. Hyperbolicity depends on the magnitudes $\lambda$ and $\mu$, not on these orientation reversals.

> **Figure clarification.** Figure 32 in the source depicts $g^{-1}$, not $g$. The visible horizontal stretching and vertical compression are therefore the reverse of the forward map described above.

### What $\mathcal Y$, $g^{-1}(\mathcal Y)$, and the strips mean

The square $\mathcal Y$ is the observation region. The preimage

$$
g^{-1}(\mathcal Y)=\{x:g(x)\in\mathcal Y\}
$$

contains all points whose next image lies in $\mathcal Y$. Hence

$$
\mathcal Y\cap g^{-1}(\mathcal Y)
=
\{x\in\mathcal Y:g(x)\in\mathcal Y\}
=
\mathcal H_0\cup\mathcal H_1.
$$

By contrast, $g(\mathcal Y)\cap\mathcal Y=\mathcal V_0\cup\mathcal V_1$ consists of points currently in $\mathcal Y$ that came from $\mathcal Y$. A horizontal strip becomes vertical because its horizontal width is multiplied by $\lambda<1/2$ while its vertical height is multiplied by $\mu>2$. Thus each branch satisfies

$$
g(\mathcal H_i)=\mathcal V_i,
\qquad
g^{-1}(\mathcal V_i)=\mathcal H_i.
$$

There are two horizontal preimage strips because there are two returning vertical branches; each $\mathcal V_i$ has its own branch preimage $\mathcal H_i$.


## 3. The Maximal Invariant Set

Most points eventually leave the square. The horseshoe dynamics is restricted to the set of points that remain in $\mathcal Y$ under all forward and backward iterates:

$$
\mathcal X
=
\left\{
x\in\mathcal Y:
g^k(x)\in\mathcal Y
\text{ for every }k\in\mathbb Z
\right\}.
$$

Equivalently,

$$
\mathcal X
=
\bigcap_{k\in\mathbb Z}g^k(\mathcal Y).
$$

Define the finite forward-survivor sets by

$$
\mathcal Y_n^+
=
\bigcap_{j=0}^{n}g^{-j}(\mathcal Y).
$$

The set $\mathcal Y_n^+$ consists of $2^n$ disjoint horizontal strips. These are the points that remain in $\mathcal Y$ for the next $n$ iterations.

Similarly,

$$
\mathcal Y_n^-
=
\bigcap_{j=0}^{n}g^j(\mathcal Y)
$$

consists of $2^n$ disjoint vertical strips. These are the points whose previous $n$ iterates lie in $\mathcal Y$.

Each additional survival condition doubles the number of strips and reduces their thickness. Consequently, the limiting set $\mathcal X$ is:

- compact;
- perfect, meaning that it has no isolated points;
- totally disconnected;
- uncountable.

Thus $\mathcal X$ is topologically a Cantor set. Geometrically, it has a product-like structure consisting of a stable Cantor set and an unstable Cantor set.

Future survival selects increasingly thin horizontal strips, while backward survival selects increasingly thin vertical strips. Their nested intersections shrink to individual points.


## 4. The Symbol Space

Let $\Sigma_2$ be the space of all bi-infinite sequences of two symbols:

$$
\Sigma_2
=
\left\{
a=(a_j)_{j\in\mathbb Z}:
a_j\in\{0,1\}
\right\}.
$$

A typical element is written as

$$
a=\ldots a_{-2}a_{-1}.a_0a_1a_2\ldots.
$$

Equip $\Sigma_2$ with the metric

$$
d(a,b)
=
\sum_{j=-\infty}^{\infty}
|a_j-b_j|\,2^{-|j|}.
$$

Two sequences are close if they agree on a long block around the index $j=0$.

The left-shift map is

$$
\sigma:\Sigma_2\to\Sigma_2,
\qquad
(\sigma(a))_j=a_{j+1}.
$$

In sequence notation,

$$
\sigma(\ldots a_{-1}.a_0a_1a_2\ldots)
=
\ldots a_{-1}a_0.a_1a_2\ldots.
$$

The use of bi-infinite sequences is essential: the horseshoe is invertible on $\mathcal X$, so both past and future itineraries are part of the dynamics.


## 5. Coding Horseshoe Orbits

For $x\in\mathcal X$, define its itinerary

$$
\phi(x)=(a_j)_{j\in\mathbb Z}
$$

by

$$
a_j=i
\quad\Longleftrightarrow\quad
g^j(x)\in\mathcal H_i,
\qquad i\in\{0,1\}.
$$

Thus $a_j$ records which horizontal strip contains the $j$-th iterate. Applying $g$ moves every symbol one position to the left:

$$
\phi(g(x))_j=a_{j+1}.
$$

Therefore,

$$
\phi\circ g=\sigma\circ\phi.
$$

The basic symbolic-geometric dictionary is:

| Symbolic information | Geometric information |
|---|---|
| $a_0$ | The horizontal strip containing the current point |
| $a_0,a_1,a_2,\ldots$ | The future itinerary |
| $\ldots,a_{-2},a_{-1}$ | The past itinerary |
| Future symbols | Determine the vertical, unstable coordinate |
| Past symbols | Determine the horizontal, stable coordinate |
| Periodic sequence | Periodic orbit |
| Long central agreement | Points lie in the same thin cylinder rectangle |
| Left shift $\sigma$ | One application of $g$ |

Points with the same future itinerary differ only in the stable direction and approach one another under forward iteration. Points with the same past itinerary are related through the unstable direction.


## 6. Horseshoe Symbolic Dynamics

### Theorem 13.1

The map

$$
g:\mathcal X\to\mathcal X
$$

is topologically conjugate to

$$
\sigma:\Sigma_2\to\Sigma_2.
$$

That is, there exists a homeomorphism $\phi:\mathcal X\to\Sigma_2$ satisfying

$$
\phi\circ g=\sigma\circ\phi.
$$

### Proof

For a finite central word

$$
b=b_{-m}\ldots b_{-1}.b_0\ldots b_n,
$$

define its cylinder rectangle by

$$
R(b)
=
\left\{
x\in\mathcal Y:
g^j(x)\in\mathcal H_{b_j}
\text{ for }-m\le j\le n
\right\}.
$$

The future conditions restrict the vertical coordinate, while the past conditions restrict the horizontal coordinate. The rectangle has dimensions

$$
\operatorname{width}R(b)=\lambda^m,
\qquad
\operatorname{height}R(b)=\mu^{-(n+1)}.
$$

Hence

$$
\operatorname{diam}R(b)
\le
\sqrt{\lambda^{2m}+\mu^{-2(n+1)}}
\longrightarrow0
$$

as $m,n\to\infty$.

#### Injectivity

Two points with the same bi-infinite itinerary belong to the same cylinder rectangle for every $m$ and $n$. Since the diameters converge to zero, the two points coincide.

#### Surjectivity

Let $a\in\Sigma_2$ be prescribed. The rectangles

$$
R(a_{-m}\ldots a_n)
$$

are nonempty, compact, and nested. Therefore,

$$
\bigcap_{m,n\ge0}R(a_{-m}\ldots a_n)\neq\varnothing.
$$

Because their diameters converge to zero, the intersection contains exactly one point. Thus every symbolic sequence is realized by a unique point of $\mathcal X$.

#### Continuity

Agreement on a sufficiently long central symbolic block places two points in the same arbitrarily small cylinder rectangle. This proves continuity. Since $\mathcal X$ is compact and $\Sigma_2$ is Hausdorff, the continuous bijection $\phi$ is a homeomorphism.

### Clarification: nested cylinders and continuity

Adding one future symbol adds one constraint, so

$$
R(b_0\ldots b_{n+1})
=
R(b_0\ldots b_n)
\cap
g^{-(n+1)}(\mathcal H_{b_{n+1}})
\subseteq
R(b_0\ldots b_n).
$$

The smaller set is a thinner horizontal substrip. Strictly speaking, the strip $R(b_0\ldots b_n)$ needs the cumulative iterate $g^{n+1}$ to become a vertical strip crossing $\mathcal Y$; this crossing guarantees that either next symbol can be appended.

The estimate $\operatorname{diam}R(b)\to0$ directly proves uniqueness of the coded point and continuity of the decoding map $\phi^{-1}$. Continuity of $\phi$ itself also uses separation of $\mathcal H_0$ and $\mathcal H_1$: for finitely many times $|j|\le N$, continuity of $g^j$ ensures that a sufficiently small perturbation of $x$ cannot switch its strip labels. Equivalently, finite cylinder sets have open preimages in $\mathcal X$.


## 7. Dynamical Consequences

The conjugacy translates elementary properties of binary sequences into geometric properties of the horseshoe.

### Periodic orbits

Every periodic sequence corresponds to a periodic point. For example,

$$
\ldots000.000\ldots
$$

and

$$
\ldots111.111\ldots
$$

correspond to fixed points, while

$$
\ldots0101.0101\ldots
$$

corresponds to a period-two orbit. There are finitely many binary sequences of each fixed period, so the set of periodic points is countably infinite.

For a period-$p$ orbit, the two multipliers have magnitudes

$$
\lambda^p<1,
\qquad
\mu^p>1.
$$

Consequently, every periodic orbit in the horseshoe is of saddle type. Periodic points are also dense: any prescribed finite symbolic block can be repeated periodically.

### Non-periodic orbits

The space $\Sigma_2$ is uncountable, whereas the set of periodic sequences is countable. Hence $\mathcal X$ contains uncountably many non-periodic orbits.

### A dense orbit

Construct a sequence whose positive tail contains every finite binary word. Shifting it eventually places any prescribed word around the central index. Its shift orbit is dense in $\Sigma_2$, so the corresponding horseshoe orbit is dense in $\mathcal X$. This does not mean that every orbit is dense.

### Hyperbolicity

The stable and unstable subspaces are

$$
E^s=\operatorname{span}(e_1),
\qquad
E^u=\operatorname{span}(e_2).
$$

For $v^s\in E^s$ and $v^u\in E^u$,

$$
\|Dg^n v^s\|
=
\lambda^n\|v^s\|
\longrightarrow0,
$$

$$
\|Dg^{-n}v^u\|
=
\mu^{-n}\|v^u\|
\longrightarrow0.
$$

Thus $\mathcal X$ is a hyperbolic invariant set. By persistence of hyperbolic sets, sufficiently small $C^1$ perturbations still contain a nearby invariant set with conjugate symbolic dynamics.

### Clarification: sequence points, orbits, and T1-T3

A single bi-infinite sequence $a\in\Sigma_2$ is one point of the symbolic phase space. Its full orbit is

$$
\mathcal O_\sigma(a)=\{\sigma^n(a):n\in\mathbb Z\}.
$$

The sequence records the entire itinerary of one horseshoe point, while all shifts of that sequence correspond to all points on the horseshoe orbit. Periodic sequences are countable because, for each period $p$, there are at most $2^p$ words. The non-periodic sequences are uncountable; since every individual orbit is countable, there must also be uncountably many distinct non-periodic orbits. A dense orbit is obtained by choosing a sequence whose forward tail contains every finite binary word, so its shifts enter every cylinder set.


## 8. Generalized Horseshoes

Straight strips are too restrictive for applications because return maps arising from differential equations usually produce curved strips.

### Curved strips

A vertical curve has the form $x=v(y)$ and satisfies

$$
0\le v(y)\le1,
\qquad
|v(y_l)-v(y_r)|
\le
\rho|y_l-y_r|,
\qquad
0<\rho<1.
$$

A horizontal curve has the form $y=h(x)$ and satisfies

$$
0\le h(x)\le1,
\qquad
|h(x_l)-h(x_r)|
\le
\rho|x_l-x_r|.
$$

Two nonintersecting vertical curves $v_1(y)<v_2(y)$ bound the strip

$$
\mathcal V
=
\left\{
(x,y)\in\mathbb R^2:
v_1(y)\le x\le v_2(y),
\ 0\le y\le1
\right\}.
$$

Similarly, $h_1(x)<h_2(x)$ bound the horizontal strip

$$
\mathcal H
=
\left\{
(x,y)\in\mathbb R^2:
h_1(x)\le y\le h_2(x),
\ 0\le x\le1
\right\}.
$$

### Cone conditions

Suppose there are $N$ disjoint horizontal strips $\mathcal H_i$ and vertical strips $\mathcal V_i$ satisfying

$$
g(\mathcal H_i)=\mathcal V_i,
\qquad
i=0,\ldots,N-1.
$$

To avoid confusing the cone parameter with the earlier expansion factor $\mu>2$, let $0<\theta<1/2$ be the cone aperture. At a point $p$, define cones of tangent vectors $\xi=(\xi_1,\xi_2)$ by

$$
K_p^u
=
\{\xi:|\xi_1|<\theta|\xi_2|\},
\qquad
K_p^s
=
\{\xi:|\xi_2|<\theta|\xi_1|\}.
$$

The required cone invariance is

$$
Dg_p(K_p^u)\subset K_{g(p)}^u,
\qquad
Dg_p^{-1}(K_p^s)\subset K_{g^{-1}(p)}^s.
$$

Uniform expansion is expressed by

$$
|(Dg_p\xi)_2|
\ge
\frac1\theta|\xi_2|,
\qquad
\xi\in K_p^u,
$$

$$
|(Dg_p^{-1}\xi)_1|
\ge
\frac1\theta|\xi_1|,
\qquad
\xi\in K_p^s.
$$

The source additionally assumes the determinant bounds

$$
|\det Dg|,
\ |\det Dg^{-1}|
\le
\frac12\theta^{-2}.
$$

### Why A1, A2, and the determinant bound are needed

The variables $\xi=(\xi_1,\xi_2)$ are tangent-vector components; the subscript $2$ in $(Dg_p\xi)_2$ means the second, vertical component. Condition A1 replaces exact invariance of the coordinate axes: near-vertical vectors remain near-vertical under $Dg$, and near-horizontal vectors remain near-horizontal under $Dg^{-1}$. Condition A2 adds quantitative expansion: the vertical component expands forward by at least $1/\theta$, while the horizontal component expands backward, equivalently contracts forward.

For the diagonal standard horseshoe, cone invariance alone holds for every cone width $\theta>0$. A2 restricts the choice because it requires

$$
\mu\ge\frac1\theta,
\qquad
\frac1\lambda\ge\frac1\theta,
$$

or equivalently $\theta\ge\max\{\lambda,1/\mu\}$. For nonlinear maps, off-diagonal derivative terms mean that even A1 only holds for a suitably chosen cone.

The determinant is the local area multiplier, but it appears here more specifically in the derivative of the slope transform. If

$$
Dg_p=
\begin{pmatrix}a&b\\c&d\end{pmatrix},
\qquad
\widehat s=\frac{as+b}{cs+d},
$$

then

$$
\frac{d\widehat s}{ds}
=
\frac{\det Dg_p}{(cs+d)^2}.
$$

A2 gives $|cs+d|\ge1/\theta$, so the determinant bound yields

$$
\left|\frac{d\widehat s}{ds}\right|
\le
|\det Dg_p|\theta^2
\le
\frac12.
$$

Thus the graph transform contracts slopes. The numerical constant $1/2$ is not essential; any uniform constant strictly below $1$ would serve.

### Theorem 13.4: Generalized Horseshoe Map

Under the strip, cone-invariance, expansion, and determinant assumptions, $g$ has a hyperbolic invariant set $\mathcal X$. On this set, $g$ is topologically conjugate to the full shift on $N$ symbols,

$$
\Sigma_N
=
\{0,1,\ldots,N-1\}^{\mathbb Z}.
$$

The proof follows the same nested-strip argument as for the standard horseshoe.


## 9. The $\lambda$-Lemma

The $\lambda$-Lemma, also called the inclination lemma, explains how transverse disks become aligned with an unstable manifold.

Let $0$ be a hyperbolic fixed point of a diffeomorphism $g$, with

$$
\dim W^u(0)=k,
\qquad
\dim W^s(0)=l.
$$

Suppose a $k$-dimensional $C^1$ manifold $\mathcal M$ intersects $W^s(0)$ transversely at $q$.

### Lemma 13.6

For any $r>0$ and $\delta>0$, there exist an integer $n_0$, a $k$-dimensional disk $D\subset\mathcal M$, and a point $x\in W^u(0)$ such that for every $n\ge n_0$,

$$
d_1\left(
B(x,r)\cap W^u(0),
g^n(D)
\right)
<\delta.
$$

Here $d_1$ denotes $C^1$-distance: both the sets and their tangent spaces are close.

### Geometric interpretation

A disk crossing the stable manifold is first drawn toward the fixed point. Its stable components contract, while its unstable components expand. It then emerges as a long disk almost tangent to the unstable manifold. The conclusion is not convergence to one point; it is convergence of manifold pieces, including their tangent directions.

### Proof mechanism

Use local stable-unstable coordinates $x=(x_s,x_u)$ and write

$$
g(x_s,x_u)
=
\left(
A^s x_s+\varphi_s(x),
A^u x_u+\varphi_u(x)
\right),
$$

where

$$
\|A^s\|\le a<1,
\qquad
\|(A^u)^{-1}\|\le a<1,
$$

and

$$
\varphi_s(x),\varphi_u(x)=o(\|x\|).
$$

For a tangent vector $v_n=(v_n^s,v_n^u)$, define its slope relative to the unstable direction by

$$
\Lambda_n
=
\frac{\|v_n^s\|}{\|v_n^u\|}.
$$

For the linearized map,

$$
\Lambda_{n+1}\le a^2\Lambda_n,
\qquad
\|v_{n+1}^u\|\ge a^{-1}\|v_n^u\|.
$$

Therefore $\Lambda_n\to0$ while the unstable extent grows. The nonlinear terms are controlled by restricting to a sufficiently small neighborhood and using cone estimates.

### What the two estimates actually prove

The ratio $\Lambda_n=\|v_n^s\|/\|v_n^u\|$ controls direction: $\Lambda_n\to0$ means the tangent space aligns with $E^u$. The growth of $\|v_n^u\|$ controls scale: a sufficiently small disk can be stretched to cover a prescribed unstable length. Either estimate alone is insufficient for $C^1$ closeness.

For the purely block-diagonal linear map, the slope improves by $a^2$. Nonlinear cross terms mix stable and unstable components, so the proof normally weakens this to a bound such as $\Lambda_{n+1}\le a\Lambda_n$; the lost factor is margin used to absorb the higher-order terms.

The calculation in the source is a proof skeleton. Turning tangent-vector growth into closeness to a radius-$r$ unstable disk additionally requires uniform derivative estimates on the whole chosen disk, integration of those estimates, and the already stated $C^0$ control.


## 10. The Smale-Birkhoff Homoclinic Theorem

Let $0$ be a hyperbolic saddle fixed point of a diffeomorphism

$$
g:\mathbb R^d\to\mathbb R^d.
$$

A point $q\neq0$ is homoclinic if

$$
q\in W^s(0)\cap W^u(0).
$$

The intersection is transverse when

$$
T_qW^s(0)+T_qW^u(0)
=
T_q\mathbb R^d.
$$

### Theorem 13.5

If $q$ is a transverse homoclinic point, then there exist a hyperbolic invariant set $\Lambda$ and a sufficiently large positive iterate $g^\kappa$ such that

$$
g^\kappa|_\Lambda
$$

is topologically conjugate to the Smale horseshoe map, and hence to the full two-symbol shift.

The essential implication is

$$
\boxed{
\text{transverse homoclinic intersection}
\Longrightarrow
\text{horseshoe dynamics}
}.
$$

### Proof sketch

1. Choose local coordinates in which $W^s(0)$ and $W^u(0)$ are coordinate axes.

2. Since $q$ is homoclinic,

   $$
   g^n(q)\to0,
   \qquad
   g^{-n}(q)\to0
   $$

   as $n\to\infty$.

3. Choose a thin horizontal rectangle $\mathcal R$ near, but separated from, the stable manifold.

4. By the $\lambda$-Lemma, a sufficiently large iterate $g^k(\mathcal R)$ becomes a long, thin strip almost parallel to $W^u(0)$.

5. The global homoclinic excursion returns this strip across $\mathcal R$, producing the geometry of horizontal and vertical horseshoe strips.

6. By increasing $k$, stable contraction and unstable expansion become strong enough to satisfy the generalized horseshoe cone conditions.

7. The generalized horseshoe theorem produces an invariant set conjugate to the binary shift.

### Where stretching, bending, and the branches come from

Write $q_j=g^j(q)$. Then $q_n\to0$ along $W^s(0)$ as $n\to+\infty$, while $q_{-n}\to0$ along $W^u(0)$. Choosing $n_s,n_u>0$ and $k=n_s+n_u$ gives the exact orbit matching

$$
g^k(q_{-n_u})=q_{n_s},
\qquad
g^k(q_{-n_u+1})=q_{n_s+1}.
$$

These marked points locate the return time and position; points themselves are not stretched and do not determine the shape of an entire strip.

Three different ingredients have distinct roles:

- local hyperbolicity and the $\lambda$-Lemma make a transverse rectangle long, thin, and $C^1$-aligned with $W^u(0)$;
- the global part of the homoclinic orbit carries that thin strip out of the local linear neighborhood, bends it, and returns it toward $W^s(0)$;
- transversality makes the returned strip cross the chosen rectangle rather than merely touch it.

Thus the $\lambda$-Lemma supplies stretching and alignment, but not bending. Bending is a global effect. In a vertical strip near $W^u$, the two long sides follow nearby global unstable curves and become the long curved boundaries of the return strip; the two short sides remain transverse end caps. Near the return to $W^s$, the strip crosses the horizontal reference rectangle approximately vertically.

To obtain the full two-symbol horseshoe, one chooses a sufficiently large iterate and suitable subrectangles so that the returned set has two separated crossing components $V_0,V_1$, with horizontal preimages $H_0,H_1$ and

$$
g^k(H_i)=V_i.
$$

Figure 34 is schematic: it illustrates the homoclinic return and boundary alignment, but it does not by itself display every component of the final two-branch Markov partition. The existence of the separated branches and the cone estimates are part of the geometric verification compressed into the proof sketch.

The conclusion concerns an iterate $g^\kappa$ restricted to a particular invariant subset, not necessarily the entire original map.

### Why Poincare maps are needed for flows

At a non-equilibrium homoclinic point of a continuous-time flow, the vector-field direction belongs to both $T_qW^s$ and $T_qW^u$. The manifolds therefore cannot intersect transversely in the full phase space. A Poincare section removes the common flow direction, allowing the return map to possess transverse homoclinic points.


## 11. The Baker's Map

The baker's map provides an explicit model of symbolic shifting:

$$
g:[0,1)^2\to[0,1)^2,
$$

$$
g(x,y)
=
\begin{cases}
\left(2x,\dfrac12y\right),
&0\le x<\dfrac12,\\[4pt]
\left(2x-1,\dfrac12y+\dfrac12\right),
&\dfrac12\le x<1.
\end{cases}
$$

Write the coordinates in binary:

$$
x=0.x_1x_2x_3\ldots,
\qquad
y=0.y_1y_2y_3\ldots.
$$

Then

$$
g(x,y)
=
\left(
0.x_2x_3x_4\ldots,
0.x_1y_1y_2y_3\ldots
\right).
$$

Thus the first binary digit of $x$ is removed from $x$ and inserted at the beginning of $y$. After arranging the digits as a bi-infinite sequence, this operation becomes a shift.

For dyadic rational numbers, one must choose a consistent binary-expansion convention because such numbers have two binary representations.

The baker's map makes the symbolic mechanism explicit without an escape set: stretching records the next symbol, while stacking stores the previous symbol in the second coordinate.


## 12. Common Pitfalls

1. **Confusing the two kinds of strips.** The set $g(\mathcal Y)\cap\mathcal Y$ consists of vertical strips, whereas $\mathcal Y\cap g^{-1}(\mathcal Y)$ consists of horizontal strips.

2. **Forgetting that Figure 32 shows $g^{-1}$.** Its visible stretching direction is the reverse of the forward map.

3. **Treating the entire square as invariant.** Only $\mathcal X$ remains in the square for every $k\in\mathbb Z$.

4. **Using one-sided sequences.** The invertible horseshoe requires bi-infinite sequences because both past and future matter.

5. **Thinking that a finite word determines one point.** A finite word determines a rectangle; a complete bi-infinite sequence determines one point.

6. **Reversing the stable and unstable directions.** For $g$, the horizontal direction is stable and the vertical direction is unstable.

7. **Letting the folding signs obscure hyperbolicity.** Contraction and expansion are determined by the magnitudes of the eigenvalues.

8. **Thinking that every orbit is dense.** The theorem only asserts the existence of at least one dense orbit.

9. **Thinking that the Cantor set is countable.** It is totally disconnected but uncountable.

10. **Reading cone variables as spatial coordinates.** Cone conditions apply to tangent vectors at base points.

11. **Misreading the $\lambda$-Lemma.** It gives $C^1$-closeness of manifold pieces, not pointwise convergence to a single point.

12. **Overstating Smale-Birkhoff.** It produces horseshoe dynamics for an iterate restricted to an invariant subset.

## 14. Summary

The logical structure of the lecture is

$$
\boxed{
\begin{aligned}
&\text{stretching, contraction, and folding}\\
&\qquad\Longrightarrow\text{an invariant Cantor set}\\
&\qquad\Longrightarrow\text{symbolic coding}\\
&\qquad\Longrightarrow\text{full shift dynamics}\\
&\qquad\Longrightarrow\text{periodic and non-periodic behavior}\\
&\qquad\Longrightarrow\text{robust hyperbolicity}.
\end{aligned}
}
$$

The standard horseshoe turns geometry into symbolic dynamics through the conjugacy

$$
\phi\circ g=\sigma\circ\phi.
$$

The Smale-Birkhoff theorem supplies the fundamental geometric criterion

$$
\boxed{
\text{transverse homoclinic point}
\Longrightarrow
\text{an embedded Smale horseshoe}
}.
$$

Therefore, a single transverse homoclinic intersection forces infinitely many periodic orbits, uncountably many non-periodic orbits, dense-orbit behavior, and a robust hyperbolic invariant set.
