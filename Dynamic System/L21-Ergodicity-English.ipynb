## 21.0 Measure-Preserving Systems

Let $(X,\mathcal F,\mu)$ be a probability space and let $g:X\to X$ be measurable. The map $g$ is **measure-preserving** if

$$
\mu(g^{-1}(A))=\mu(A)
$$

for every $A\in\mathcal F$. Consequently, for every $h\in L^1$,

$$
\int_X h\circ g\,d\mu=\int_Xh\,d\mu.
$$

For an indicator function, this follows directly from

$$
\int_X\mathbf 1_A\circ g\,d\mu
=\mu(g^{-1}(A))
=\mu(A)
=\int_X\mathbf 1_A\,d\mu.
$$

Linearity and approximation extend the result to all integrable functions. Every iterate $g^j$ is also measure-preserving, so

$$
\int_X|f\circ g^j|\,d\mu=\int_X|f|\,d\mu.
$$

### Why must $g$ be measurable?

If $g$ is not measurable, $f\circ g$ need not be measurable even when $f$ is. The Koopman operator, orbit sums, maximal functions, and their integrals may then fail to be defined. The standard definition of a measure-preserving map already includes measurability.


## 21.1 Definition of Ergodicity

A measure-preserving map $g$ is **ergodic** if every invariant measurable set has zero or full measure:

$$
g^{-1}(A)=A
\quad\Longrightarrow\quad
\mu(A)=0\ \text{or}\ 1.
$$

Thus, the system cannot be decomposed into two invariant regions that both have positive measure.

### Ergodicity does not mean reaching almost every point

The orbit of $x$ is

$$
\mathcal O(x)=\{x,g(x),g^2(x),\ldots\}.
$$

In a continuous space, this orbit is usually countable and may satisfy $\mu(\mathcal O(x))=0$. The correct interpretation is:

> For almost every initial point, the orbit visits each fixed positive-measure region with the correct long-term frequency; it does not necessarily pass through almost every individual point.

If $X$ is a compact metric space, $\mu$ has full support, and $g$ is ergodic, then almost every orbit is dense in $X$. This is a consequence of Birkhoff's theorem, not the definition of ergodicity. Without full support, one generally obtains density only in $\operatorname{supp}\mu$, and the conclusion concerns almost every orbit rather than every orbit.


## 21.2 Invariant-Function Characterization

A measurable function $f:X\to\mathbb R$ is **invariant** if

$$
f\circ g=f
\qquad\text{almost everywhere}.
$$

### Lemma 21.2

The map $g$ is ergodic if and only if every measurable invariant function is constant almost everywhere.

### Proof idea

Suppose $g$ is ergodic. For every $c\in\mathbb R$, the set

$$
A_c=\{x:f(x)<c\}
$$

is invariant, so $\mu(A_c)\in\{0,1\}$. If $f$ were not almost everywhere constant, one could choose $c$ between two essential values of $f$, making both $A_c$ and its complement have positive measure. This contradicts ergodicity.

Conversely, if $A$ is invariant, then

$$
\mathbf 1_A\circ g=\mathbf 1_A.
$$

If every invariant function is constant almost everywhere, $\mathbf 1_A$ must equal either $0$ or $1$ almost everywhere. Hence $\mu(A)=0$ or $1$.

### Important clarification

Even when $g$ is ergodic, an arbitrary $f\in L^1$ need not be constant. The lemma applies only to functions satisfying $f\circ g=f$. In Birkhoff's theorem, the original observable $f$ is arbitrary; the limiting time-average $f^*$ is invariant and therefore becomes constant in the ergodic case.


## 21.3 Example 21.3: Circle Rotations

Consider the rotation

$$
g(x)=x+r\pmod1
$$

on $S^1\simeq[0,1]/(0\sim1)$ with Lebesgue measure. Then

$$
g\text{ is ergodic}
\quad\Longleftrightarrow\quad
r\notin\mathbb Q.
$$

Let an invariant $f\in L^2(S^1)$ have Fourier expansion

$$
f(x)=\sum_{n\in\mathbb Z}a_ne^{2\pi inx}.
$$

Since $f(x+r)=f(x)$, uniqueness of Fourier coefficients gives

$$
a_ne^{2\pi inr}=a_n.
$$

If $r$ is irrational, then $e^{2\pi inr}\neq1$ for every $n\neq0$, so $a_n=0$ for all $n\neq0$. Only the constant mode remains, and Lemma 21.2 implies ergodicity.

If $r=p/q$ is rational, then the nonconstant function

$$
f(x)=\cos(2\pi qx)
$$

satisfies $f\circ g=f$, so the rotation is not ergodic.

For irrational $r$, every orbit is dense in the circle. Nevertheless, each orbit is countable and has Lebesgue measure zero. Density means approaching every point arbitrarily closely, not passing through almost every point exactly.


## 21.4 The Koopman Operator

Define

$$
(\mathcal L_gf)(x)=f(g(x)).
$$

Its iterates satisfy

$$
\mathcal L_g^jf=f\circ g^j.
$$

If $g$ is measure-preserving, then

$$
\|\mathcal L_gf\|_{L^p}=\|f\|_{L^p},
$$

so $\mathcal L_g$ is an isometry. It is also positive:

$$
f\ge0\quad\Longrightarrow\quad\mathcal L_gf\ge0.
$$

Measure preservation ensures that the Koopman operator maps $L^p$ into itself while preserving the relevant norms and integrals.


## 21.5 The Maximal Ergodic Theorem

For $f\in L^1$, define the partial sums

$$
f_0=0,
\qquad
f_n=f+\mathcal Lf+\cdots+\mathcal L^{n-1}f,
$$

so that

$$
f_n(x)=\sum_{j=0}^{n-1}f(g^j(x)).
$$

Define the maximal function

$$
F_N(x)=\max_{0\le n\le N}f_n(x)\ge0.
$$

The quantity $F_N(x)$ compares partial sums of different lengths, not the individual values of $f$ along the orbit. Let

$$
A_N=\{x:F_N(x)>0\}.
$$

Then

$$
x\in A_N
\iff
\exists\,1\le n\le N:\ f_n(x)>0.
$$

Thus, $A_N$ contains the initial points whose first $N$ orbit steps produce a positive partial sum at some time. The theorem integrates the original function $f$:

$$
\boxed{\int_{A_N}f\,d\mu\ge0.}
$$

### Main proof steps

Since $F_N\ge f_{n-1}$ and the Koopman operator is positive,

$$
\mathcal LF_N+f
\ge\mathcal Lf_{n-1}+f
=f_n.
$$

On $A_N$, the maximum cannot come only from $f_0=0$. Therefore

$$
\mathcal LF_N+f\ge F_N,
\qquad
f\ge F_N-\mathcal LF_N.
$$

On $X\setminus A_N$, we have $F_N\ge0$ and $F_N\not>0$, so $F_N=0$. Hence

$$
\int_{A_N}F_N\,d\mu=\int_XF_N\,d\mu.
$$

Using $\mathcal LF_N\ge0$ and measure preservation,

$$
\int_{A_N}f\,d\mu
\ge
\int_XF_N\,d\mu-\int_X\mathcal LF_N\,d\mu
=0.
$$


## 21.6 The Maximal Inequality

For $h\in L^1$, define

$$
B_\alpha
=
\left\{
x:
\sup_{m\ge1}\frac1m\sum_{j=0}^{m-1}h(g^j(x))>\alpha
\right\}.
$$

Thus, $x\in B_\alpha$ means that some finite-time average starting from $x$ exceeds $\alpha$. If $A$ is invariant, Corollary 21.6 states

$$
\int_{B_\alpha\cap A}h\,d\mu
\ge
\alpha\,\mu(B_\alpha\cap A).
$$

The proof applies the Maximal Ergodic Theorem to $h-\alpha$ on the invariant subsystem $A$ and lets the finite time range tend to infinity. This inequality is later used to rule out a gap between the upper and lower limiting time averages.


## 21.7 The Birkhoff Ergodic Theorem

Define the time average

$$
S_nf(x)=\frac1n\sum_{j=0}^{n-1}f(g^j(x)).
$$

For every $f\in L^1$:

1. There exists $f^*\in L^1$ such that $S_nf(x)\to f^*(x)$ for almost every $x$.
2. The limit satisfies $f^*\circ g=f^*$ and

   $$
   \int_Xf^*\,d\mu=\int_Xf\,d\mu.
   $$

3. If $g$ is ergodic, then

   $$
   f^*(x)=\int_Xf\,d\mu
   \qquad\text{a.e.}
   $$

Thus, in the ergodic case, the time average equals the space average.

### Proof structure

Define

$$
\overline f=\limsup_{n\to\infty}S_nf,
\qquad
\underline f=\liminf_{n\to\infty}S_nf.
$$

The identity

$$
S_nf(g(x))
=\frac{n+1}{n}S_{n+1}f(x)-\frac{f(x)}n
$$

shows that $\overline f$ and $\underline f$ are invariant. This shift-invariance comes from the form of the time average and does not itself require measure preservation.

For rational numbers $\beta<\alpha$, define

$$
E_{\alpha,\beta}
=\{x:\underline f(x)<\beta<\alpha<\overline f(x)\}.
$$

Since $E_{\alpha,\beta}\subseteq B_\alpha$, Corollary 21.6 gives

$$
\int_{E_{\alpha,\beta}}f\,d\mu
\ge\alpha\mu(E_{\alpha,\beta}).
$$

Applying the same result to $-f$ gives

$$
\int_{E_{\alpha,\beta}}f\,d\mu
\le\beta\mu(E_{\alpha,\beta}).
$$

Because $\beta<\alpha$, we obtain $\mu(E_{\alpha,\beta})=0$. A countable union over rational pairs proves $\underline f=\overline f$ almost everywhere, so the limit exists.

Fatou's Lemma and measure preservation give

$$
\int_X|f^*|\,d\mu
\le\liminf_n\int_X|S_nf|\,d\mu
\le\int_X|f|\,d\mu,
$$

so $f^*\in L^1$. Dividing the range of $f^*$ into intervals of length $1/n$, applying Corollary 21.6 on the resulting invariant sets, and repeating the argument with $-f$ yields

$$
\int_Xf^*\,d\mu=\int_Xf\,d\mu.
$$

Finally, if $g$ is ergodic, the invariant function $f^*$ is constant almost everywhere. Since $\mu(X)=1$, its value is $\int_Xf\,d\mu$.

### Visit frequencies and dense orbits

Taking $f=\mathbf 1_B$ gives, for every fixed measurable set $B$ in an ergodic system,

$$
\lim_{n\to\infty}
\frac1n\sum_{j=0}^{n-1}\mathbf 1_B(g^j(x))
=\mu(B)
$$

for almost every $x$. If $\mu(B)>0$, a typical orbit enters $B$ infinitely often with asymptotic frequency $\mu(B)$. With a countable topological basis and full support, taking a countable intersection over the basis elements shows that a typical orbit is dense.


## 21.8 The von Neumann $L^p$ Ergodic Theorem

Let $1\le p<\infty$ and $f\in L^p$. Then there exists an invariant function $f^*\in L^p$ such that

$$
\|S_nf-f^*\|_{L^p}\longrightarrow0.
$$

### Proof idea

Choose $h\in L^\infty$ such that

$$
\|f-h\|_{L^p}<\frac{\varepsilon}{3}.
$$

Birkhoff's theorem gives almost-everywhere convergence of $S_nh$. Since $|S_nh|\le\|h\|_\infty$, the Dominated Convergence Theorem upgrades this to $L^p$ convergence.

Ergodic averages are contractions:

$$
\|S_nu\|_{L^p}\le\|u\|_{L^p}.
$$

A standard $\varepsilon/3$ argument therefore shows that $(S_nf)$ is a Cauchy sequence in $L^p$. The limit is invariant because

$$
(S_nf)\circ g-S_nf
=\frac{f\circ g^n-f}{n},
$$

whose $L^p$ norm is at most $2\|f\|_{L^p}/n$.

### What the theorem does not imply by itself

The von Neumann theorem only guarantees convergence to an invariant function $f^*$. In a non-ergodic system, $f^*$ need not be constant. More precisely,

$$
f^*=\mathbb E[f\mid\mathcal I],
$$

where $\mathcal I$ is the invariant sigma-algebra. Only with the additional assumption of ergodicity do we obtain

$$
f^*=\int_Xf\,d\mu.
$$

The logical structure is therefore

$$
\boxed{
\text{measure preservation}\Longrightarrow
\text{time averages converge to an invariant function};
\qquad
\text{ergodicity}\Longrightarrow
\text{that limit equals the space average}.
}
$$
