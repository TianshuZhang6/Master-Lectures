# Part I - L7-8: From ERM to PAC and uniform convergence


# 1. Where the course has arrived and why learning theory matters

*Slide spine: L7-8, pp. 1-3.*

Before L7-8, the course covered:

1. classification and the idea of empirical risk minimization;
2. overfitting, inductive bias, and hypothesis classes;
3. linear classifiers, logistic regression, and support vector machines;
4. linear regression and ordinary least squares;
5. nonlinear classifiers and nonlinear features;
6. kernels, their theory, and the representer theorem.

L7-8 asks a more fundamental question: **what does it mean to learn from data?** Learning theory is studied in order to

- discover better models through inductive bias, implicit regularization, and structure;
- discover algorithms that make appropriate statistical and computational tradeoffs;
- understand and explain scientifically why learning procedures succeed or fail.

The goal is to learn from a finite training set and predict accurately on unseen data. The small “Learn / ML” book graphic on slide 3 expresses the transition from merely using machine learning to understanding learning itself.


# 2. Statistical setup: data, hypotheses, loss, and risk

*Slide spine: L7-8, pp. 4-6.*

Let the data domain be an arbitrary set $\mathcal X$, often $\mathbb R^d$, and let the label domain be a discrete set $\mathcal Y$, such as $\{0,1\}$ or $\{-1,1\}$. A feature map may be written $\phi(x)$ when one wants to emphasize the representation of an input.

Let

$$
\mathcal Z=\mathcal X\times\mathcal Y,
\qquad
Z=(X,Y)\sim\mathbb P,
$$

where $\mathbb P$ is a fixed but unknown distribution. The training sample is

$$
S=(Z_1,\ldots,Z_n)
=((X_1,Y_1),\ldots,(X_n,Y_n))
\sim\mathbb P^n.
$$

The abbreviation i.i.d. means independent and identically distributed. A hypothesis, prediction rule, or classifier is a map

$$
h:\mathcal X\to\mathcal Y.
$$

The notation $h_S$ emphasizes that the learned rule depends on the random sample $S$.

For a loss function $\ell$, the population risk, true risk, or generalization error is

$$
L_{\mathbb P}(h)
=\mathbb E_{(X,Y)\sim\mathbb P}[\ell(h,X,Y)].
$$

The empirical risk is

$$
L_S(h)=\frac1n\sum_{i=1}^n\ell(h,X_i,Y_i).
$$

For binary classification, the $0$-$1$ loss is

$$
\ell_{0/1}(h,(x,y))
=\mathbf 1\{h(x)\ne y\}
=
\begin{cases}
1,&h(x)\ne y,\\
0,&h(x)=y.
\end{cases}
$$

Hence

$$
L_{\mathbb P}(h)=\mathbb P(h(X)\ne Y).
$$

The population risk is what we care about, but it cannot normally be computed because $\mathbb P$ is unknown. The empirical risk is observable and is therefore used as a proxy.

## 2.1 Supplement: the Bayes classifier

For $\mathcal Y=\{0,1\}$, define

$$
\eta(x)=\mathbb P(Y=1\mid X=x).
$$

Under $0$-$1$ loss, predicting $1$ has conditional error $1-\eta(x)$ and predicting $0$ has conditional error $\eta(x)$. The Bayes classifier is therefore

$$
h_{\mathrm{Bayes}}(x)=\mathbf 1\left\{\eta(x)\ge\frac12\right\}.
$$

It minimizes risk among all measurable classifiers, but it is generally unavailable because the data distribution is unknown. The No-Free-Lunch discussion later explains why no data-driven method can uniformly recover this ideal on every possible problem.


# 3. ERM and the central generalization question

*Slide spine: L7-8, pp. 7-11.*

Given a hypothesis class $\mathcal H$, empirical risk minimization (ERM) returns

$$
h_S\in\operatorname*{argmin}_{h\in\mathcal H}L_S(h).
$$

The central question is:

> If ERM makes $L_S(h_S)$ small, what does that imply about $L_{\mathbb P}(h_S)$?

The desired guarantee is

$$
L_{\mathbb P}(h_S)
\le
\inf_{h\in\mathcal H}L_{\mathbb P}(h)+\varepsilon.
$$

If $L_S(h)$ approximates $L_{\mathbb P}(h)$ **simultaneously for every** $h\in\mathcal H$, then ERM cannot exploit a chance downward fluctuation of empirical risk. This idea will become uniform convergence.

For a hypothesis $h$ fixed independently of $S$,

$$
\mathbb E_S[L_S(h)]=L_{\mathbb P}(h).
$$

This unbiasedness is not enough for $h_S$, because $h_S$ is selected after seeing the same data used to evaluate $L_S$. Learning theory must control this data dependence. The slides also warn that class-level analysis is only one view: optimization and the particular solution selected by an algorithm may matter as well.


# 4. Inductive bias: restricting the ERM search

*Slide spine: L7-8, p. 8.*

ERM over all functions can memorize the sample and behave arbitrarily elsewhere. An inductive bias restricts the search space before the training data are observed. Examples include linear models, neural-network architectures, and random forests.

The learning rule is then

$$
\operatorname{ERM}_{\mathcal H}(S)
\in\operatorname*{argmin}_{h\in\mathcal H}L_S(h).
$$

Ideally, $\mathcal H$ reflects prior knowledge about the data. A class can still be too rich and overfit, while an overly restrictive class can underfit. There is no learning without some inductive bias; the later No-Free-Lunch theorem makes this necessity formal.


# 5. Approximation and estimation error

*Slide spine: L7-8, pp. 12-17.*

Let

$$
h_{\mathcal H}^*
\in\operatorname*{argmin}_{h\in\mathcal H}L_{\mathbb P}(h).
$$

Then

$$
\begin{aligned}
L_{\mathbb P}(h_S)
&=L_{\mathbb P}(h_{\mathcal H}^*)
+\left[L_{\mathbb P}(h_S)-L_{\mathbb P}(h_{\mathcal H}^*)\right]\\
&=\varepsilon_{\mathrm{apx}}+\varepsilon_{\mathrm{est}},
\end{aligned}
$$

where

$$
\varepsilon_{\mathrm{apx}}
=\inf_{h\in\mathcal H}L_{\mathbb P}(h)
$$

is the approximation error and

$$
\varepsilon_{\mathrm{est}}
=L_{\mathbb P}(h_S)-\inf_{h\in\mathcal H}L_{\mathbb P}(h)
$$

is the estimation error.

The approximation error measures the price of the inductive bias: it depends on how well $\mathcal H$ can express the underlying relationship and does not directly depend on $n$ or on the realized sample. The estimation error measures the price of learning from finite data: empirical risk is only a proxy for population risk.

The slides ask the reader to investigate:

- how prior knowledge and the choice of $\mathcal H$ affect approximation error;
- how class capacity, sample size, the realized sample, distribution shift, and robustness affect estimation error.

**Supplement.** If ERM is solved only approximately, one may add optimization error. For an algorithmic output $\widetilde h_S$,

$$
L_{\mathbb P}(\widetilde h_S)-L_{\mathbb P}(h_{\mathrm{Bayes}})
=\underbrace{L_{\mathbb P}(h_{\mathcal H}^*)-L_{\mathbb P}(h_{\mathrm{Bayes}})}_{\text{approximation}}
+\underbrace{L_{\mathbb P}(h_S)-L_{\mathbb P}(h_{\mathcal H}^*)}_{\text{estimation}}
+\underbrace{L_{\mathbb P}(\widetilde h_S)-L_{\mathbb P}(h_S)}_{\text{optimization}}.
$$


# 6. Classical capacity tradeoff and double descent

*Slide spine: L7-8, pp. 18-19.*

The classical plot has model capacity on the horizontal axis and risk on the vertical axis. Training risk decreases as capacity grows. Test risk is U-shaped: a weak class underfits, an intermediate capacity gives a “sweet spot,” and a very rich class overfits.

The modern double-descent plot adds an interpolation threshold. In the under-parameterized classical regime, the first U-shaped curve appears. Near the point at which training risk first becomes zero, test risk can spike. In the over-parameterized modern interpolating regime, test risk may decrease again even though training risk remains zero.

The slide cites Mikhail Belkin, Daniel Hsu, Siyuan Ma, and Soumik Mandal, *Reconciling modern machine-learning practice and the classical bias-variance trade-off*. The lesson is not that error decomposition is false. Rather, raw parameter count may be a poor complexity measure, and optimization can impose an implicit bias toward particular interpolating solutions.


# 7. A formal learnability model: realizability and a finite class

*Slide spine: L7-8, pp. 20-23.*

## 7.1 Assumption 1: realizability

The realizability assumption states that

$$
\exists h^*\in\mathcal H
\quad\text{such that}\quad
L_{\mathbb P}(h^*)=0.
$$

Then $L_S(h^*)=0$ almost surely. One proof uses

$$
\mathbb E_S[L_S(h^*)]=L_{\mathbb P}(h^*)=0
$$

and the fact that a nonnegative random variable with expectation zero is zero almost surely. Since $h^*$ achieves empirical risk zero, every ERM solution also satisfies

$$
L_S(h_S)=0.
$$

This does **not** imply $h_S=h^*$ or $L_{\mathbb P}(h_S)=0$: several hypotheses may interpolate the training data.

## 7.2 Assumption 2: finite hypothesis class

Assume

$$
|\mathcal H|<\infty.
$$

The slides call this a very strong inductive bias. The question is whether realizability plus finiteness is sufficient for ERM to learn. It is.


# 8. The finite realizable class theorem

*Slide spine: L7-8, pp. 24 and 29.*

If realizability holds and $|\mathcal H|<\infty$, then every ERM solution satisfies

$$
\mathbb P_{S\sim\mathbb P^n}
\left(L_{\mathbb P}(h_S)>\varepsilon\right)
\le |\mathcal H|e^{-n\varepsilon}.
$$

Consequently, if

$$
n\ge\frac{\log(|\mathcal H|/\delta)}{\varepsilon},
$$

then, with probability at least $1-\delta$,

$$
L_{\mathbb P}(h_S)\le\varepsilon.
$$

The two parameters have different roles: $\varepsilon$ is the accuracy tolerance and $\delta$ is the failure-probability or confidence parameter. The theorem is statistical; it says nothing about whether ERM can be computed efficiently.


# 9. Complete proof by misleading samples and a union bound

*Slide spine: L7-8, pp. 25-28.*

Define the bad hypotheses

$$
\mathcal H_B
=\{h\in\mathcal H:L_{\mathbb P}(h)>\varepsilon\}
$$

and the misleading-sample event

$$
M
=\{S:\exists h\in\mathcal H_B\text{ with }L_S(h)=0\}.
$$

Realizability gives $L_S(h_S)=0$. Therefore, if ERM fails, then $h_S$ itself is a bad hypothesis that looks perfect on the sample:

$$
\{S:L_{\mathbb P}(h_S)>\varepsilon\}\subseteq M.
$$

For a fixed $h\in\mathcal H_B$,

$$
\begin{aligned}
\mathbb P^n(L_S(h)=0)
&=(1-L_{\mathbb P}(h))^n\\
&\le(1-\varepsilon)^n\\
&\le e^{-n\varepsilon}.
\end{aligned}
$$

Independence supplies the product, and $1-u\le e^{-u}$ supplies the final inequality. Since

$$
M=\bigcup_{h\in\mathcal H_B}\{S:L_S(h)=0\},
$$

the union bound gives

$$
\begin{aligned}
\mathbb P^n(L_{\mathbb P}(h_S)>\varepsilon)
&\le\mathbb P^n(M)\\
&\le\sum_{h\in\mathcal H_B}\mathbb P^n(L_S(h)=0)\\
&\le|\mathcal H_B|e^{-n\varepsilon}\\
&\le|\mathcal H|e^{-n\varepsilon}.
\end{aligned}
$$

Requiring the last quantity to be at most $\delta$ yields the theorem. The coin-flip intuition on slide 25 is that simultaneous, independent good luck for many genuinely bad candidates becomes exponentially unlikely.

The proof sketch on L7-8 slide 25 cites Section 2.3.1 of Shalev-Shwartz and Ben-David, abbreviated **[SSS]** on the slides.


# 10. PAC learning

*Slide spine: L7-8, pp. 29-31; repeated at L9, pp. 2-3.*

Under realizability, a class $\mathcal H$ is probably approximately correct learnable, or PAC learnable, if there exist a learning algorithm and a sample-complexity function $m_{\mathcal H}(\varepsilon,\delta)$ such that, for every $\varepsilon,\delta\in(0,1)$ and every realizable distribution $\mathbb P$, whenever

$$
n\ge m_{\mathcal H}(\varepsilon,\delta),
$$

the output $h$ satisfies

$$
\mathbb P_{S\sim\mathbb P^n}
\left(L_{\mathbb P}(h)\le\varepsilon\right)
\ge1-\delta.
$$

“Probably” is necessary because a random finite sample may be unrepresentative. “Approximately” is necessary because a finite sample may omit relevant details. The slide includes Leslie Valiant, who introduced the PAC framework, and notes his 2010 Turing Award.

Two problems remain: what if realizability fails, and what if $\mathcal H$ is infinite?


# 11. No-Free-Lunch: why inductive bias is unavoidable

*Slide spine: L7-8, p. 33; repeated at L9, p. 4.*

The slides label the following statement **Theorem 5.1 (No-Free-Lunch)**. For any binary classification algorithm $A$ under $0$-$1$ loss, if

$$
m<\frac{|\mathcal X|}{2},
$$

then there is a distribution $\mathcal D$ over $\mathcal X\times\{0,1\}$ such that

1. some function $f:\mathcal X\to\{0,1\}$ has $L_{\mathcal D}(f)=0$;
2. with probability at least $1/7$ over $S\sim\mathcal D^m$,

   $$
   L_{\mathcal D}(A(S))\ge\frac18.
   $$

The constants are not the conceptual point. The theorem says that for every learner there is a task on which it fails, although a learner with a different task-specific bias may succeed. There is no universally superior learner across all possible data-generating mechanisms.

**Slide exercise.** Give a learner that succeeds on the constructed task while $A$ fails.

**Answer.** The theorem guarantees a zero-risk target $f$. The rule $B(S)=f$, with $f$ hard-coded, has zero risk. This is an existence argument, not a claim that a practical learner knows $f$ in advance; it exposes the role of prior information.


# 12. Agnostic PAC learning

*Slide spine: L7-8, pp. 34-35; L9, p. 5.*

Without realizability, it may be impossible to make absolute risk smaller than $\varepsilon$. The weaker and meaningful target is comparison with the best hypothesis in the chosen class:

$$
L_{\mathbb P}(h_S)
\le
\inf_{h\in\mathcal H}L_{\mathbb P}(h)+\varepsilon.
$$

A class is agnostic PAC learnable if one algorithm achieves this inequality with probability at least $1-\delta$, for every distribution $\mathbb P$, once $n\ge m_{\mathcal H}(\varepsilon,\delta)$. The definition applies to general risks

$$
L_{\mathbb P}(h)=\mathbb E_{\mathbb P}[\ell(h,X,Y)],
$$

not only to $0$-$1$ classification.

The approximation error is now irreducible within the chosen class. Learning controls the estimation error uniformly over all distributions. This distribution-free strength is also a limitation: the bound may be conservative because it cannot exploit benign properties of the actual distribution. L7-8 slide 35 points forward to [SSS], Section 6.1, for infinite but learnable classes.


# 13. Representative samples and uniform convergence

*Slide spine: L7-8, pp. 36-38; L9, pp. 6-8.*

A sample $S$ is $\eta$-representative for $\mathcal H$ if

$$
\sup_{h\in\mathcal H}
|L_S(h)-L_{\mathbb P}(h)|
\le\eta.
$$

If $S$ is $\varepsilon/2$-representative and $h_S$ is an ERM solution, then for every $h\in\mathcal H$,

$$
\begin{aligned}
L_{\mathbb P}(h_S)
&\le L_S(h_S)+\frac\varepsilon2\\
&\le L_S(h)+\frac\varepsilon2\\
&\le L_{\mathbb P}(h)+\varepsilon.
\end{aligned}
$$

Taking the infimum over $h$ gives

$$
L_{\mathbb P}(h_S)
\le
\inf_{h\in\mathcal H}L_{\mathbb P}(h)+\varepsilon.
$$

Thus uniform convergence is sufficient for ERM to be an agnostic PAC learner. The slides cite the formal statement as **Definition 4.3 (Uniform Convergence)**: $\mathcal H$ has the uniform convergence property if there exists $m_{\mathcal H}^{\mathrm{UC}}(\eta,\delta)$ such that, for every $\mathbb P$, whenever $n\ge m_{\mathcal H}^{\mathrm{UC}}(\eta,\delta)$,

$$
\mathbb P_{S\sim\mathbb P^n}
\left(
\sup_{h\in\mathcal H}|L_S(h)-L_{\mathbb P}(h)|\le\eta
\right)
\ge1-\delta.
$$

The slide text sometimes calls the sample $\varepsilon$-representative while the lemma uses $\varepsilon/2$. The precise rule is: a uniform deviation of $\eta$ yields ERM excess risk at most $2\eta$.

## 13.1 Supplement: finite-class uniform convergence

For bounded losses in $[0,1]$, Hoeffding's inequality gives, for a fixed $h$,

$$
\mathbb P\left(|L_S(h)-L_{\mathbb P}(h)|>\eta\right)
\le2e^{-2n\eta^2}.
$$

A union bound over finite $\mathcal H$ gives

$$
\mathbb P\left(
\sup_{h\in\mathcal H}|L_S(h)-L_{\mathbb P}(h)|>\eta
\right)
\le2|\mathcal H|e^{-2n\eta^2}.
$$

This already yields the agnostic scaling

$$
n=O\left(\frac{\log|\mathcal H|+\log(1/\delta)}{\varepsilon^2}\right).
$$

The next lecture replaces the literal finite cardinality by an effective number of labelings.


# Part II - L9: Growth functions, shattering, and VC dimension


# 14. L9's two extensions of PAC learning

*Slide spine: L9, pp. 1-9.*

L9 begins by repeating the PAC definition and the roles of $\varepsilon$ and $\delta$. It then develops two extensions:

1. replace realizability by the agnostic comparison

   $$
   L_{\mathbb P}(h_S)
   \le
   \inf_{h\in\mathcal H}L_{\mathbb P}(h)+\varepsilon;
   $$

2. replace the literal size $|\mathcal H|$, which is useless for an infinite class, by the number of distinct behaviors the class can exhibit on finitely many points.

The No-Free-Lunch theorem, representative-sample lemma, and formal uniform-convergence definition on L9 slides 4-8 are exactly the bridge developed in Sections 11-13 above. L9 slide 9 announces that the right complexity notion will be VC dimension.


# 15. Effective size and the growth function

*Slide spine: L9, pp. 10-12.*

L9 slides 10-16 credit this sequence of growth-function slides to **Y. Polyanskiy (MIT)**.

Consider homogeneous hyperplane classifiers

$$
\widehat Y(x)=\operatorname{sgn}(x^T w).
$$

There are infinitely many parameter vectors $w$. Yet after fixing points $x^{(1)},\ldots,x^{(N)}$, only finitely many label vectors can occur. This number, not the parameter-space cardinality, is the effective size of the class on the data.

For $C=\{c_1,\ldots,c_m\}\subseteq\mathcal X$, define the restriction

$$
\mathcal H_C
=\{(h(c_1),\ldots,h(c_m)):h\in\mathcal H\}.
$$

The growth function, cited again on L10 slide 5 as **Definition 6.9**, is

$$
\tau_{\mathcal H}(m)
=\max_{\substack{C\subseteq\mathcal X\\|C|=m}}|\mathcal H_C|.
$$

Equivalently,

$$
\tau_{\mathcal H}(m)
=\max_{x_1,\ldots,x_m}
\left|
\{(h(x_1),\ldots,h(x_m)):h\in\mathcal H\}
\right|.
$$

Always

$$
1\le\tau_{\mathcal H}(m)\le2^m.
$$

The slides' key claim is that $|\mathcal H|$ in a finite-class argument can often be replaced by $\tau_{\mathcal H}(N)$. Slide 12 contains an isolated “Lebesgue integral” side remark; it is a lecturer aside, not part of the growth-function definition.


# 16. One-dimensional affine classifiers

*Slide spine: L9, p. 13; L10, p. 6.*

For one-dimensional affine classifiers whose threshold direction may be reversed, two points can receive all four binary labelings:

$$
\tau_{\mathcal H}(2)=2^2=4.
$$

For three ordered points, the alternating patterns $(+,-,+)$ and $(-,+,-)$ are impossible with one boundary. The other six patterns are possible, so

$$
\tau_{\mathcal H}(3)=6<2^3.
$$

More generally, this bidirectional affine class has $\tau_{\mathcal H}(m)=2m$ for $m\ge1$. Do not confuse it with the one-direction threshold class $h_a(x)=\mathbf1\{x<a\}$, which has $m+1$ patterns and VC dimension $1$.


# 17. The growth-function dichotomy

*Slide spine: L9, pp. 14-17.*

The Vapnik-Chervonenkis dichotomy says that a binary class has one of two behaviors:

- $\tau_{\mathcal H}(m)=2^m$ for every $m$; or
- after some finite point, $\tau_{\mathcal H}(m)$ grows at most polynomially.

When the polynomial exponent is $d$, the slide gives the realizable-case heuristic

$$
\varepsilon
<\frac1N\log\frac{\tau_{\mathcal H}(N)}{\delta}
=O\left(
\frac dN\log\frac Nd+
\frac1N\log\frac1\delta
\right).
$$

This is a scale calculation, not the sharp agnostic result. A standard realizable upper bound is

$$
N=O\left(
\frac{d\log(1/\varepsilon)+\log(1/\delta)}{\varepsilon}
\right),
$$

whereas agnostic estimation has the $1/\varepsilon^2$ scale.

If $\tau_{\mathcal H}(m)=2^m$ for every $m$, the class can assign arbitrary labels to any finite set. In a worst-case construction, the sample reveals nothing about the next unseen point and the error there is $1/2$, the performance of a random guess. This is an NFL intuition, not a claim that every algorithm has risk exactly $1/2$ under every fixed distribution.

The slide summary is crucial: learnability depends on the maximum number of classifications the class can realize on finite subsets, not on literal cardinality. Excessive power permits a bad distribution that preserves realizability on observed points while making the learner fail elsewhere.


# 18. Shattering and VC dimension

*Slide spine: L9, p. 18.*

A finite set $C=\{c_1,\ldots,c_m\}$ is shattered by $\mathcal H$ if every binary labeling is realized:

$$
\forall(y_1,\ldots,y_m)\in\{0,1\}^m,
\quad
\exists h\in\mathcal H,
\quad
h(c_i)=y_i\ \text{for all }i.
$$

Equivalently,

$$
|\mathcal H_C|=2^{|C|}.
$$

The Vapnik-Chervonenkis dimension is

$$
\operatorname{VCdim}(\mathcal H)
=\sup\{|C|:\mathcal H\text{ shatters }C\}.
$$

The quantifiers matter. To prove a lower bound, one suitable set must be shattered. To prove an upper bound, **every** larger set must fail to be shattered. Infinite VC dimension implies that $\mathcal H$ is not PAC learnable.


# 19. Shattering examples and the role of parameters

*Slide spine: L9, p. 19.*

Three non-collinear points in $\mathbb R^2$ are shattered by affine linear classifiers. A line can isolate any one point, and reversing its orientation swaps positive and negative labels. The slide illustrates these separations with oriented lines; the figure is credited to C. Burges.

No four points can be shattered. If four points are in convex position, an alternating XOR labeling is not linearly separable. If one point lies in the convex hull of the other three, labeling the interior point opposite to all outer points is impossible. Therefore

$$
\operatorname{VCdim}(\text{affine halfspaces in }\mathbb R^2)=3.
$$

**Supplement: one parameter can still have infinite VC dimension.** Parameter count is not a universal capacity measure. A classical oscillatory family has the form

$$
h_\alpha(x)=\operatorname{sign}(\sin(\alpha x)).
$$

Carefully chosen inputs allow changes in the single frequency parameter $\alpha$ to realize arbitrarily many labelings. This example comes from the textbook material underlying the earlier note; it is not visible on the rendered L9 slide 19.


# 20. How to prove a VC-dimension result

*Slide spine: L9, p. 20.*

To prove $\operatorname{VCdim}(\mathcal H)=d$, prove both:

1. there exists a set of size $d$ shattered by $\mathcal H$;
2. no set of size $d+1$ is shattered by $\mathcal H$.

Thus a bound $d_1\le\operatorname{VCdim}(\mathcal H)\le d_2$ requires a shattered $d_1$-point witness and an impossibility proof for every $(d_2+1)$-point set. Showing that one particular large set is not shattered does not establish an upper bound.


# 21. Threshold functions and finite classes

*Slide spine: L9, p. 21.*

The slide presents the threshold class as **Example 6.2**. Consider

$$
\mathcal H=\{h_a:a\in\mathbb R\},
\qquad
h_a(x)=\mathbf1\{x<a\}.
$$

One point is shattered: put the threshold on either side. If $c_1\le c_2$, the labeling $(0,1)$ is impossible, because $h_a(c_1)=0$ implies $a\le c_1\le c_2$ and hence $h_a(c_2)=0$. Thus

$$
\operatorname{VCdim}(\mathcal H)=1.
$$

The class is infinite but PAC learnable.

For a finite class, if $d$ points are shattered, at least $2^d$ distinct hypotheses are needed on those points. Hence

$$
2^d\le|\mathcal H|
\quad\Longrightarrow\quad
\operatorname{VCdim}(\mathcal H)\le\log_2|\mathcal H|.
$$

The slide explicitly asks students to read Chapter 6 of Shalev-Shwartz and Ben-David and its exercises.


# 22. The fundamental theorem of statistical learning

*Slide spine: L9, p. 22; repeated and quantified at L10, p. 2.*

The slides cite this result as **Theorem 6.7 (The Fundamental Theorem of Statistical Learning)**. For binary classification with $0$-$1$ loss, the following are equivalent:

1. $\mathcal H$ has the uniform convergence property.
2. Every ERM rule is a successful agnostic PAC learner for $\mathcal H$.
3. $\mathcal H$ is agnostic PAC learnable.
4. $\mathcal H$ is PAC learnable.
5. Every ERM rule is a successful PAC learner for $\mathcal H$.
6. $\operatorname{VCdim}(\mathcal H)<\infty$.

The easy implication map noted on the slide is

$$
1\Rightarrow2\Rightarrow3\Rightarrow4,
\qquad
2\Rightarrow5.
$$

The implications $4\Rightarrow6$ and $5\Rightarrow6$ follow from an NFL construction. L10 proves the difficult direction $6\Rightarrow1$.

## 22.1 Quantitative version

**Theorem 6.8 (Quantitative Version), shown on L10 slide 2.** If $d=\operatorname{VCdim}(\mathcal H)<\infty$, then for absolute constants $C_1,C_2$,

$$
m_{\mathcal H}^{\mathrm{UC}}(\varepsilon,\delta)
=\Theta\left(\frac{d+\log(1/\delta)}{\varepsilon^2}\right),
$$

$$
m_{\mathcal H}^{\mathrm{agnostic}}(\varepsilon,\delta)
=\Theta\left(\frac{d+\log(1/\delta)}{\varepsilon^2}\right),
$$

and in the realizable setting

$$
C_1\frac{d+\log(1/\delta)}{\varepsilon}
\le
m_{\mathcal H}(\varepsilon,\delta)
\le
C_2\frac{d\log(1/\varepsilon)+\log(1/\delta)}{\varepsilon}.
$$

The agnostic case estimates small differences between nonzero risks and therefore has the mean-estimation scale $1/\varepsilon^2$. In the realizable case, a bad hypothesis must avoid making any observed error, an event whose probability decays like $e^{-n\varepsilon}$, giving the faster $1/\varepsilon$ scale.


# 23. Other complexity notions and PAC-Bayes

*Slide spine: L9, pp. 23-24.*

L9 closes by recapping representative samples, shattering, VC dimension, and the two-part VC proof template. It then lists other complexity tools:

- Pollard pseudo-dimension, a threshold-based extension of VC dimension to real-valued functions;
- covering numbers, the number of metric balls needed to cover a class at a chosen resolution;
- metric entropy, the logarithm of a covering number;
- Rademacher complexity, the ability of a function class to correlate with random signs;
- Gaussian complexity, the analogous quantity with Gaussian multipliers;
- local Rademacher complexity, which measures a relevant low-risk neighborhood rather than the whole class.

PAC-Bayes puts a prior distribution on hypotheses and analyzes a data-dependent posterior. Its bounds typically combine empirical risk, a KL-divergence complexity term, and confidence. The slide emphasizes that such generalization bounds can be data dependent and points to Chapter 13 of Shalev-Shwartz and Ben-David.


# Part III - L10: Finite VC dimension, uniform convergence, and stability


# 24. The $6\Rightarrow1$ strategy

*Slide spine: L10, pp. 1-3.*

L10 proves the difficult implication

$$
\operatorname{VCdim}(\mathcal H)<\infty
\Longrightarrow
\mathcal H\text{ has uniform convergence}.
$$

The crux is

$$
\text{finite VC dimension}
\Longrightarrow
\text{polynomial growth function}
\Longrightarrow
\text{vanishing uniform discrepancy}.
$$

Slide 2 displays the informal rate $O(\log\tau_{\mathcal H}(N)/\sqrt N)$; the formal theorem uses $\sqrt{\log\tau_{\mathcal H}(N)}/\sqrt N$. The formal statement below is the one proved.

**Theorem 6.11 (growth-function uniform-discrepancy bound).** For every distribution and $\delta\in(0,1)$, with probability at least $1-\delta$,

$$
\sup_{h\in\mathcal H}|L_{\mathbb P}(h)-L_S(h)|
\le
\frac{4+\sqrt{\log\tau_{\mathcal H}(2N)}}{\delta\sqrt{2N}}.
$$

If $d=\operatorname{VCdim}(\mathcal H)<\infty$, Sauer's lemma gives

$$
\tau_{\mathcal H}(2N)
\le\left(\frac{2eN}{d}\right)^d,
$$

so the displayed theorem implies

$$
N_{\mathcal H}^{\mathrm{UC}}(\varepsilon,\delta)
\le\widetilde O\left(\frac{d}{(\delta\varepsilon)^2}\right).
$$

**Slide exercise: derive this corollary.** Sauer's lemma first gives

$$
\log\tau_{\mathcal H}(2N)
\le
d\log\left(\frac{2eN}{d}\right).
$$

Substitution into Theorem 6.11 yields

$$
Z(S)
\le
\frac{4+\sqrt{d\log(2eN/d)}}{\delta\sqrt{2N}}.
$$

Choose $N$ so that the right-hand side is at most $\varepsilon$. Since $N$ also occurs inside the logarithm, solving the inequality gives the stated soft-order bound

$$
N=\widetilde O\left(\frac{d}{\delta^2\varepsilon^2}\right).
$$

This $\delta$ dependence is loose because the proof ends with Markov's inequality. Sharper concentration yields the optimal $\log(1/\delta)$ dependence in Section 22.1.


# 25. Restriction, shattering, and the set-system formulation

*Slide spine: L10, pp. 4-5.*

**Definition 6.2 (Restriction).** For $C=\{c_1,\ldots,c_m\}$,

$$
\mathcal H_C
=\{(h(c_1),\ldots,h(c_m)):h\in\mathcal H\}.
$$

The word “function” here can be read as a labeling of the points in $C$. **Definition 6.3 (Shattering)** says that the class shatters $C$ exactly when $|\mathcal H_C|=2^{|C|}$.

There is an equivalent set-system formulation. Associate each classifier with its positive region

$$
S_h=\{x:h(x)=1\},
\qquad
\mathcal F=\{S_h:h\in\mathcal H\}.
$$

Then $\mathcal H$ shatters $C$ if and only if

$$
\{C\cap S:S\in\mathcal F\}=2^C.
$$

Indeed, a binary labeling is uniquely determined by the subset of $C$ assigned label $1$. This answers the equivalence exercise on slide 4. The slide also uses the word “covering” informally for realizing every subset; it should not be confused with metric covering numbers.


# 26. Sauer-Shelah-Perles lemma

*Slide spine: L10, pp. 5-7.*

**Lemma 6.10 (Sauer--Shelah--Perles).** If $\operatorname{VCdim}(\mathcal H)\le d<\infty$, then

$$
\tau_{\mathcal H}(m)
\le
\sum_{i=0}^{d}\binom mi.
$$

For $1\le d<m$,

$$
\sum_{i=0}^{d}\binom mi
\le
\left(\frac{em}{d}\right)^d.
$$

Thus finite VC dimension forces polynomial rather than exponential growth. Infinite VC dimension gives $\tau_{\mathcal H}(m)=2^m$ for every $m$.

**Slide exercise: prove the second inequality.** Here is one standard proof. If $d\le m/2$, choose $x=d/(m-d)\le1$. Since $x^i\ge x^d$ for $i\le d$,

$$
x^d\sum_{i=0}^d\binom mi
\le
\sum_{i=0}^d\binom mi x^i
\le(1+x)^m.
$$

Consequently,

$$
\sum_{i=0}^d\binom mi
\le
\left(\frac md\right)^d
\left(\frac m{m-d}\right)^{m-d}
\le
\left(\frac{em}{d}\right)^d,
$$

where $(1+d/(m-d))^{m-d}\le e^d$. If $d>m/2$, use $\sum_{i=0}^d\binom mi\le2^m$ and

$$
2^m\le\left(\frac{em}{d}\right)^d;
$$

the latter follows by writing $r=d/m\in(1/2,1)$ and observing that $r(1-\log r)\ge\log2$.

The example

$$
\mathcal F=\{S\subseteq[m]:|S|\le d\}
$$

has size $\sum_{i=0}^d\binom mi$ and cannot shatter a $(d+1)$-point set, because the full $(d+1)$-point subset is absent. It shows the combinatorial bound is tight.

## 26.1 Supplement: recursive proof

Let $g_d(m)$ be the maximum number of labelings on $m$ points for a class of VC dimension at most $d$. Split the labelings on the first $m-1$ points into those that have one extension to the last point and those that have both extensions. The latter family has VC dimension at most $d-1$; otherwise adjoining the last point would shatter $d+1$ points. Hence

$$
g_d(m)\le g_d(m-1)+g_{d-1}(m-1).
$$

With $g_0(m)=g_d(0)=1$ and Pascal's identity,

$$
g_d(m)\le\sum_{i=0}^d\binom mi.
$$

The slide's historical notes are:

- 1968: publication by Vapnik and Chervonenkis, with proof in 1969;
- 1970: Erdős raised the question at the International Congress of Mathematicians;
- 1972: Sauer answered it in the *Journal of Combinatorial Theory (A)*;
- the humorous remark, “Nobody can find it in Shelah.”


# 27. From expected discrepancy to high probability

*Slide spine: L10, pp. 8-9.*

Define

$$
Z(S)=\sup_{h\in\mathcal H}|L_{\mathbb P}(h)-L_S(h)|.
$$

The proof first establishes

$$
\mathbb E_S[Z(S)]
\le
\frac{4+\sqrt{\log\tau_{\mathcal H}(2m)}}{\sqrt{2m}}.
$$

Since $Z\ge0$, Markov's inequality gives

$$
\mathbb P\left(Z>\frac{\mathbb E Z}{\delta}\right)\le\delta,
$$

which produces the theorem on slide 3. The slide attributes the result and proof plan to **Theorem 6.11** of [SSS].

The six proof steps listed on slide 9 are:

1. variable doubling with a ghost sample;
2. Jensen's inequality;
3. symmetrization with random signs;
4. restriction to the finite set appearing in two samples;
5. Hoeffding plus a union bound;
6. conversion of a probability tail into an expectation bound.


# 28. Step 1: variable doubling with a ghost sample

*Slide spine: L10, pp. 10-11.*

Let $S'=(Z_1',\ldots,Z_m')\sim\mathbb P^m$ be independent of $S$. For fixed $h$,

$$
L_{\mathbb P}(h)=\mathbb E_{S'}[L_{S'}(h)].
$$

Therefore

$$
\mathbb E_S\sup_h|L_{\mathbb P}(h)-L_S(h)|
=
\mathbb E_S\sup_h
\left|
\mathbb E_{S'}[L_{S'}(h)-L_S(h)]
\right|.
$$

The unknown population expectation has been replaced by an expectation over another empirical sample.


# 29. Step 2: Jensen and expansion of empirical risks

*Slide spine: L10, pp. 12-14.*

Using $|\mathbb EX|\le\mathbb E|X|$ and $\sup_h\mathbb E f_h\le\mathbb E\sup_h f_h$,

$$
\mathbb E_S\sup_h|L_{\mathbb P}(h)-L_S(h)|
\le
\mathbb E_{S,S'}\sup_h|L_{S'}(h)-L_S(h)|.
$$

Writing $z_i=(x_i,y_i)$ and $z_i'=(x_i',y_i')$ gives

$$
\mathbb E_{S,S'}
\left[
\sup_{h\in\mathcal H}
\left|
\frac1m\sum_{i=1}^m
\bigl(\ell(h,z_i')-\ell(h,z_i)\bigr)
\right|
\right].
$$

Slides 13 and 14 are proof-rhythm slides: the former restates the bound and the latter substitutes the empirical-risk definition.


# 30. Step 3: the symmetrization trick

*Slide spine: L10, pp. 15-16.*

Let $\sigma_1,\ldots,\sigma_m$ be independent Rademacher signs, uniformly distributed on $\{-1,+1\}$. Because each pair $(z_i,z_i')$ is exchangeable, swapping the pair changes the sign of

$$
\ell(h,z_i')-\ell(h,z_i)
$$

without changing its distribution. Consequently the previous expectation equals

$$
\mathbb E_{S,S',\sigma}
\left[
\sup_{h\in\mathcal H}
\left|
\frac1m\sum_{i=1}^m
\sigma_i\bigl(\ell(h,z_i')-\ell(h,z_i)\bigr)
\right|
\right].
$$

The dice graphic on the slides represents the random signs. This is the core mechanism behind Rademacher-complexity arguments.


# 31. Step 4: reduce the infinite class to a finite restriction

*Slide spine: L10, p. 17.*

Condition on $S,S'$ and let the set of **input points** be

$$
C_X=\{x_1,\ldots,x_m,x_1',\ldots,x_m'\},
\qquad |C_X|\le2m.
$$

The phrase “appearing in both samples” on the slide means the union of the two samples, not their intersection. Only predictions on $C_X$ matter: two hypotheses with the same restriction to $C_X$ produce the same losses on the fixed labeled samples. Let $\widetilde{\mathcal H}_{C_X}\subseteq\mathcal H$ contain one representative from each distinct restriction class. Then

$$
|\widetilde{\mathcal H}_{C_X}|
=|\mathcal H_{C_X}|
\le\tau_{\mathcal H}(2m),
$$

and the supremum over $\mathcal H$ becomes a maximum over these representatives. This distinction is important: $\mathcal H_{C_X}$ itself is a set of labeling vectors, whereas the representatives remain functions on which $\ell$ can be evaluated.

For $h\in\widetilde{\mathcal H}_{C_X}$, define

$$
\theta_h
=\frac1m\sum_{i=1}^m
\sigma_i\bigl(\ell(h,z_i')-\ell(h,z_i)\bigr).
$$

Conditional on $S,S'$, $\mathbb E_\sigma[\theta_h]=0$, and the summands are independent and bounded.


# 32. Step 5: Hoeffding and the union bound

*Slide spine: L10, p. 18.*

**Slide exercise: prove the Hoeffding step and the union-bound step.** Put

$$
a_i=\ell(h,z_i')-\ell(h,z_i),
\qquad |a_i|\le1.
$$

Conditional on $S,S'$, $\theta_h=m^{-1}\sum_i\sigma_i a_i$ is a mean-zero Rademacher sum. Hoeffding's inequality therefore has the general form

$$
\mathbb P_\sigma(|\theta_h|>\rho)
\le2e^{-c m\rho^2}
$$

for an absolute constant $c>0$. The standard range calculation $\sigma_i a_i/m\in[-|a_i|/m,|a_i|/m]$ gives $c=1/2$. The slide uses a different range normalization and explicitly states, for every $\rho>0$,

$$
\mathbb P_\sigma(|\theta_h|>\rho)
\le2e^{-2m\rho^2}.
$$

The exact constant does not affect the learning rate; the displayed $c=2$ is retained because it is the slide's formula. The bad maximum event is the union of the individual bad events, hence

$$
\begin{aligned}
&\mathbb P_\sigma\left(
\max_{h\in\widetilde{\mathcal H}_{C_X}}|\theta_h|>\rho
\right)\\
&\quad\le
\sum_{h\in\widetilde{\mathcal H}_{C_X}}
\mathbb P_\sigma(|\theta_h|>\rho)\\
&\quad\le
2|\widetilde{\mathcal H}_{C_X}|e^{-2m\rho^2}.
\end{aligned}
$$

Using the slide's shorthand $|\mathcal H_C|$ for the number of distinct restrictions, this is its displayed inequality written with the rigorous representative-class notation:

$$
\mathbb P_\sigma\left(
\max_{h\in\widetilde{\mathcal H}_{C_X}}|\theta_h|>\rho
\right)
\le2|\mathcal H_{C_X}|e^{-2m\rho^2}.
$$


# 33. Step 6: Massart or tail integration

*Slide spine: L10, p. 19.*

The finite-class maximal inequality used on the slide gives

$$
\mathbb E_\sigma
\left[
\max_{h\in\widetilde{\mathcal H}_{C_X}}|\theta_h|
\right]
\le
\frac{4+\sqrt{\log|\mathcal H_{C_X}|}}{\sqrt{2m}}.
$$

The same type of result follows from

$$
\mathbb EX=\int_0^\infty\mathbb P(X>t)\,dt
$$

and the preceding tail bound. Since $|C_X|\le2m$,

$$
|\mathcal H_{C_X}|\le\tau_{\mathcal H}(2m).
$$

Combining all six steps proves

$$
\boxed{
\mathbb E_S
\sup_{h\in\mathcal H}|L_{\mathbb P}(h)-L_S(h)|
\le
\frac{4+\sqrt{\log\tau_{\mathcal H}(2m)}}{\sqrt{2m}}
}.
$$

Markov's inequality then completes the high-probability theorem. Together with Sauer's lemma, this establishes $6\Rightarrow1$ in the fundamental theorem. The slide cites **Massart's lemma** and [SSS], Appendix A.4, for this tail-to-expectation step.

## 33.1 Supplement: Rademacher complexity and a sharper perspective

For a real-valued class $\mathcal F$, the empirical Rademacher complexity is

$$
\widehat{\mathfrak R}_S(\mathcal F)
=\mathbb E_\sigma
\left[
\sup_{f\in\mathcal F}
\frac1m\sum_{i=1}^m\sigma_i f(X_i)
\right].
$$

It measures the ability of the class to align with random noise. The L10 proof is a finite-labeling version of this method. Bounded-difference concentration applied after symmetrization produces the more standard confidence dependence

$$
\sup_{h\in\mathcal H}|L_{\mathbb P}(h)-L_S(h)|
=O\left(
\sqrt{\frac{d+\log(1/\delta)}{m}}
\right),
$$

up to the exact theorem and constants being used.


# 34. Generalization through algorithmic stability

*Slide spine: L10, pp. 20-21.*

Uniform convergence controls an entire hypothesis class and can be pessimistic. Algorithmic stability instead asks how much a particular algorithm changes when its input data are slightly perturbed.

The slides list four perturbations:

1. resample one point;
2. leave one point out;
3. corrupt one point arbitrarily;
4. add noise to the algorithm itself.

If the output is insensitive to such changes, it is less likely to have memorized an idiosyncratic training point, so the generalization gap should be small.

**Slide Explore prompt.** Compare how these perturbations affect stability and generalization. Resampling measures replacement stability and matches the hybrid-sample proof below; leaving one out measures deletion stability; arbitrary corruption asks for adversarial or worst-case robustness; and internal algorithmic noise makes stability a property of a randomized learner. Each can be studied in expectation or in the worst case, so the corresponding guarantees have different strengths.


# 35. Stability setup and hybrid samples

*Slide spine: L10, p. 22.*

Let

$$
S=(z_1,\ldots,z_n),
\qquad
S'=(z_1',\ldots,z_n')
$$

be independent samples. View an algorithm as

$$
A:\mathcal Z^n\to\mathcal H.
$$

The slide writes $A:\mathbb P^n\to\mathcal H$; the standard type above distinguishes the sample space from its distribution.

For each $i$, form the hybrid sample

$$
S^i=(z_1,\ldots,z_{i-1},z_i',z_{i+1},\ldots,z_n).
$$

This is the same ghost-sample idea as before, now used to compare algorithm outputs.


# 36. Average stability equals the expected generalization gap

*Slide spine: L10, pp. 23-24.*

For $z=(x,y)$, the slide first abbreviates

$$
\ell(f,z)=\ell(f(x),y).
$$

It labels the following notion **average stability (data dependent)**. Define it by

$$
\Delta(A)
=\mathbb E_{S,S'}
\left[
\frac1n\sum_{i=1}^n
\left(
\ell(A(S),z_i')-
\ell(A(S^i),z_i')
\right)
\right].
$$

Here $z_i'$ is unseen by $A(S)$ but seen by $A(S^i)$. Thus $\Delta(A)$ is both an average one-point sensitivity and an average seen-versus-unseen loss difference.

The proposition is the exact identity

$$
\boxed{
\mathbb E_S
\left[
L_{\mathbb P}(A(S))-L_S(A(S))
\right]
=\Delta(A)
}.
$$

Proof:

$$
\begin{aligned}
&\mathbb E[L_{\mathbb P}(A(S))-L_S(A(S))]\\
&=
\mathbb E\left[\frac1n\sum_i\ell(A(S),z_i')\right]
-
\mathbb E\left[\frac1n\sum_i\ell(A(S),z_i)\right].
\end{aligned}
$$

Exchangeability of $z_i$ and $z_i'$ gives

$$
\mathbb E[\ell(A(S),z_i)]
=\mathbb E[\ell(A(S^i),z_i')].
$$

Substitution yields $\Delta(A)$. The slides call the left-hand side “expected excess risk” and immediately parenthesize “generalization gap.” Strictly, excess risk usually means $L_{\mathbb P}(h)-\inf_{g\in\mathcal H}L_{\mathbb P}(g)$; the identity here is about the expected generalization gap.


# 37. Uniform stability

*Slide spine: L10, p. 25.*

Let $d_H(S,S')$ be Hamming distance between samples. Uniform stability is the worst-case loss change when exactly one example is replaced:

$$
\Delta_{\sup}(A)
=
\sup_{\substack{S,S'\\d_H(S,S')=1}}
\sup_{z}
\left|
\ell(A(S),z)-\ell(A(S'),z)
\right|.
$$

The slide writes the second supremum as $z\sim\mathbb P$; the standard reading is a supremum over test points in the support. Since

$$
|\Delta(A)|\le\Delta_{\sup}(A),
$$

uniform stability controls the expected generalization gap as well.

**Supplement.** Strong convexity created by regularization often makes regularized ERM uniformly stable. This is an algorithm-dependent explanation that can be far sharper than a worst-case bound for every $h\in\mathcal H$.

# 41. Final proof map

The complete theory can be remembered through four routes.

**Finite realizable class:**

$$
\text{bad hypothesis fits all }n\text{ examples}
\Longrightarrow e^{-n\varepsilon}
\xRightarrow{\text{union bound}}
|\mathcal H|e^{-n\varepsilon}.
$$

**VC route:**

$$
\operatorname{VCdim}(\mathcal H)=d
\xRightarrow{\text{Sauer}}
\tau_{\mathcal H}(m)\le(em/d)^d
\Longrightarrow\text{uniform convergence}
\Longrightarrow\text{agnostic ERM}.
$$

**Rademacher route:**

$$
\text{ghost sample}
\Longrightarrow\text{symmetrization}
\Longrightarrow\text{random-sign supremum}
\Longrightarrow\text{generalization bound}.
$$

**Stability route:**

$$
S\text{ changes in one point}
\Longrightarrow A(S)\text{ changes little}
\Longrightarrow
\mathbb E[L_{\mathbb P}(A(S))-L_S(A(S))]\text{ is small}.
$$

Together, the slides show two complementary explanations of generalization: control the effective capacity of the entire hypothesis class, or control the sensitivity of the particular learning algorithm.
